### Problem 3: Exercise 

Part A: Since energy is conserved, $E_1$, the energy when speed is $v_1$ and distance is $l_1$, can be set equal to $E_2$, the energy when speed is $v_2$ and distance is $l_2$, yielding
$$\frac{1}{2} mv_1^2 - G\frac{mM}{l_1} = \frac{1}{2} mv_2^2 - G\frac{mM}{l_2}$$

We also know that at aphelion, $l_2 v_2 = l_1 v_1$. Rearranging the first equation and substituting $\frac{l_1 v_1}{v_2}$ for $l_2$ yields
$$v_2^2-\frac{2GM}{v_1 l_1}v_2-\left(v_1^2-\frac{2GM}{l_1}\right)=0$$
the roots of which can be found using the quadratic formula where $a = 1$, $b = \frac{2GM}{v_1 l_1}$, and $c = -\left(v_1^2-\frac{2GM}{l_1}\right)$. Solving the quadratic formula yields
$$v_2 = \frac{GM - \sqrt{G^2 M^2 +v_1^4 l_1^2 +2GMv_1^2 l_1}}{v_1 l_1}$$
the other larger root is $v_1$.
Using $l_2 = \frac{l_1 v_1}{v_2}$, $l_2$ is calculated to be
$$l_2 = \frac{v_1^2 l_1^2}{GM - \sqrt{G^2 M^2 +v_1^4 l_1^2 +2GMv_1^2 l_1}}$$

```python
#allow reasonable user entries and then convert to actual values
peri_dist_Gig = float(input("Enter the distance from the Sun at perihelion in Gigameters (10^9 meters)")) #l_1 in Gm
peri_speed_kms = float(input("Enter the speed at perihelion in km/s (10^3 m/s)")) #v_1 in km/s
peri_dist = peri_dist_Gig*(10**9) #l_1 in m
peri_speed = peri_speed_kms*1000 #v_1 in m/s

#defining constants and calculating values
G = 6.6738*(10**(-11)) #m^3/(kg s^2)
M = 1.9891*(10**30) #kg
grav = 2*G*M/peri_dist #commonly used variables combined to simplify equation
ap_speed = (grav - np.sqrt((grav**2)-4*(peri_speed**2)*((-peri_speed**2)+grav)))/(2*peri_speed) #v_2 in m/s
ap_dist = (peri_dist*peri_speed)/ap_speed #l_2 in meters
semi_major = (1/2)*(peri_dist+ap_dist)  #a in meters
semi_minor = np.sqrt(peri_dist*ap_dist) #b in meters
period = (2*math.pi*semi_major*semi_minor)/(peri_speed*peri_dist) #T in seconds
ecc = (ap_dist-peri_dist)/(ap_dist+peri_dist)

#converting to useful units
period_years = period/60/60/24/365
ap_dist_Gig = ap_dist/(10**9)
ap_speed_kms = ap_speed/1000

print("The distance to the Sun at aphelion is", ap_dist_Gig,"Gm.")
print("The velocity at aphelion is", ap_speed_kms, "km/s.")
print("The orbital period is approximately", period_years, "years.")
print("The eccentricity of the orbit is", ecc,".")
```

#### Explanation: 

In order to assure the user can enter reasonable values between 1 and 1000, I ask for the distance at perihelion in Gm and the speed in km/s, and then convert the values into m and m/s. I define G, M, and a variable "grav" which is 2GM divided by the distance at perihelion. Using these three variables I calculate the speed and the distance from the sun at aphelion. These values are then used to calculate the semi-major and semi-minor axes, the period and the eccentricity. Period is in seconds, so I converted it to years. I also converted the distance to Gm and the speed to km/s for easy reading.
