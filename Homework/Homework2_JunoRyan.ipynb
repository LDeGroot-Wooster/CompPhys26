```python
def CtoF(C_temp):
    F_temp = [0] * len(C_temp)  #list of zeroes of same length as input list
    for i in range(len(C_temp)):
        F_temp[i] = (C_temp[i]*9/5)+32  #translates each temperature into fahrenheit
    return F_temp

print(CtoF([5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75]))
```

Part A: Since energy is conserved, $E_1$, the energy when speed is $v_1$ and distance is $l_1$, can be set equal to $E_2$, the energy when speed is $v_2$ and distance is $l_2$, yielding
$$\frac{1}{2} mv_1^2 - G\frac{mM}{l_1} = \frac{1}{2} mv_2^2 - G\frac{mM}{l_2}$$

We also know that at aphelion, $l_2 v_2 = l_1 v_1$. Rearranging the first equation and substituting $\frac{l_1 v_1}{v_2}$ for $l_2$ yields
$$v_2^2-\frac{2GM}{v_1 l_1}v_2-\left(v_1^2-\frac{2GM}{l_1}\right)=0$$
the roots of which can be found using the quadratic formula where $a = 1$, $b = \frac{2GM}{v_1 l_1}$, and $c = -\left(v_1^2-\frac{2GM}{l_1}\right)$. Solving the quadratic formula yields
$$v_2 = \frac{GM - \sqrt{G^2 M^2 +v_1^4 l_1^2 +2GMv_1^2 l_1}}{v_1 l_1}$$
the other larger root is $v_1$.
Using $l_2 = \frac{l_1 v_1}{v_2}$, $l_2$ is calculated to be
$$l_2 = \frac{v_1^2 l_1^2}{GM - \sqrt{G^2 M^2 +v_1^4 l_1^2 +2GMv_1^2 l_1}}$$

```python
prime_list = [2]

def isPrime(n,primes):   #returns 0 if not prime, returns value if prime
        for j in primes:
            if j > np.sqrt(n):
                return n
            if n%j == 0:  
                return 0

#checks if output of function above is 0 for all values in range, if not, appends output to prime list
for i in range(3,10001):
    n = isPrime(i,prime_list)
    if n!=0:
        prime_list.append(n)

print(prime_list)
```

#### Explanation: 

I begin with a list of primes containing 2, the lowest prime number. My function, isPrime, takes an input of a number and a list. For each value in the list, it first checks if the value is greater than the square root of the given number. Since all prime factors of a number must be less than it's square root, if the lowest value in the primes list is greater than the number's square root, the number must be prime. If this is the case, the function returns the given number. If not, the function checks if the given number is divisible by the list value. If this is true the number cannot be prime and the function returns 0. If neither of these cases are true, the function checks the next value on the list and so on until it reaches a value that is greater than the square root of the given number (in which case the number is prime) or it finds a prime factor of the number (in which case it is not prime).

To implement this function, I use a for loop going from 3 to 10001. At each value, I run the function and check whether the it returns a non-zero value. If it does, I append it to the prime list so it can be used to be check future values.

```python
#plots sunspot data
data = np.loadtxt("sunspots.txt",float)
months = data[:1000,0]
spots = data[:1000,1]
plt.scatter(months,spots,s=6,c="g")
plt.xlabel("Months since January 1749")
plt.ylabel("Number of Sunspots")
plt.title("Number of Sunspots Observed on the Sun over Time")

#empty list to be filled with runnign average data
avg_spots = []
r=5
def run_avg(month):
    if month < r or month > 1000 - (r+1): #checks if running average possible. If not, returns 0
        return 0
    else:                           #performs running average calculation by summing five values
        total_spots = 0             #below and above, plus given value, and dividing by number
        for i in range(-r,r+1):     #of values
            total_spots+=spots[month+i]
        avg_spots = total_spots/(2*r+1)
        return avg_spots

#populates list and plots data
for month in range(len(months)):
    avg_spots.append(run_avg(month))
plt.plot(months,avg_spots)
plt.show
```

#### Explanation: 

First I plotted the first 1000 points (up to but not including index 1000) of the data as a nice labeled scatter plot. Then I defined a function to calculate the runnign average. The average should be calculated over 11 values, five below and five above, so the function first checks whether there are enough values below and above to calculate properly. If not, it returns 0. Then, the calculation is performed using a for loop to sum the 11 values and then dividing by 11. I create an empty list for the data and then use a for loop that calls the function on each of the first 1000 months to populate the list. I then plot the data.
