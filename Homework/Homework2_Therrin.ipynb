```python
a1 = 15.8; a2 = 18.3; a3 = 0.714; a4 = 23.2

def Binding_energy_func(A,Z):
    if A%2==0:
            if Z%2==0:
                    a5=12
            else:   a5=-12
    else:           a5=0
    return a1*A - a2*(A**(2/3))-a3*((Z**2)/(A**(1/3)))-a4*(((A-2*Z)**2)/(A))+a5/(A**(1/2))


A_input = int(input("A value: "))
Z_input = int(input("Z value: "))



print("Binding energy in MeV:",Binding_energy_func(A_input,Z_input))
print("Binding energy in MeV per Nucleon:", Binding_energy_func(A_input,Z_input)/A_input)

```

```python
Z_input = int(input("Z value: "))
A_list = np.arange(Z_input,3*Z_input+1)
BperA_list = []
for i in range(np.size(A_list)):
    BperA_list.append(
        Binding_energy_func(A_list[i],Z_input) / A_list[i]
        )
print("A that maximimizes binding energy is:", A_list[np.argmax(BperA_list)])

```

### Problem 4: Exercise 

```python
Sunspots_data = np.loadtxt("sunspots.txt")
Time_array = Sunspots_data[:,0]
Count_array = Sunspots_data[:,1]

Time_1000_array = Time_array[:1000]
Count_1000_array = Count_array[:1000]

bounds = 5
Average = np.zeros(len(Count_1000_array)) / (2*bounds)
for i in range(bounds, len(Count_1000_array) - bounds):
    Average[i] = np.sum(Count_1000_array[i-bounds:i+bounds])
    
plt.figure()
plt.plot(Time_1000_array, Count_1000_array, label="Sunspot number")
plt.plot(Time_1000_array, Average, label="Running average bounds: 5")
plt.xlabel("Month")
plt.ylabel("Number of sunspots")
plt.legend()
plt.show()
```

#### Explanation:
