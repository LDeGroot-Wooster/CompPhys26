```python
# function to return the electronic wave-function
def elec_wave_func(x, y, z):
    r = np.sqrt(x**2 + y**2 + z**2)
    answer = z*(2-r)*np.exp(-r)
    return answer

#part b

#Use your function to make a density plot of the probability density abs(\Psi)^2 \
# of the electron in the xz plane for values of x and z between -2 and 2

# values of the xz plane
x = np.linspace(-2, 2, 300)
z = np.linspace(-2, 2, 300)

# making arrays of the electronic wave-function
xz_waves = [] #starting them as lists, will convert them later

for i in range(len(x)+1):
    xz_waves.append(elec_wave_func(x,0,z))

array = np.array(xz_waves)
array_density = array ** 2


plt.imshow(array_density, origin='lower')
plt.viridis()
plt.title("Electron Probability Density in the xz-plane")
plt.xlabel("z")
plt.ylabel("x")
plt.colorbar(label='Density $|\Psi|^2$')
plt.show()
```
