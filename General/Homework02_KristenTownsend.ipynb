```python
# set-up inputs
#A = 58 #float(input("Input mass number A:"))
#Z = 28 #float(input("Input atomic number Z:"))

# establishing constants
a_1 = 15.8
a_2 = 18.3
a_3 = 0.714
a_4 = 23.2

# calculate binding energy
def calc_B(A, Z):
    # determining a_5
    if A%2 == 1:
        a_5 = 0
    else:
        if Z%2 == 0:
            a_5 = 12.0
        else:
            a_5 = -12.0
    # the weird equation
    B = a_1*A - a_2*(A**(2/3)) - a_3*((Z**2)/(A**(1/3))) - a_4*(((A-2*Z)**2) /A) + a_5/(A**(1/2))
    return B

def find_stable(Z):
    # initialize
    A = Z
    B_max = 0
    stable_A = 0

    while A <= 3*Z:
        # calculate binding energy per nucleon
        new_B_per_nucleon = calc_B(A, Z) / A
        
        # keep track of the strongest bind per nucleon
        if new_B_per_nucleon >= B_max:
            B_max = new_B_per_nucleon
            stable_A = A
        A += 1
    return(stable_A, B_max)

# for part d, iterating over values of Z
max_bind = 0

for Z in range(1, 101):
    loop_A, B_per_A = find_stable(Z)
    # print results
    print("The most stable atomic mass number A for atomic number ", Z," is ", loop_A, \
    ", with a binding energy per nucleon value of ", B_per_A, " MeV.", sep="")
    # keep track of best bind results for the Z value range
    if B_per_A >= max_bind:
        max_bind = B_per_A
        best_Z = Z
# print the stringest bind
print("The best bind per nucleon is ", max_bind, " MeV at atomic number ", best_Z, ".", sep="")
```

### Problem 2: Conversion Function

```python
# temps to be converted
C_temp = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75]

# the conversion
def temp_convert(celsius):
    F_temp = []
    for i in range( len(celsius) ):
        C = celsius[i]
        F = (9/5) * C + 32
        F_temp.append(F)
    return F_temp

# show and tell
print(temp_convert(C_temp))
```

### Problem 3: Exercise 2.11, p. 80

```python
# starting list of primes
primes = [2]

for num in range(3,10001):
    # innocent until proven guilty moment
    prime = True
    #check until sqrt(num) to save computing
    for i in range(2, int(math.sqrt(num)) + 1):
        #does this divide evenly? If so, it's not prime!
        if num % i == 0:
            prime = False
    #numbers that pass the if statement end up here
    if prime:
       primes.append(num)
       #print (num)
                
print(primes)        

```

```python
# part (c)
sun_data = np.loadtxt("sunspots.txt", float)

# extract data, only first 1000
N = 1000 #limits how many points are graphed
time = sun_data[:N,0] #months
num_spots = sun_data[:N,1] # number of sunspots

#plot it
plt.plot(time, num_spots) #takes x,y as an argument

# finding the running average
r = 5 # index to each side for average

#initialize array for running average
avg_sunspots = np.zeros(N, float)

# start and stop point for indexing
start_index = -r
end_index = r + 1

for i in range(0,len(num_spots)):

    # slice the list
    sub_list = num_spots[start_index:end_index]

    # Calculate the average sunspot
    avg = np.mean(sub_list)

    # plug into array of average sunspots
    avg_sunspots[i] = avg

    # update to next slot
    start_index += 1
    end_index += 1

# plot it
plt.plot(time, avg_sunspots, c="r")

plt.xlabel("time (months)")
plt.ylabel("number of sunspots")
plt.title("Sunspots over time")

plt.show()
```

Running average!
