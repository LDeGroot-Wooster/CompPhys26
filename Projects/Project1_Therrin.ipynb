```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
coordinate_0 = np.array((0,float(input())))
while coordinate_0[1]<0:
    coordinate_0 = np.array((0,float(input())))
    
angle_0 = float(input())
while angle_0>np.pi/2 or angle_0<0:
    angle_0 = float(input())

velocity_0 = float(input())*np.array((np.cos(angle_0),np.sin(angle_0)))
gravity = 9.8

time_density = 100

X_in_time = []
Y_in_time = []
coordinate = coordinate_0
velocity = velocity_0/time_density
dv = 2*np.array((0,-gravity))/(time_density**2)

idx = 0
while coordinate[1]>0:
    
    X_in_time.append(coordinate[0]) 
    Y_in_time.append(coordinate[1]) 
    
    coordinate += velocity + dv/2
    velocity += dv
    idx+=1

X_in_time = np.array(X_in_time)
Y_in_time = np.array(Y_in_time)

plt.grid()
plt.plot(X_in_time,Y_in_time, "ro")
plt.xlabel("Distance from start (m)")
plt.ylabel("Height (m)")
plt.show()
```
