```python
def factorial_int(n): 
    result = 1 # by listing first value as an integer, sets type for rest of problem 
    for k in range(1,n+1):
        result *= k
    return result

def factorial_float(n): 
    result = 1.0 # by defining first value as a float, sets type for rest of program
    for k in range(1,n+1):
        result *= k
    return result

# ask for user input
number = int(input("Please provide a number to take the factorial of: "))

# integers
print(factorial_int(number))

# floats
print(factorial_float(number))
```

####  Calculating a derivative (adapted from textbook)

Using the definition of a derivative below we can use a computer take the derivative of a function $f(x)$ at a point $x$:
$$ \frac{df}{dx} = \lim_{h \to 0} \frac{f(x+h)-f(x)}{h}$$
Actually taking $h$ to zero is not reasonable, but we can at least make $h$ small.


Write a code to do the following:
- Define a function `f1(x)` that returns the value of $x(x-1)$ [this should only take you 2 lines]
- Define a second function `f_deriv(x, h)` that takes in an x and h value and uses your `f1(x)` function to calculate the derivative of the function at the point $x = 1$ using the formula above and $h = 10^{-2}$. Have your function return both the result of the derivative and h (in that order). 
- Put the result in a markdown cell and comment on the accuracy by calculating the actual derivative at $x=1$ yourself (you can do this on your own, I know you can!). 
- Repeat the calculation for $h = 10^{-4}, 10^{-6}, 10^{-8}, 10^{-10}, 10^{-12}, 10^{-14}$. You should see an increase in accuracy and then a decrease. In the markdown cell, include all the results and comment on why this occurs. 

When your program is complete, add a markdown cell below it to explain your process and program and include any values/comparisons. 

```python
# define function that want to take derivative of 
def f1(x):
    return x*(x-1)

def f_deriv(x, h): 
    # define constants
    h = 1e-14
    # use function and constants along with derivative equation to calculate derivative 
    df = (f1(x+h)-f1(x))/h
    return df, h

print(f_deriv(1,1e-14))
```

(b) As we can see, the estimates do indeed get better then worst again. They get better initailly because smaller $h$ gives a better approximation to the true definition of a derivative. But then they get worst again because in the numerator we are subtracting two numbers that are nearly equal and that produces a large fractional error in the result. The most accurate result is achieved when $h = 10^{-8}$, which is not a coincidence: $10^{-8}$ is the square root of the machine precision or error constant. As explained in the book in Chapter 5, there is a good reason to pick exactly this value for $h$ when calculating derivatives.

####  Calculating integrals (adapted from textbook)

Like you did with a derivative, we can also use the definition of an integral to evaluate integrals with a computer. If we want to evaluate the integral 
 

$$ I = \int_{-1}^{1} \sqrt{1-x^2} \,dx, $$


we can divide the domain of the integration in to a larger number $N$ of slices of width $h = 2/N$ each. The Riemann definition can then be applied: 
$$ I = \lim_{N \to\infty} \sum_{k=1}^{N} hy_k, $$
where 
$$ y_k = \sqrt{1-x_k^2} \hspace{0.2in} \text{and}\hspace{0.2in} x_k = -1+hk.$$


We just need to make $N$ large to get a good approximation (since we can't in practice take the limit as $N\rightarrow \infty$). The expected value of this integral is $\pi/2$. 

Write a code to do the following: 
- Define a function `f2(x)` that returns the value of $\sqrt{1-x^2}.$  [this should only take you 2 lines]
- Define a second function `f_int(N)` that takes in in an N value and uses your `f2(x)` function to evaluate the integral. Have your function return both the result of the derivative and N (in that order).
- $N = 100$ and compare with the expected result. [perhaps a percent error makes a good comparison!]
- Increase the value of $N$ to make it more accurate. If the program must run in 1 second or less, how accurate can you get? 


When your program is complete, add a markdown cell below it to explain your process and program and include any comparisions. 

```python
start = time.time()
# define function
def f2(x):
    return np.sqrt(1-x**2)

def f_int(N): 
    # define constants
    h = 2/N
    I = 0.0

    # calculate the integral of the function using equation
    for k in range(N):
        I += h*f2(h*k-1)

    return I, N # return result of integral and N used 

end = time.time()
print("time taken (in s):", end-start) # calculate time taken    

# print result
#print(f_int(1000))
result = f_int(2000000)
print(result[0], ",",result[1])

print(np.pi/2) # print to compare with result 
```
