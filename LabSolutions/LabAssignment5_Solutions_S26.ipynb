### Write a script 1 (25 pts)
#### Making a density plot of a function

Write a code to do the following: 
- Create a 2D array that shows how the values change linearly from the center outwards based on the equation: 
$$ B = 7.5r - 2.8 $$
    - the overall grid should be 200 x 200 units 
    - you should use a 1000x1000 array
- Plot the array as a density plot, change the color scale to one of your choice, and include a color bar

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
import numpy as np
import matplotlib.pyplot as plt

# This is really similar to the wave interference example from class
# set the total number of points that want, but then scale to 200 x 200
n_points = 1000
side_size = 200

# set the central position of the function as the center of the arrays 
x1 = side_size/2.0
y1 = side_size/2.0

# relate the number of points to the size of the grid 
spacing = side_size/n_points

# create empty numpy array to insert B values into
B = np.zeros([n_points,n_points], float)

# use for loop to apply B calculation to every element in array
# make sure to calculate r from the x and y array elements & considering the central x1 and y1 position 
for i in range(n_points):
    y = spacing*i
    for j in range(n_points):
        x = spacing*j
        r = np.sqrt(((x-x1)**2+(y-y1)**2))
        B[i,j] =7.5*r - 2.8
        
# make the plot
plt.imshow(B,origin="lower", extent=[0,side_size,0,side_size])
plt.jet()
plt.colorbar()
plt.show()
```

```python
# another way of interpreting the problem 

import numpy as np
import matplotlib.pyplot as plt
n_points = 1000
x1 = n_points/2.0
y1 = n_points/2.0
spacing = 1
side_size = 200
B = np.zeros([n_points,n_points], float)

for i in range(n_points):
    y = spacing*i
    for j in range(n_points):
        x = spacing*j
        r = np.sqrt(((x-x1)**2+(y-y1)**2))
        B[i,j] =7.5*r - 2.8
        
# make the plot
# simply use extent to set grid. Notice slight differences in the calculated values in color bar
plt.imshow(B,origin="lower", extent=[0,side_size,0,side_size])
plt.jet()
plt.colorbar()
plt.show()
```
