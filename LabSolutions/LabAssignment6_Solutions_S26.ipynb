### Animation #1 (22 pts)
#### A Growing Cone 

- Create a cone of some initial height and radius
- Make it grow to 5 times it’s size  
- Use the appropriate rate to make this occur over 10 seconds
- Make the cone a color of your choice 

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
scene1 = canvas(width=800, height=500, autoscale=False, center=vector(5,0,0), range=10)

# define number of steps and total run time
nsteps = 50
total_time = 10 #seconds

# calculate rate value 
time_between = int(nsteps/total_time) #equivalent to `int(1/(total_time/nsteps))`
#print(time_between)

# define starting radius and height as well as how much increases each step
rad = 0.5
rad_inc = 0.2
height = 0.5
height_inc = 0.2

# define starting cone
c=cone(pos=vector(5,-2,0), axis=vector(0,height,0), radius = rad, color=color.red)

# loop through to increase both radius and height with time
for i in range(nsteps):
    rate(time_between) # tells computer to wait 1/10th of a second has passed since last time rate() was called
    rad=rad+rad_inc
    height = height+height_inc
    c.radius=rad
    c.axis = vector(0,height,0)
```
