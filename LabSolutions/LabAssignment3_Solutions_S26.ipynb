### Write a script 2 (25 pts)
#### Find the sum of odds

Write a code to add all the odd numbers from 1 to $N$. 
- The function should return the sum *and* the time in milliseconds needed to perform this sum. (Google is your friend!)
- Use your function to report the sum of all the odd integers from 1 to 1,000,000 as well as how long it took to do this.

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
# define a function to calculate sum of odd values 
def odd_sum(N):
    #determine start time of function in seconds - did have to add import time to top 
    start_time = time.time()
    # define sum currently equal to zero and then loop through N to see if values odd
    # if so, add them to the sum 
    sum_val = 0
    for i in range(N):
        if i%2 == 1:
            sum_val += i
    #calculate end time and take different to find total run time 
    end_time = time.time()
    elapsed_time = (end_time-start_time)*1000 # multiply by 1000 to convert to milliseconds 
    return sum_val, elapsed_time

# call on function 
calc_sum, elapsed_time = odd_sum(1000000)

# print out sum
print("The sum of integers from 1 to 1,000,000 is", calc_sum)
# print out elapsed time 
print("Total run time is", elapsed_time, "milliseconds")
```

**Explanation** I had to find a way to calculate time, which I found the time module through a google search. By creating a variable `start_time` at the beginning of the function and `end_time` after the completion, the different gives the time elapsed in secons. 

To determine if a value is odd, I used the modulo `%`. If taking the modulo of the number with respect to 2 is 1, then the number must be odd (since a remainder of 1 is left after division by 2). If that is the case, I added that value to the sum. 

I had my function return both the elapseded time and the sum value. I double-checked this for a value I could hand calculate before running it on 1,000,000

The equation for standard deviation is given by: 
$$ \sigma = \left[ \frac{1}{N} \sum\limits_{i = 1}^{N} \left( a_i -\overline{a} \right)^2 \right]^{1/2} $$
where $\overline{a}$ is the average value of the array, and $a_i$ is each element of the array of length $N$. 
In the cell below, write a function to calculate the standard deviation for an input array, a:
- First compute the average of the elements in a to define $\overline{a}$ [hint: there is a function called np.mean()] **I have provided the array a for you - be sure to understand what it does**
- Next compute the sum over the squares of $a - \overline{a}$
- Then divide the sum by the number of elements in the array
- Finally take the square root (you can use np.sqrt())
Test your function on a random array, and compare to the built-in np.std() function

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
def std_dev(arr):
    # takes in an array and returns the standard deviation 
    
    # calculate the average of the array - part (a)
    a_bar = np.mean(arr)
    #print(a_bar)
    
    # take the sum of the difference between each array value and the average - part (b)
    sum_square = np.sum((arr - a_bar)**2)
    #print(sum_square)
    
    # divide sum by number of elements in arr - part (C)
    sum_square_div = sum_square/len(arr)
    #print(sum_square_div)

    # take the square root - part (d)
    std_dev = np.sqrt(sum_square_div)
    
    return std_dev


# first create input array a
a = np.random.random_sample(20)*20
print(a)

print(std_dev(a))

# part (e)
print(np.std(a))
```
