### Write a script 1 (25 pts) - Lists 

Let's build and analyze a list of measurements:
- Create an empty list called "meas_list"
- Ask the user 10 times for a number between 1 and 10 & append to the list 
    - Print the length of the list to make sure it is correct! 
- Calculate the average, maximum, and minimum of the list & print these values
- Assuming all of these are values of time, convert the values into distance (in meters) by multiplying by the constant velocity the object is traveling (55 mph) & print your results

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
# first create empty list
meas_list = []
print(meas_list)

# ask user for numbers to add to list
while len(meas_list) < 10:
    new_num = float(input("Please give a number between 1 and 10: "))
    if new_num > 0 and new_num <= 10: 
        meas_list.append(new_num)
    else: 
        print("the number must be between 1 and 10. Try again!")
print(meas_list)

# print length of list
print("the length of the list is ", len(meas_list))
    
# Convert to distance
v_mph = 55 # need to convert to m/s
v_m = v_mph*1609.34/3600

# Calculate stats
list_avg = sum(meas_list)/len(meas_list)
list_max = max(meas_list)
list_min = min(meas_list)

#print
print("The average, max, and min values of the list are:", list_avg, list_max, list_min)

#calculate distances
dist_avg = list_avg*v_m
dist_max = list_max*v_m
dist_min = list_min*v_m
print("The average, max, and min values of distance are:", dist_avg, ",", dist_max, ",", dist_min, "meters")
```

```python
g = 9.81

#define height array 
height = np.arange(0,50.5,0.5)
print(height)

# input a mass
mass = 5.0 # kg 

# copy height array to create potential and kinetic arrays
potential = np.copy(height)
kinetic = np.copy(height)

#calculate potential energy
potential = potential*mass*g
print(potential)

#calculate kinetic energy 
kinetic = potential[-1]-kinetic*mass*g
print(kinetic)


# yes this makes sense. When the kinetic is max, the potential is zero. When the kinetic is 0, the potential is max
```
