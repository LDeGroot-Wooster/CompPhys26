#### We can create a collection of spheres 

```python
scene3 = canvas(background=color.black)
Count = 5
Radius = 0.3
for i in range(-(Count)//2,+(Count)//2):
    for j in range(-(Count)//2,+(Count)//2):
        for k in range (-(Count)//2,+(Count)//2):
            sphere(pos=vector(i/(2*Radius),j/(2*Radius),k/(2*Radius)), radi=Radius, color=color.white)
```

#### You can also store 3D shapes in an array 

```python
shapes_array = np.empty(100,sphere)
scene4 = canvas(range=10)
n = 0
for i in np.linspace(0,9,100):
    shapes_array[n] = sphere(pos=vector(5*cos(i**2),5*sin(2*sqrt(i)),0))
    n+=1
```

#### Looping through position change
- Create a loop to change the position of the sphere 50 times 


```python
scene7=canvas(width= 200, height = 200, autscale=False)
xpos = 1
sphere_shape = sphere(pos=vec(xpos,0,0))

for i in range(50):
    rate(10)
    if xpos%2==0:
        xpos/2
    else: 5*xpos+3
    sphere_shape.pos = vector(xpos,0,0)
    
```

#### Example: A Moving Sphere

A program to move a sphere around on the screen in a uniform circle

```python
scene8 = canvas(width= 200, height = 200, autscale=False)
xpos = 1
ypos = 0

s = sphere(pos=vec(xpos,ypos,0),radi=0.1,color=vec(1,0.5,0.1))

for i in range(100):
    rate(10)
    xpos = cos(2*pi*i/50)
    ypos = sin(2*pi*i/50)
    sphere_shape.pos = vector(xpos,0,0)
```

How could I make the sphere get larger? how could I make it rotate? 

```python
scene8 = canvas(width= 200, height = 200, autscale=False)
xpos = 1
ypos = 0

s = sphere(pos=vec(xpos,ypos,0),radi=0.1,color=vec(1,0.5,0.1))

for i in range(100):
    rate(10)
    xpos = 10*cos(2*np.pi*i/50)
    ypos = 10*sin(2*np.pi*i/50)
    s.pos = vector(xpos,0,0)
    s.radi = 0.1 * cos(2*np.pi*i/50)
```
