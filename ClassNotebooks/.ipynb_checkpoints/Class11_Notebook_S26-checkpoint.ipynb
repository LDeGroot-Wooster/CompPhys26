### Forward, Backward, and Central difference 

Evaluate the derivative of $f(x) = e^{\sin(2x)}$ at $x = 0.5$

The derivative is $2\cos{(2x)}e^{\sin{(2x)}}$ which equals approximately 2.507.

```python
def f_prime(x):
    return 2*np.cos(2*x)*math.e**(np.sin(2*x))

true_val = f_prime(0.5)
print(true_val)
```

Now apply forward/backward difference 

```python
def f(x):
    return math.e**(np.sin(2*x))
x = 0.5

def fwd_diff(f,x,h):
    return (f(x+h)-f(x))/h

def bkwd_diff(f,x,h):
    return (f(x)-f(x-h))/h

h_fwd = []
h_bkwd = []
for i in range (1,12):
    h_fwd.append(10**(-i))
    h_bkwd.append(10**(-i))

fwd_res = []
for i in range(len(h_fwd)):
    fwd_res.append(fwd_diff(f,x,h_fwd[i]))
bkwd_res = []
for i in range(len(h_bkwd)):
    bkwd_res.append(bkwd_diff(f,x,h_bkwd[i]))

fwd_err = []
for i in range(len(h_fwd)):
    fwd_err.append(np.abs(fwd_res[i]-true_val))
bkwd_err = []
for i in range(len(h_bkwd)):
    bkwd_err.append(np.abs(bkwd_res[i]-true_val))

print("the true value is ", true_val)
print("h     forward diff       error              backward diff      error")
rowf = "{0:1.0e} {1:1.16f} {2:1.16f} {3:1.16f} {4:1.16f}"
for i in range(len(h_fwd)):
    print(rowf.format(h_fwd[i],fwd_res[i],fwd_err[i],bkwd_res[i],bkwd_err[i]))
```

Now apply central difference 

```python
def cent_diff(f,x,h):
    return (f(x+(h/2))-f(x-(h/2)))/h

h_cent = []
for i in range (1,12):
    h_cent.append(10**(-i))

cent_res = []
for i in range(len(h_cent)):
    cent_res.append(cent_diff(f,x,h_cent[i]))

cent_err = []
for i in range(len(h_cent)):
    cent_err.append(np.abs(cent_res[i]-true_val))

print("the true value is ", true_val)
print("h     central diff       error")
rowf = "{0:1.0e} {1:1.16f} {2:1.16f}"
for i in range(len(h_cent)):
    print(rowf.format(h_cent[i],cent_res[i],cent_err[i],))
```

How do they compare? 

Central difference acheives optimum error at h=1e-5 and worsens after that. Forward and backward difference acheive optimum error at h=1e-8. Central difference is usually more accurate until very low values of h, where rounding error takes over and the errors for all 3 are similar.

### Second derivatives

Let's take the second derivative use central difference of the same function $f(x) = e^{\sin(2x)}$ at $x = 0.5$ & compare with the true value 

$$ \frac{d^2}{dx^2}e^{\sin(2x)} = 2 e^{\sin(2x)}(-2\sin(2x)+\cos(4x)+1)  $$

```python
def f_prime_prime(x):
    return 2*math.e**(np.sin(2*x))*((-2*np.sin(2*x))+np.cos(4*x)+1)

true_val_doub = f_prime_prime(0.5)

def doub_diff(f,x,h):
    return (f(x+h)-2*f(x)+f(x-h))/(h**2)

h_doub = []
for i in range (1,8):
    h_doub.append(10**(-i))

doub_res = []
for i in range(len(h_doub)):
    doub_res.append(doub_diff(f,x,h_doub[i]))

doub_err = []
for i in range(len(h_doub)):
    doub_err.append(np.abs(doub_res[i]-true_val_doub))

print("the true value is ", true_val)
print("h     second diff         error")
rowf = "{0:1.0e} {1:1.16f} {2:1.16f}"
for i in range(len(h_doub)):
    print(rowf.format(h_doub[i],doub_res[i],doub_err[i]))
```

### Automatic Differentiation - book example

Store value/derivative pairs as two-element lists in Python of the form [value, derivative]

Define some functions to perform basic mathematics operations on these inputted lists. 

Apply this to overall figure out the value of the function $f(x) = 3x^2 +e^x$ and its derivative at $t = 1/2$

To do this, perform the following sequence of operations: 

$$ u_1 = x^2$$
$$ u_2 = 3u_1$$
$$ u_3 = e^x$$

and finally 

$$ f(x) = u_2 + u_3$$

Each each mathematical operation will be performed using the generalized operations listed above and the end result will be a pair $[f,f']$, representing the value of $f$ and its derivative.

Final bit of info: to calculate the derivative $f'(t)$ at any value of $t$, need an input list to evaluate our function $f(𝑥)$. Use $[t,1]$ since for whatever value of $t$ we are interested in, by definition, this is the correct value/derivative pair for the independent variable $t$, because $dt/dt = 1$ always. So if we use this pair as the argument of our function, we will get out a pair representing the value of the function $f(t)$ and its derivative $f'(t)$, at the chosen value of $t$.


```python
def exp(x):
    vx,dx = x
    return[math.exp(vx), dx*math.exp(vx)]

def power(x,c):
    vx,dx = x
    return[vx**c, dx*(c*vx**(c-1))]

def int_mult(x,c):
    vx,dx = x
    return[c*vx, c*dx]

def add(x,y):
    vx,dx = x
    vy,dy = y
    return[vx+vy, dx+dy]

def f(x):
    u1 = power(x,2)
    u2 = int_mult(u1, 3)
    u3 = exp(x)
    val = add(u2, u3)
    return val

print(f([0.5,1]))
```

The actual value of the function and the derivative at $f(t)$ at $t = 1/2$ is 
$$f (t) = 3t^2 + e^t = \frac{3}{4} + e^{1/2} = 2.3987 ...$$
$$ f'(t) = 6t + e^t = 3 + e^{1/2} = 4.6487 ...$$

Look how exact we are!
