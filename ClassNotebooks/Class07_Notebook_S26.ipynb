#### You can also store 3D shapes in an array 

```python
s4 = canvas(width = 400, height = 400, background = color.black)
s4.select()
i = 0
spheres = []
while i < 5:
    spheres.append(sphere(pos = vec((6 - (3 * i)),0,-0), radius = 1, color = color.green))
    i += 1
```

##  Vectors & Rate

#### Create an object & move it
- Create a sphere and give it a specific location 
- Change the position 

What do you notice? 

```python
s5 = canvas(width = 400, height = 400, background = color.black)
s5.select()
i = 0
velocity = vec(-1, 0, 0)
ball = sphere(pos = vec(5 , 0, 0), radius = 1)
while i < 20:
    ball.pos = ball.pos + velocity * (0.1 * i)
    i += 1
```

#### Using rate()
 
 - Do the same as above, but use the rate function in between the creation and position 
 - What happens as you change the value in rate()? 


```python
s6 = canvas(width = 400, height = 400, background = color.black)
s6.select()
i = 0
velocity = vec(-1, 0, 0)
ball = sphere(pos = vec(5 , 0, 0), radius = 1)
while i < 15:
    rate(30)
    ball.pos = ball.pos + velocity * (0.1 * i)
    i += 1
```

#### Looping through position change
- Create a loop to change the position of the sphere 50 times 


```python
s7 = canvas(width = 400, height = 400, background = color.black)
s7.select()
i = 0
velocity = vec(0, 0, 0)
ball = sphere(pos = vec(-1 , 0, 0), radius = 1)
while i < 50:
    rate(30)
    ball.pos = ball.pos + velocity * (math.sin(0.1 * i))
    i += 1
```
