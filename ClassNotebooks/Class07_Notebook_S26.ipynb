#### Example: A Moving Sphere

A program to move a sphere around on the screen in a uniform circle

```python

theta = 0
r = 10

x = r*cos(theta)
y = r*sin(theta)


scene6 = canvas(width=800,height=800,autoscale=False)
s = sphere(pos=vector(x,y,0),radius = 0.1,color=color.orange)

for theta in np.arange(0,20*np.pi,0.1):
    rate(30)
    x = r*cos(theta)
    y = r*sin(theta)
    s.pos = vector(x,y,0)

```
