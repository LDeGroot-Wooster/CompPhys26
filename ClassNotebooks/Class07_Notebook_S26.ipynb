#### Now make another sphere that is green!

#### We can create a collection of spheres 

```python
scene2 = canvas(width=600,height=600,center=vector(5,0,0),background=color.white)
L = 5
R = 0.3
for i in range(-L,L+1):
    for j in range(-L,L+1):
        sphere(radius=R,pos=vector(i,j,0),color=color.cyan)
```

#### You can also store 3D shapes in an array 

```python
s = np.empty(10,sphere)

scene3 = canvas(width=600,height=600,center=vector(5,0,0),background=color.white)
n=0
for x in np.linspace(-9,9,10):
    s[n] = sphere(pos=vector(n,0,0))
    n+= 1
```

#### Example: A Moving Sphere

A program to move a sphere around on the screen in a uniform circle

```python
xpos = 1
ypos = 0
scene5 = canvas(width=600,height=600,center=vector(5,0,0))
s= sphere(pos=vector(xpos,ypos,0),radius=0.1)
for theta in np.arange(10,10*np.pi,0.1):
    rate(30)
    xpos=1.0*np.cos(theta)
    ypos=1.0*np.sin(theta)
    s.pos = vector(xpos,ypos,0)
```
