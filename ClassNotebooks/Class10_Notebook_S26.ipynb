## Applying Trapezoidal Rule

Example 5.1: Calculate the integral of $x^4-2x+1$ from $x=0$ to $x=2$ using 10 slices

$$ I(a,b) = h\left[ \frac{1}{2}f(a) + \frac{1}{2}f(b) + \sum_{k=1}^{N-1}f(a+kh)\right] $$

```python
# define function of function want to integrate
def f(x):
    return x**4-2*x+1

def trapezoidal_int(a,b,N,f): 
    # a = lower limit of integration 
    # b = upper limit of integration 
    # N = number of steps 
    # func = function that you are integrating, should be a user-defined function 
    
    # calculate h - width of each step
    h = (b-a)/N

    #integrate using extended trapezoidal rule
    s = 0.5*f(a)+0.5*f(b)
    for k in range(1,N):
        s += f(a+k*h)
    I = h*s

    # return value of the integral 
    return(I)

# define my limits of integration & number of steps
a = 0.0
b = 2.0
N = 10


# get first result 
I1_t = trapezoidal_int(a,b,N,f)
print("the result using Trapezoidal rule for N = 10 is", I1_t)

# double number of steps 
N2 = 20
# get first result 
I2_t = trapezoidal_int(a,b,N2,f)
print("the result using Trapezoidal rule for N = 20 is",I2_t)


# approximation error 
err_trap = np.abs((1/3)*(I2_t-I1_t))
print("the error in the result is =", err_trap)
```

## Simpson's Rule

Same integral - Calculate the integral of $x^4-2x+1$ from $x=0$ to $x=2$ using 10 slices

$$ I(a,b) = \frac{1}{3}h\left[ f(a) + f(b) + 4\sum_{k \text{ odd}}^{N-1}f(a+kh) + 2\sum_{k \text{ even}}^{N-2}f(a+kh)\right] $$

```python
def f(x):
    return x**4 - 2*x + 1

def simpsons_int(a,b,N,f): 
    # a = lower limit of integration 
    # b = upper limit of integration 
    # N = number of steps 
    # func = function that you are integrating, should be a user-defined function 
    
    # calculate h - width of each step
    h = (b-a)/N

    # evaluate using simpson's rule
    s = f(a) + f(b)
    for k in range(1,N):
        if k%2 == 1: # checking for odd
            s+= 4*f(a+k*h)
        else:
            s += 2*f(a+k*h)
    I = h/3*s

    # return value of the integral 
    return(I)


a = 0
b = 2
N = 10

# get first result 
I1_s = simpsons_int(a,b,N,f)
print("the result using Simpson's rule for N = 10 is", I1_s)

# double number of steps 
N2 = 20
# get first result 
I2_s = simpsons_int(a,b,N2,f)
print("the result using Simpson's rule for N = 20 is",I2_s)


      
## error
delta_s = np.abs((I2_s-I1_s)/15)
print("the error in the result is =", delta_s)
```

$$ x_k^\prime = \frac{1}{2}(b-a)x_k + \frac{1}{2}(b+a) $$
$$ w_k^\prime = \frac{1}{2}(b-a)w_k $$

```python
def f(x):
    return x**4 -2*x +1
a=0.0
b=2.0
N=3
x, w = gaussxw(N, a, b)
print(x, w)

s = 0.0
for k in range(N):
    s += w[k] * f(x[k])
print(s)
```

## Multiple Integrals

Evalute the following integral using Gaussian quadrature

$$ I = \int_1^2 \int_{-1}^2 xy^2 dx dy $$

Answer should be: $3.5$

```python
def h(x, y):
    return x*y**2
ax = -1
bx = 2
ay = 1
by = 2
N = 10

xx, wx = gaussxw(N, ax, bx)
yy, wy = gaussxw(N, ay, by)

s = 0.0
for i in range(N):
    for j in range(N):
        s += wx[i]*wy[j]*h(xx[i], yy[j])
print(s)
```
