## Saving Figures from Jupyter Notebooks

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create array of theta values 
t = np.linspace(0,20,500)

# make x and y values based on theta 
x = 3*np.cos(t*3.2)
y = 5*np.sin(t*2)

#plot the values, label axes 
plt.scatter(t,y, label="Data 1")
plt.scatter(t,x, label="Data 2")
plt.xlabel("time (s)")
plt.title("Harmonic Waves", fontsize=14)

# Save the figure to a file
plt.savefig('harmonic_plot.png') # Saves as a PNG image

# always save the figure before plt.show() or will get blank file 
plt.show()
```
