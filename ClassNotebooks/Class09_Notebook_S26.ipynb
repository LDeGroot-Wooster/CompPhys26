## Applying Trapezoidal Rule

Example 5.1: Calculate the integral of $x^4-2x+1$ from $x=0$ to $x=2$ using 10 slices

$$ I(a,b) = h\left[ \frac{1}{2}f(a) + \frac{1}{2}f(b) + \sum_{k=1}^{N-1}f(a+kh)\right] $$

```python
def f(x):
    func = x**4-2*x+1
    return func

b = 2
a = 0
N = 10
h=(b-a)/N
k_sum = 0
for k in range(1,N):
    val = f(a+k*h)
    k_sum += val

area = h*(0.5*f(a)+0.5*f(b)+k_sum)
print(area)
```

```python
# double number of steps 
N = 150
h=(b-a)/N
k_sum = 0
for k in range(1,N):
    val = f(a+k*h)
    k_sum += val
area = h*(0.5*f(a)+0.5*f(b)+k_sum)
print(area)
```

## Simpson's Rule

Same integral - Calculate the integral of $x^4-2x+1$ from $x=0$ to $x=2$ using 10 slices

$$ I(a,b) = \frac{1}{3}h\left[ f(a) + f(b) + 4\sum_{k \text{ odd}}^{N-1}f(a+kh) + 2\sum_{k \text{ even}}^{N-2}f(a+kh)\right] $$

```python
def f(x):
    func = x**4-2*x+1
    return func

b = 2
a = 0
N = 10
h=(b-a)/N
k_odd_sum_1 = 0
for k in range(1,N,2):
    val = f(a+k*h)
    k_odd_sum_1 += val

k_even_sum_1 = 0
for k in range(2,N,2):
    val = f(a+k*h)
    k_even_sum_1 += val

I1 = (1/3)*h*(f(a)+f(b)+4*k_odd_sum_1+2*k_even_sum_1)
print(I1)

N = 20
h=(b-a)/N
k_odd_sum = 0
for k in range(1,N,2):
    val = f(a+k*h)
    k_odd_sum += val

k_even_sum = 0
for k in range(2,N,2):
    val = f(a+k*h)
    k_even_sum += val

I2 = (1/3)*h*(f(a)+f(b)+4*k_odd_sum+2*k_even_sum)
print(I2)
```

## Error on Simpson's Rule

Double the number of steps used to evaluate the integral above and evaluate the error on the result. 

 $$\delta = \frac{1}{15}(I_2 - I_1)$$

```python
delta = (1/15)*(I2-I1)
print(delta)
```

## Choosing Number of Steps

Simpsons: 

 $$\delta_i = \frac{1}{15}(I_i - I_{i-1})$$

Trapezoidal: 

 $$\delta_i = \frac{1}{3}(I_i - I_{i-1})$$

```python
def f(x):
    func = x**4-2*x+1
    return func

b = 2
a = 0
N = 10
h=(b-a)/N
k_sum = 0
for k in range(1,N):
    val = f(a+k*h)
    k_sum += val

I1 = h*(0.5*f(a)+0.5*f(b)+k_sum)
print(I1)

#double steps
N2 = 20
h2 =(b-a)/N
k_sum2 = 0
for k2 in range(1,N2):
    if k2%2 == 1:
        k_sum2 += f(a+k2*h2)

I2 = 0.5*I1 + h2*k_sum2
print(I2)

delta = (1/3)*(I2-I1)
```
