## Applying Trapezoidal Rule

Example 5.1: Calculate the integral of $x^4-2x+1$ from $x=0$ to $x=2$ using 10 slices

$$ I(a,b) = h\left[ \frac{1}{2}f(a) + \frac{1}{2}f(b) + \sum_{k=1}^{N-1}f(a+kh)\right] $$

```python
h = 0.2
n = 10
a = 0.0
b = 2.0
sum = (0.5 * f(a)) + (0.5 * f(b))
def f(x):
    return (x ** 4) - (2 * x) + 1

for k in range(1, n):
    sum += f(a + k*h)

I = h * sum
print(I)
    
```

```python
# double number of steps 

n = 10000000
a = 0.0
b = 2.0
h = (b-a)/n
sum = (0.5 * f(a)) + (0.5 * f(b))
def f(x):
    return (x ** 4) - (2 * x) + 1

for k in range(1, n):
    sum += f(a + k*h)

I = h * sum
print(I)
```

## Simpson's Rule

Same integral - Calculate the integral of $x^4-2x+1$ from $x=0$ to $x=2$ using 10 slices

$$ I(a,b) = \frac{1}{3}h\left[ f(a) + f(b) + 4\sum_{k \text{ odd}}^{N-1}f(a+kh) + 2\sum_{k \text{ even}}^{N-2}f(a+kh)\right] $$

```python
n = 10
a = 0.0
b = 2.0
h = (b-a)/n

def f(x):
    return (x ** 4) - (2 * x) + 1
    
sum = (f(a)) + (f(b))

for k in range(1, n):
    if k%2 == 1:
        sum += 4*f(a + k*h)
    else:
        sum += 2*f(a + k*h)
I1_s = h/3 * sum
print(I1_s)

n = 20
a = 0.0
b = 2.0
h = (b-a)/n

def f(x):
    return (x ** 4) - (2 * x) + 1
    
sum = (f(a)) + (f(b))

for k in range(1, n):
    if k%2 == 1:
        sum += 4*f(a + k*h)
    else:
        sum += 2*f(a + k*h)

I2_s = h/3 * sum
print(I2_s)
```

## Error on Simpson's Rule

Double the number of steps used to evaluate the integral above and evaluate the error on the result. 

 $$\delta = \frac{1}{15}(I_2 - I_1)$$

```python

delta = 1/15 * (I2_s - I1_s)
print(delta)
```

## Choosing Number of Steps

Simpsons: 

 $$\delta_i = \frac{1}{15}(I_i - I_{i-1})$$

Trapezoidal: 

 $$\delta_i = \frac{1}{3}(I_i - I_{i-1})$$

```python
n = 10
a = 0.0
b = 2.0
h = (b-a)/n
sum = (0.5 * f(a)) + (0.5 * f(b))
def f(x):
    return (x ** 4) - (2 * x) + 1

for k in range(1, n):
    sum += f(a + k*h)

I1 = h * sum
print(I1)

# double

n2 = 2*n
h2 = (b-a)/n2

I2 = 0.5 * I1
sum2 = 0
for k in range(1, n2):
    if k%2 == 1:
        sum2 += f(a + k*h2)

I2 += h2*sum2
print(I2, I1)
delta = np.abs(I2-I1)/3
print(delta)
```
