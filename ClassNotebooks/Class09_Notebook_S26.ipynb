## Applying Trapezoidal Rule

Example 5.1: Calculate the integral of $x^4-2x+1$ from $x=0$ to $x=2$ using 10 slices

$$ I(a,b) = h\left[ \frac{1}{2}f(a) + \frac{1}{2}f(b) + \sum_{k=1}^{N-1}f(a+kh)\right] $$

```python
#How many slices and my other constants
n = 20000
a = 0. #Lower limit
b = 2. #Upper limit

#Calculate h
h = (b-a)/n

#Defining my function
def f(x):
    return x**4 - 2*x + 1

#Integrate using the trapezoidal rule
# Define sum variable
S = 0.5*f(a)+0.5*f(b)
for k in range(1,n):
    S += f(a+k*h)
I1_s = S*h

print(I1_s)

```

## Simpson's Rule

Same integral - Calculate the integral of $x^4-2x+1$ from $x=0$ to $x=2$ using 10 slices

$$ I(a,b) = \frac{1}{3}h\left[ f(a) + f(b) + 4\sum_{k \text{ odd}}^{N-1}f(a+kh) + 2\sum_{k \text{ even}}^{N-2}f(a+kh)\right] $$

```python
#How many slices and my other constants
n = 10
a = 0. #Lower limit
b = 2. #Upper limit

#Calculate h
h = (b-a)/n

#Defining my function
def f(x):
    return x**4 - 2*x + 1

#Integrate using the trapezoidal rule
# Define sum variable
S = f(a)+f(b)
for k in range(1,n):
    if k%2 == 0: #Is an even number
        S += 2*f(a+k*h)
    if k%2 == 1: #Is an odd number
        S += 4*f(a+k*h)

I = S*h/3

print(I)
```

## Choosing Number of Steps

Simpsons: 

 $$\delta_i = \frac{1}{15}(I_i - I_{i-1})$$

Trapezoidal: 

 $$\delta_i = \frac{1}{3}(I_i - I_{i-1})$$

```python
#How many slices and my other constants
n = 10
a = 0. #Lower limit
b = 2. #Upper limit

#Calculate h
h = (b-a)/n

#Defining my function
def f(x):
    return x**4 - 2*x + 1

#Integrate using the trapezoidal rule
# Define sum variable
S = 0.5*f(a)+0.5*f(b)
for k in range(1,n):
    S += f(a+k*h)
I1 = S*h

print(I1)

# Double the steps
N2 = 2*n
h2 = (b-a)/N2
s2=0

#Evalaulate the sum
I2 = 0.5*I1
for k in range(1,N2):
    if k%2 == 1: #Am i odd?
        s2+= f(a+k*h2)
I2 += h2*s2
print(I2)

delta = np.abs(I2-I1)/3
print(delta)
```
