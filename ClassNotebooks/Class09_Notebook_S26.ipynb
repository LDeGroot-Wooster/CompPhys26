## Applying Trapezoidal Rule

Example 5.1: Calculate the integral of $x^4-2x+1$ from $x=0$ to $x=2$ using 10 slices

$$ I(a,b) = h\left[ \frac{1}{2}f(a) + \frac{1}{2}f(b) + \sum_{k=1}^{N-1}f(a+kh)\right] $$

```python
def func(x):
    return x**4 -2*x + 1
a = 0.0
b = 2.0
N = 10
h = (b-a) / N
s = 0.5*func(a)+0.5*func(b)
for k in range(1, N):
    s += func(a+k*h)
I = h*s
print(I)
```

## Simpson's Rule

Same integral - Calculate the integral of $x^4-2x+1$ from $x=0$ to $x=2$ using 10 slices

$$ I(a,b) = \frac{1}{3}h\left[ f(a) + f(b) + 4\sum_{k \text{ odd}}^{N-1}f(a+kh) + 2\sum_{k \text{ even}}^{N-2}f(a+kh)\right] $$

```python
def func(x):
    return x**4 - 2*x + 1
a = 0.0
b = 2.0
n = 10 # MUST BE EVEN
h = (b-a) / n
s = func(a) + func(b)
N = np.linspace(1, n-1, int(n/2))
for k in N:
    print(s)
    s += 4*func(a+k*h) + 2*func(a+(k+1)*h)
I = (1/3)*h*s
print(s)
print(I)
```

```python
def func(x):
    return x**4 -2*x + 1
a = 0.0
b = 2.0
n = 10 # MUST BE EVEN
h = (b-a) / n
s = func(a) + func(b)
for k in range(1, n):
    if k%2 == 1:
        s += 4*func(a+k*h)
    else:
        s += 2*func(a+k*h)
I = (1/3)*h*s
print(I)
```
