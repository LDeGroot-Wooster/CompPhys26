### Gaussian Elimination with backsubstitution 

Let's solve the example from the book: Use Gaussian Elimination to find $w, x, y,$ and $z$. 

$$\begin{pmatrix}
2 & 1  & 4  & 1  \\
3 & 4  & -1 & -1 \\
1 & -4 & 1  & 5  \\
2 & -2 & 1  & 3  
\end{pmatrix} 
\begin{pmatrix}
w \\
x \\
y \\
z 
\end{pmatrix}
=
\begin{pmatrix}
-4 \\
3  \\
9  \\
7  
\end{pmatrix}$$

```python
# define matrix & vectors 
A = np.array([[2, 1,  4,  1],
             [3, 4, -1, -1],
             [ 1, -4, 1,  5], 
             [ 2, -2, 1,  3]], float)
v = np.array([-4, 3, 9, 7], float)

# determine the number of variables solving for
N = len(v)
# create array for results, x
x = np.zeros(N, float)
print(A)
```

```python
# gaussian elimination with backsubtitution 
for m in range(N):
    div = A[m, m]
    A[m,:] /= div
    v[m] /= div
    for i in range(m+1, N):
        mult = A[i,m]
        A[i,:] -=  mult*A[m,:]
        v[i] -= mult*v[m]
for m in range(N-1, -1, -1):
    x[m] = v[m]
    for i in range(m+1, N):
        x[m] -= A[m,i]*x[i]
print(x)
```

**note** this works for matrices of any size

### LU Substitution 

Apply LU substitution to the same matrices 

```python
#creates a number of matrices L, then multiplies them together, the product of which can be solved using backsubstitution
#this is efficient because it means we can calculate the matrices once and be done with it, which is the most time consuming part
#we could write this ourselves, but numpy has the .linalg module, which has a command 'solve' which does this all for us
```

```python
A = np.array([[0, 1,  4,  1],
             [3, 4, -1, -1],
             [ 1, -4, 1,  5], 
             [ 2, -2, 1,  3]], float)
v = np.array([-4, 3, 9, 7], float)
x = np.linalg.solve(A, v)
print(x)
```
