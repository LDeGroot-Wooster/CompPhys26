### Gaussian Elimination with backsubstitution 

Let's solve the example from the book: Use Gaussian Elimination to find $w, x, y,$ and $z$. 

$$\begin{pmatrix}
2 & 1  & 4  & 1  \\
3 & 4  & -1 & -1 \\
1 & -4 & 1  & 5  \\
2 & -2 & 1  & 3  
\end{pmatrix} 
\begin{pmatrix}
w \\
x \\
y \\
z 
\end{pmatrix}
=
\begin{pmatrix}
-4 \\
3  \\
9  \\
7  
\end{pmatrix}$$

```python
# define matrix & vectors 
A = np.array([[2, 1,  4,  1],
             [3, 4, -1, -1],
             [ 1, -4, 1,  5], 
             [ 2, -2, 1,  3]], float)
v = np.array([-4, 3, 9, 7], float)

# determine the number of variables solving for
N = len(v)
# create array for results, x

print(A)
```

```python
# gaussian elimination with backsubtitution 
def Gaussian_Elimination(A,v):
    N = len(v)
    for m in range(N):
        factor = A[m,m]
        
        A[m,:] /= factor
        v[m] /= factor
        for i in range(m+1,N):
            factor = A[i,m]
            A[i,:] -= factor*A[m,:]
            v[i] -= factor*v[m]
    return A,v
A,v = Gaussian_Elimination(A,v)
print(A)
print(v)
def Backward_Substitution(A,v):
    N = len(v)
    x = np.zeros(N, float)
    for m in range(N-1,-1,-1):
        x[m]=v[m]
        for i in range(m+1,N):
            x[m] -= A[m,i]*x[i]
    return x
x = Backward_Substitution(A,v)
print(x)
```

**note** this works for matrices of any size 

### Pivoting 

We need to interchange rows of the matrix for certain situations. Let's see an example below. 

$$\begin{pmatrix}
0 & 1  & 4  & 1  \\
3 & 4  & -1 & -1 \\
1 & -4 & 1  & 5  \\
2 & -2 & 1  & 3  
\end{pmatrix} 
\begin{pmatrix}
w \\
x \\
y \\
z 
\end{pmatrix}
=
\begin{pmatrix}
-4 \\
3  \\
9  \\
7  
\end{pmatrix}$$

```python
# define matrix & vectors 
A = np.array([[0, 1,  4,  1],
             [3, 4, -1, -1],
             [ 1, -4, 1,  5], 
             [ 2, -2, 1,  3]], float)
v = np.array([-4, 3, 9, 7], float)

# determine the number of variables solving for
N = len(v)
# create array for results, x
x = np.zeros(N, float)
print(A)
```

What happens when we run Gaussian Elimination on this system of equations? 

```python
def Pivot(A):
    N = len(A)
    for m in range(N):
        pivot,largest = m,np.abs(A[m,m])
        for i in range(m+1,N):
            if np.abs(A[i,m])>largest:
                pivot, largest = i,np.abs(A[i,m])
        for i in range(N):
            A[m,i], A[pivot,i] = A[pivot,i],A[m,i]
    return A
Gaussian_Elimination(Pivot(A),v)
```

### LU Substitution 

Apply LU substitution to the same matrices 

```python
# define matrix & vectors 
A = np.array([[2, 1,  4,  1],
             [3, 4, -1, -1],
             [ 1, -4, 1,  5], 
             [ 2, -2, 1,  3]], float)
v = np.array([-4, 3, 9, 7], float)

# determine the number of variables solving for
N = len(v)
# create array for results, x
x = np.linalg.solve(A,v)
print(x)
```
