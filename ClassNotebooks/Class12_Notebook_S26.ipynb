### Gaussian Elimination with backsubstitution 

Let's solve the example from the book: Use Gaussian Elimination to find $w, x, y,$ and $z$. 

$$\begin{pmatrix}
2 & 1  & 4  & 1  \\
3 & 4  & -1 & -1 \\
1 & -4 & 1  & 5  \\
2 & -2 & 1  & 3  
\end{pmatrix} 
\begin{pmatrix}
w \\
x \\
y \\
z 
\end{pmatrix}
=
\begin{pmatrix}
-4 \\
3  \\
9  \\
7  
\end{pmatrix}$$

```python
# define matrix & vectors 
A = np.array([[2, 1,  4,  1],
             [3, 4, -1, -1],
             [ 1, -4, 1,  5], 
             [ 2, -2, 1,  3]], float)
v = np.array([-4, 3, 9, 7], float)

# determine the number of variables solving for
N = len(v)
# create array for results, x
x = np.zeros(N, float)
print(A)
```

```python
# gaussian elimination with backsubtitution
for m in range(N):
    diag = A[m,m] #diagonal component of row A
    A[m,:] = A[m,:]/diag #divide each component in row by diagonal component
    v[m]= v[m]/diag #divide row m in v by same value

    for i in range(m+1,N):
        first = A[i,m] #first nonzero element of each lower row
        A[i,:] -= first*A[m,:] #subtract appropriate multiple of top row from lower rows
        v[i] -= first*v[m]

print(A)
print(v)

#backsubstitution
for m in range(N-1,-1,-1): #3,2,1,0 for above matrix
    x[m] = v[m] #x[3] = v[3]
    for i in range(m+1, N): 
        x[m] = x[m] - A[m,i]*x[i] #for example x = v1 - a12y - a13z

print(x)
```

**note** this works for matrices of any size 

### Pivoting

```python
# define matrix & vectors 
A = np.array([[0, 1,  4,  1],
             [3, 4, -1, -1],
             [ 1, -4, 1,  5], 
             [ 2, -2, 1,  3]], float)
v = np.array([-4, 3, 9, 7], float)

# determine the number of variables solving for
N = len(v)
# create array for results, x
x = np.zeros(N, float)
print(A)
```

What happens when we run Gaussian Elimination on this system of equations?

```python
A = np.array([[0, 1,  4,  1],
             [3, 4, -1, -1],
             [ 1, -4, 1,  5], 
             [ 2, -2, 1,  3]], float)
v = np.array([-4, 3, 9, 7], float)

# determine the number of variables solving for
N = len(v)
# create array for results, x
x = np.zeros(N, float)

for m in range(N):
    biggest_first = np.abs(A[m,m])
    biggest_first_row = m
    current_row = A[m,:].copy()
    current_row_v = v[m].copy()
    for j in range(m, N):
        if np.abs(A[j,m]) > biggest_first:
            biggest_first = np.abs(A[j,m])
            biggest_first_row = j
    #swap rows for A
    A[m,:] = A[biggest_first_row,:]
    A[biggest_first_row,:] = current_row
    #swap rows for v
    v[m] = v[biggest_first_row]
    v[biggest_first_row] = current_row_v
    
    diag = A[m,m] #diagonal component of row A
    A[m,:] = A[m,:]/diag #divide each component in row by diagonal component
    v[m]= v[m]/diag #divide row m in v by same value

    for i in range(m+1,N):
        first = A[i,m] #first nonzero element of each lower row
        A[i,:] -= first*A[m,:] #subtract appropriate multiple of top row from lower rows
        v[i] -= first*v[m]

#backsubstitution
for m in range(N-1,-1,-1): #3,2,1,0 for above matrix
    x[m] = v[m] #x[3] = v[3]
    for i in range(m+1, N): 
        x[m] = x[m] - A[m,i]*x[i] #for example x = v1 - a12y - a13z

print(x)
```

### LU Substitution 

Apply LU substitution to the same matrices 

```python
# define matrix & vectors 
A = np.array([[2, 1,  4,  1], #using multiple brackets to create a proper 2d array
             [3, 4, -1, -1],
             [ 1, -4, 1,  5], 
             [ 2, -2, 1,  3]], float)
v = np.array([-4, 3, 9, 7], float)

# determine the number of variables solving for
N = len(v)
# create array for results, x
x = np.linalg.solve(A,v)

print(A)
print(x)
```
