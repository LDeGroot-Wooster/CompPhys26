### Gaussian Elimination with backsubstitution 

Let's solve the example from the book: Use Gaussian Elimination to find $w, x, y,$ and $z$. 

$$\begin{pmatrix}
2 & 1  & 4  & 1  \\
3 & 4  & -1 & -1 \\
1 & -4 & 1  & 5  \\
2 & -2 & 1  & 3  
\end{pmatrix} 
\begin{pmatrix}
w \\
x \\
y \\
z 
\end{pmatrix}
=
\begin{pmatrix}
-4 \\
3  \\
9  \\
7  
\end{pmatrix}$$

```python
# define matrix & vectors 
A = np.array([[2, 1,  4,  1],
             [3, 4, -1, -1],
             [ 1, -4, 1,  5], 
             [ 2, -2, 1,  3]], float)
v = np.array([-4, 3, 9, 7], float)

# determine the number of variables solving for
N = len(v)
# create array for results, x
x = np.zeros(N, float)
print(A)
```

```python
# gaussian elimination with backsubtitution 

for m in range(N):
   
    #divide by diagonal elements
    div = A[m,m]
    A[m,:] /= div
    v[m] /= div

    #subtract ffrom lover rows
    for i in range(m+1, N):
        mult = A[i,m]
        A[i,:] -= mult*A[m,:]
        v[i] -= mult*v[m]

# apply back sub
for m in range(N-1, -1, -1):    
    x[m] = v[m]
    for i in range(m+1,N):
        x[m] -= A[m,i]*x[i]

print(x)   
```

**note** this works for matrices of any size 

### Pivoting 

We need to interchange rows of the matrix for certain situations. Let's see an example below. 

$$\begin{pmatrix}
0 & 1  & 4  & 1  \\
3 & 4  & -1 & -1 \\
1 & -4 & 1  & 5  \\
2 & -2 & 1  & 3  
\end{pmatrix} 
\begin{pmatrix}
w \\
x \\
y \\
z 
\end{pmatrix}
=
\begin{pmatrix}
-4 \\
3  \\
9  \\
7  
\end{pmatrix}$$

```python
# define matrix & vectors 
A = np.array([[0, 1,  4,  1],
             [3, 4, -1, -1],
             [ 1, -4, 1,  5], 
             [ 2, -2, 1,  3]], float)
v = np.array([-4, 3, 9, 7], float)

# determine the number of variables solving for
N = len(v)

# create array for results, x
x = np.zeros(N, float)
print(A)
```

What happens when we run Gaussian Elimination on this system of equations? 

```python
for m in range(N):

    pivot = m
    largest = A[m,m]
    for i in range(m+1, N):
        if np.abs(A[i,m]) > largest:
            largest = np.abs(A[i,m])
            pivot = i
    # swap
    for i in range(N):
        A[i,m], A[pivot, i] = A[pivot, i], A[m,i]
```

```python
for m in range(N):
    for m in range(N):
        pivot = m
        largest = A[m,m]
        for i in range(m+1, N):
            if np.abs(A[i,m]) > largest:
                largest = np.abs(A[i,m])
                pivot = i
    # swap
        for i in range(N):
            A[i,m], A[pivot, i] = A[pivot, i], A[m,i]
        v[m], v[pivot] = v[pivot], v[m]
            
    #divide by diagonal elements
    div = A[m,m]
    A[m,:] /= div
    v[m] /= div

    #subtract ffrom lover rows
    for i in range(m+1, N):
        mult = A[i,m]
        A[i,:] -= mult*A[m,:]
        v[i] -= mult*v[m]

# apply back sub
for m in range(N-1, -1, -1):    
    x[m] = v[m]
    for i in range(m+1,N):
        x[m] -= A[m,i]*x[i]

print(x)   
```

### LU Substitution 

Apply LU substitution to the same matrices 

```python
# define matrix & vectors 
A = np.array([[0, 1,  4,  1],
             [3, 4, -1, -1],
             [ 1, -4, 1,  5], 
             [ 2, -2, 1,  3]], float)
v = np.array([-4, 3, 9, 7], float)

# determine the number of variables solving for
N = len(v)

# create array for results, x
x = np.linalg.solve(A,v)
print(x)
```
