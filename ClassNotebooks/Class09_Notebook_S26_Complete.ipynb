## Applying Trapezoidal Rule

Example 5.1: Calculate the integral of $x^4-2x+1$ from $x=0$ to $x=2$ using 10 slices

$$ I(a,b) = h\left[ \frac{1}{2}f(a) + \frac{1}{2}f(b) + \sum_{k=1}^{N-1}f(a+kh)\right] $$

```python
def f(x): 
    return x**4 - 2*x + 1

# define our constants
a = 0.0 # lower limit
b = 2.0 # upper limit
N = 10 # number of steps

# calculate h
h = (b-a)/N

# integrate using the trapezoidal rule
# define sum varialbe
s = 0.5*f(a)+0.5*f(b)
for k in range(1,N):
    s += f(a+k*h)
I = h*s
print(I)
```

## Simpson's Rule

Same integral - Calculate the integral of $x^4-2x+1$ from $x=0$ to $x=2$ using 10 slices

$$ I(a,b) = \frac{1}{3}h\left[ f(a) + f(b) + 4\sum_{k \text{ odd}}^{N-1}f(a+kh) + 2\sum_{k \text{ even}}^{N-2}f(a+kh)\right] $$

```python
# define our constants
a = 0.0 # lower limit
b = 2.0 # upper limit
N = 10 # number of steps

# calculate h
h = (b-a)/N

# evaluate using simpson's rule
s = f(a)+f(b)
for k in range(1,N):
    if k%2 ==1: #checking for odd
        s += 4*f(a+k*h)
    else:
        s+= 2*f(a+k*h)
I1_s = h/3*s
print(I1_s)
```

```python
### alternative 
s1 = 0.0
for k in range(1,N,2):
    s1 += f(a+k*h)
s2 = 0.0
for k in range(2,N,2):
    s2 += f(a+k*h)   
I1_alt = h/3*(f(a)+f(b)+4*s1+2*s2)
print(I1_alt)
```

## Error on Simpson's Rule

Double the number of steps used to evaluate the integral above and evaluate the error on the result. 

 $$\delta = \frac{1}{15}(I_2 - I_1)$$

```python
delta_s = (I2_s - I1_s)/15
print("the error on the integral calculated using simpsons rule is", delta_s)
```

## Choosing Number of Steps

Simpsons: 

 $$\delta_i = \frac{1}{15}(I_i - I_{i-1})$$

Trapezoidal: 

 $$\delta_i = \frac{1}{3}(I_i - I_{i-1})$$

```python
def f(x): 
    return x**4 - 2*x + 1

# define our constants
a = 0.0 # lower limit
b = 2.0 # upper limit
N = 10 # number of steps

# calculate h
h = (b-a)/N

# integrate using the trapezoidal rule
# define sum varialbe
s = 0.5*f(a)+0.5*f(b)
for k in range(1,N):
    s += f(a+k*h)
I1 = h*s
print(I1)

# double the steps
N2 = 2*N
h2 = (b-a)/N2

#evalauate the sum
I2 = 0.5*I1
s2 = 0
for k in range(1,N2):
    if k%2 ==1:
        s2 += f(a + k*h2)
I2 += h2*s2
print("I1, I2 =", I1, I2)
delta = np.abs(I2-I1)/3
print(delta)
```
