##  For loops in Python

- Let's write a simple `for` loop that prints out each element of an array 

```python
import numpy as np
n = 1
for i in np.arange(0, 20, 2):
    print("Element", n, 'in the array is:', i)
    n += 1
```

-  let's write a simple for loop that includes an if/else statement 

```python
base_arr = np.arange(0, 50, 1)
odd_arr = []
even_arr = []
for i in base_arr:
    if i / 2 == round(i / 2):
        even_arr.append(float(i))
    else:
        odd_arr.append(float(i))

print(even_arr)
print(odd_arr)
```

#### Defining a function

- Let's create a user-defined function that accomplishes the following:
    - the function requires two parameters: a & b
    - if a < b, the function returns the product of the two
    - if a > b, then the function returns the results of a/b

```python
def ex_func(a, b):
    if a < b:
        return a * b
    else:
        return a / b

c = ex_func(3, 7)
print(c)
```

- Evaluate the factorial code from the book

```python
def factorial(n):
    f = 1.0
    for k in range(1,n+1):
        f *= k
    return f

factorial_val = factorial(100)
print(factorial_val)
```
