#### Nested for loops

-  Create a nested for loop that iterates through 2 different lists, element-by-element 

```python
Upper_Bound = int(input("Upper Bound?"))
n = 0
k = 0
for Bob in range(1, Upper_Bound):
    for i in range(2, Upper_Bound):
        if Bob%i==0:
            k = k+1
        else:
            n = n+1
print(n)
print(k)
```

#### Defining a function

- Let's create a user-defined function that accomplishes the following:
    - the function requires two parameters: a & b
    - if a < b, the function returns the product of the two
    - if a >= b, then the function returns the results of a/b

```python
def function_junction(a,b):
    if a<b: 
        return a*b
    else:
        return a/b
    
A = float(input("Your a-value is: "))
B = float(input("Your b-value is: "))
Result = function_junction(A,B)

print(Result)
```

- Evaluate the factorial code from the book

```python
def factorial(n):
    f = 1.0
    for k in range(1,n+1):
        f *= k
    return f

factorial_val = factorial(100)
print(factorial_val)
```
