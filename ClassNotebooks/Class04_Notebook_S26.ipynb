To create an example dictionary, let's say that I have the following data:
- voltage = 2 V, 4 V, 6 V
- current = 1 A, 2 A, 3 A
- uncertainty in voltage = 0.05 V for all 

Below we will: 
- Make the dictionary based on these values and keys
- Add the key and elements: resistance = 2.0 ohms for all 
- list the keys & dictionary 
- print out the second element of one of the key arrays in the dictionary 

```python
voltage = [2, 4, 6] # in volts
current = [1, 2, 3] # in amps
volt_unc = [0.05, 0.05, 0.05]

ohm_dict = {"voltage":voltage, "current":current, "volt_unc":volt_unc}
print(ohm_dict["voltage"][1]) # key, second index value

# print out keys
dict_keys = list(ohm_dict.keys())
print(dict_keys)

# add resistance
resistance = [2, 2, 2] # ohm's law V = IR
ohm_dict["resistance"] = resistance
# ohm_dict["resistance"] = [2, 2, 2]     Another option
print(ohm_dict)
```

## User-defined functions 


#### Defining a function

- Let's create a user-defined function that accomplishes the following:
    - the function requires two parameters: a & b
    - if a < b, the function returns the product of the two
    - if a > b, then the function returns the results of a/b

```python
def example_function(a, b):
    if a < b:
        return a*b
    else:
        return a/b, a**3 # can return multiple things as a list
```

```python
print(example_function(3,2))

c = example_function(6,7)
print(c)
```

- Evaluate the factorial code from the book

```python
def factorial(n):
    f = 1.0
    for k in range(1,n+1):
        f *= k
    return f

factorial_val = factorial(100)
print(factorial_val)
```
