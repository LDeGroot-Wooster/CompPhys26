### The Relaxation Method 
#### Single Variable 

Outline: 
1) Guess an initial value for the unknown variable, plug into the right hand side. 

2) Repeat the process by taking this value and plugging back into RHS.

3) Continue until eventually converges to a fixed point of the equation – until “relaxes” into solution.

4) Continue until eventually converges to a fixed point of the equation – until “relaxes” into solution. If the process converges, gives you a solution to the equation since feed one value into RHS and get same out on LHS. 






#### Let's apply for the function:

$$ y = \cos y +\frac{1}{3} $$


```python

def f(y):
    return np.cos(y) + 1/3

N = 50 #number of iterations

y = 1/3 #guess
for _ in range(N):
    print(y)
    y = f(y)

```

### Relaxation Method - Multi-variable 

#### Glycolosis (book problem 6.13)


The biochemical process of glyocolysis, the breakdown of glucose in the body to release energy, can be modeled by the equations
$$ \frac{dx}{dt} = -x + ay + x^2 y, \qquad
\frac{dy}{dt} = b - ay - x^2 y.
$$
Here $x$ and $y$ represent concentrations of two chemicals, ADP and F6P, and $a$ and $b$ are positive constants.  One of the important features of nonlinear linear equations like these is their stationary points, meaning values of $x$ and $y$ at which the derivatives of both variables become zero simultaneously, so that the variables stop changing and become constant in time.  Setting the derivatives to zero above, the stationary points of our glycolysis equations are solutions of
$$
-x + ay + x^2 y = 0, \qquad b - ay - x^2 y = 0.
$$

### Binary Search 

Outline: 
1) Given an initial pair of points $x_1, x_2$ check that $f(x_1)$ and $f(x_2)$ have opposite signs. Also choose a target accuracy for the answer you want. 

2) Calculate the midpoint $x' = \frac{1}{2}(x_1+x_2)$ and evaluate $f(x')$

3) If $f(x')$ has the same sign as $f(x_1)$ then set $x_1 = x'$. Otherwise set $x_2 = x'$. 

4) If $|{x_1-x_2}|$ is greater than the target accuracy, repeat from step 2. Otherwise, calculate $\frac{1}{2}(x_1+x_2)$ once more and this is the final estimate of the position of the root. 


#### Let's apply for the function:

$$ f(x) = e^{x-\sqrt{x}}-x$$

```python
def f(x):
    return np.exp(x-np.sqrt(x)) - x



def binary_solve(x1,x2,f,acc):
    f1 = f(x1)
    f2 = f(x2)

    if f1*f2 < 0: #negative * positive always equals negative
        while abs(x2-x1) > acc:
            mid = 0.5*(x1+x2)
            fmid = f(mid)
            if fmid*f1 > 0:
                x1 = mid
            elif fmid*f2 > 0:
                x2 = mid
        return 0.5*(x1+x2)
    else:
        return -999.99 #used in astronomy to inform that an answer is wrong



#initial guesses
x1 = 0.2
x2 = 2.0
targetaccuracy = 1e-8



result = binary_solve(x1,x2,f,targetaccuracy)
print(result)
```
