### Forward, Backward, and Central difference 

Evaluate the derivative of $f(x) = e^{\sin(2x)}$ at $x = 0.5$

If you took the derivative by hand, what would it be? 

```python
def fprime(x):
    return 2*np.cos(2*x)*np.exp(np.sin(2*x))

x = 0.5
true_val =fprime(x)
print(true_val)
```

Now apply forward/backward difference 

```python
def f(x):
    return np.exp(np.sin(2*x))

#apply forward difference
def calc_fd(f,x,h):
    fd = (f(x+h)-f(x))/h
    return fd

#backward difference
def calc_bd(f,x,h):
    fd = (f(x)-f(x-h))/h
    return fd
#this can be called and used just like the fd func


x= 0.5
h_fd = [10**(-i) for i in range(1,12)]

fd_array = np.zeros(len(h_fd))

for j in range(len(fd_array)):
    fd_array[j]= calc_fd(f,x,h_fd[j])


fd_error = np.abs(fd_array - true_val)

print("the true value is ", true_val)
print(" h      forward diff        error")
rowf = "{0:1.0e} {1:1.16f} {2:1.16f}"
for i in range(len(h_fd)):
    print(rowf.format(h_fd[i], fd_array[i], fd_error[i]))
```

Now apply central difference 

```python

def calc_cd(f,x,h):
    cd = (f(x+h/2)-f(x-h/2))/h
    return cd

cd_array = np.zeros(len(h_fd))

for j in range(len(cd_array)):
    cd_array[j]= calc_cd(f,x,h_fd[j])

cd_error = np.abs(cd_array - true_val)

print("the true value is ", true_val)
print(" h      central diff        error")
rowf = "{0:1.0e} {1:1.16f} {2:1.16f}"
for i in range(len(h_fd)):
    print(rowf.format(h_fd[i], cd_array[i], cd_error[i]))
```

### Automatic Differentiation - book example

Store value/derivative pairs as two-element lists in Python of the form [value, derivative]

Define some functions to perform basic mathematics operations on these inputted lists. 

Apply this to overall figure out the value of the function $f(x) = 3x^2 +e^x$ and its derivative at $t = 1/2$

To do this, perform the following sequence of operations: 

$$ u_1 = x^2$$
$$ u_2 = 3u_1$$
$$ u_3 = e^x$$

and finally 

$$ f(x) = u_2 + u_3$$

Each each mathematical operation will be performed using the generalized operations listed above and the end result will be a pair $[f,f']$, representing the value of $f$ and its derivative.

Final bit of info: to calculate the derivative $f'(t)$ at any value of $t$, need an input list to evaluate our function $f(𝑥)$. Use $[t,1]$ since for whatever value of $t$ we are interested in, by definition, this is the correct value/derivative pair for the independent variable $t$, because $dt/dt = 1$ always. So if we use this pair as the argument of our function, we will get out a pair representing the value of the function $f(t)$ and its derivative $f'(t)$, at the chosen value of $t$.


```python
def power(x,c):
    vx, dx = x
    return [vx**c, c*vx**(c-1)*dx]
            
def mult(x,c):
    vx,dx = x
    return [c*vx,c+dx]


def expon(x):
    vx,dx = x
    return [np.exp(vx), np.exp(vx)*dx]

def add(x,y):
    vx,dx = x
    vy,dy = y
    return [vx+vy, dx+dy]


def f_ad(x):
    u1 = power(x,2)
    u2 = mult(u1,3)
    u3 = expon(x)
    u4 = add(u2,u3)

    return u4


print(f_ad([0.5,1]))

```

The actual value of the function and the derivative at $f(t)$ at $t = 1/2$ is 
$$f (t) = 3t^2 + e^t = \frac{3}{4} + e^{1/2} = 2.3987 ...$$
$$ f'(t) = 6t + e^t = 3 + e^{1/2} = 4.6487 ...$$

Look how exact we are!
