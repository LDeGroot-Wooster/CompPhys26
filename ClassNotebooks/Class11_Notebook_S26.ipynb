### Forward, Backward, and Central difference 

Evaluate the derivative of $f(x) = e^{\sin(2x)}$ at $x = 0.5$

If you took the derivative by hand, what would it be? 

$$\frac{df}{dx}=2\text{e}^{\sin{2x}}\cos{2x}$$
$$\left.\frac{df}{dx}\right|_{0.5}\hspace{-2mm}=2\text{e}^{\sin{1}}\cos{1}\approx 2.506761534986894$$

```python
def e_sin(x):
    return np.exp(np.sin(2*x))
def e_sin_derv(x):
    return np.exp(np.sin(2*x))*2*np.cos(2*x)
true_val = e_sin_derv(0.5)
print(true_value)
```

Now apply forward/backward difference 

```python
def fwd_diff(x,h,f):
    return (f(x+h)-f(x))/h

x_fwd_diff = 0.5
h_fwd_diff = np.array([10**(-i) for i in range(1,12)])

fwd_appr_val = fwd_diff(x_fwd_diff,h_fwd_diff,e_sin)

print("the true value is ", true_val)
print(" h      forward diff        error")
rowf = "{0:1.0e} {1:1.16f} {2:1.16f}"

for i in range(len(fwd_appr_val)):
    fwd_err = abs(fwd_appr_val[i]-true_val)
    print(rowf.format(h_fwd_diff[i], fwd_appr_val[i], fwd_err))
```

backward diff:

```python
def bwd_diff(x,h,f):
    return (-f(x-h)+f(x))/h

x_bwd_diff = 0.5
h_bwd_diff = np.array([10**(-i) for i in range(1,12)])

bwd_appr_val = bwd_diff(x_bwd_diff,h_bwd_diff,e_sin)

print("the true value is ", true_val)
print(" h      backward diff        error")
rowf = "{0:1.0e} {1:1.16f} {2:1.16f}"

for i in range(len(bwd_appr_val)):
    bwd_err = abs(bwd_appr_val[i]-true_val)
    print(rowf.format(h_bwd_diff[i], bwd_appr_val[i], bwd_err))
```

Now apply central difference 

```python
def cent_diff(x,h,f):
    return (f(x+h)-f(x-h))/(2*h)

x_cent_diff = 0.5
h_cent_diff = np.array([10**(-i) for i in range(1,12)])

cd_appr_val = cent_diff(x_cent_diff,h_cent_diff,e_sin)

print("the true value is ", true_val)
print(" h      central diff        error")
rowf = "{0:1.0e} {1:1.16f} {2:1.16f}"

for i in range(len(cd_appr_val)):
    cd_err = abs(cd_appr_val[i]-true_val)
    print(rowf.format(h_cent_diff[i], cd_appr_val[i], cd_err))
```

### Second derivatives

Let's take the second derivative use central difference of the same function $f(x) = e^{\sin(2x)}$ at $x = 0.5$ & compare with the true value 

$$ \frac{d^2}{dx^2}e^{\sin(2x)} = 2 e^{\sin(2x)}(-2\sin(2x)+\cos(4x)+1)  $$

```python
def sec_diff(x,h,f):
    return (f(x+h)+f(x-h)-2*f(x))/(h**2)
def e_sin_derv_2(x):
    return 2*np.exp(np.sin(2*x))*(-2*np.sin(2*x)+np.cos(4*x)+1)

x_sec_diff = 0.5
h_sec_diff = np.array([10**(-i) for i in range(1,12)])

true_val_doub = e_sin_derv_2(0.5)

appr_val_doub = sec_diff(x_sec_diff,h_sec_diff,e_sin)

print("the true value is ", true_val_doub)
print(" h      second diff        error")
rowf = "{0:1.0e} {1:1.16f} {2:1.16f}"

for i in range(len(appr_val_doub)):
    sec_err = abs(appr_val_doub[i]-true_val_doub)
    print(rowf.format(h_sec_diff[i], appr_val_doub[i], sec_err))
```

### Automatic Differentiation - book example

Store value/derivative pairs as two-element lists in Python of the form [value, derivative]

Define some functions to perform basic mathematics operations on these inputted lists. 

Apply this to overall figure out the value of the function $f(x) = 3x^2 +e^x$ and its derivative at $t = 1/2$

To do this, perform the following sequence of operations: 

$$ u_1 = x^2$$
$$ u_2 = 3u_1$$
$$ u_3 = e^x$$

and finally 

$$ f(x) = u_2 + u_3$$

Each each mathematical operation will be performed using the generalized operations listed above and the end result will be a pair $[f,f']$, representing the value of $f$ and its derivative.

Final bit of info: to calculate the derivative $f'(t)$ at any value of $t$, need an input list to evaluate our function $f(𝑥)$. Use $[t,1]$ since for whatever value of $t$ we are interested in, by definition, this is the correct value/derivative pair for the independent variable $t$, because $dt/dt = 1$ always. So if we use this pair as the argument of our function, we will get out a pair representing the value of the function $f(t)$ and its derivative $f'(t)$, at the chosen value of $t$.


```python
def power(x,p):
    val_x,dx =  x # x = [x,dx]
    return [val_x**p, p*val_x**(p-1) * dx]
    
def multiplication(x,c):
    val_x,dx = x # x = [x,dx]
    return [c*val_x, c*dx]

def exponential(x):
    val_x,dx =  x # x = [x,dx]
    return [math.exp(val_x), math.exp(val_x) * dx]

def addition(x,y):
    val_x,dx = x
    val_y,dy = y
    return [val_x + val_y, dx + dy]

def Auto_Differentiation(x):
    u1 = power(x,2)
    u2 = multiplication(u1,3)
    u3 = exponential(x)
    u4 = addition(u2,u3)
    return u4

print(Auto_Differentiation([0.5,1]))
```

The actual value of the function and the derivative at $f(t)$ at $t = 1/2$ is 
$$f (t) = 3t^2 + e^t = \frac{3}{4} + e^{1/2} = 2.3987 ...$$
$$ f'(t) = 6t + e^t = 3 + e^{1/2} = 4.6487 ...$$

Look how exact we are!
