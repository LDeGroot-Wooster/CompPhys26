#### Plotting the trajectory of an object in freefall 

-  Create an array of 15 different times, equally spaced between 0 and 10 s
- Using a starting height of 500 m, create an array of heights of an object above the ground as a function of your time array 
- plot height vs. time
    - change the marker/line style 
- label all axes 
- title your graph
- add gridlines 

```python
t = np.linspace(1,10,15)

g = 9.81
h = 500 - (1/2)*g*(t**2)
plt.figure(1)
plt.plot(t,h,"bo")
plt.xlabel("time (s)")
plt.ylabel("height (m)")
plt.title("Height of object in free fall")
plt.grid()
z=t**2
plt.figure(2)
plt.plot(t,h,"bo")
plt.plot(t,z,"m:")
plt.show()
```

## Scatter Plots 


#### color-magnitude diagram 

- Download and import the "stars.txt" data 
- Make a scatter plot of magnitude (second column of data) vs. temperature (first column of data)
    - change the size and color of the points 
- label the axes 

```python
data1 = np.loadtxt("stars.txt",float)
x_data1 = data1[:,0]
y_data1 = data1[:,1]
plt.scatter(x_data1,y_data1,s=6,c="g")
plt.xlabel("Temp (K)")
plt.ylabel("Magnitude")
plt.show
```

##  Adding customizations to scatter plots 

#### Adding a legend 
- Create 3 different arrays:
    - an array of x values 
    - two arrays (y and z) that each depend on x
- Create a single scatter plot: 
    - y vs. x, called Data 1
    - z vs. x, called Data 2
- Make sure they are markers and different colors 
- Add a legend to the plot

```python
%matplotlib ipympl

x = np.linspace(1,100,20)
y = 2*(x**(1/3))
z = x/4

plt.scatter(x,y,c="k",label="Data 1")
plt.scatter(x,z,c="r",label="Data 2")
plt.legend(fontsize=10)
```

#### Adding error bars  
 
 - For the plot below, add error bars on the y-axis of a consistent value
     - make sure you still have 'circuit.txt' in your directory


```python
data = np.loadtxt("circuit.txt",float)
voltage = data[:,0]
current = data[:,1]

voltage_err = np.ones(len(voltage))*15
plt.scatter(current,voltage)
plt.errorbar(current,voltage,yerr=voltage_err,fmt = "None", capsize = 2)
```
