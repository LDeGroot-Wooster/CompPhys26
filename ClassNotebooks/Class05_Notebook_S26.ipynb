#### Plotting a numpy array 
- Create a numpy array `t` of 40 points equally spaced between 0 and 20. 
- Create a second array of the equation: 
$$ N(t) = N_0 e^{-\lambda t} $$
- With $N_0 = 5.0, \lambda = 0.5$
- Make a plot of `N` vs. `t`

```python
t = np.linspace(0, 20, 40)
N = []
for i in t:
    N.append(float(5 * (math.e ** (-0.5 * i))))

plt.plot(t, N, marker = 'o')
plt.show()
```

#### Plotting the trajectory of an object in freefall 

-  Create an array of 15 different times, equally spaced between 0 and 10 s
- Using a starting height of 500 m, create an array of heights of an object above the ground as a function of your time array 
- plot height vs. time
    - change the marker/line style 
- label all axes 
- title your graph
- add gridlines 

```python
t = np.linspace(0, 10, 15)
h = []
h_mars = []
for i in t:
    h.append(float(500 - (0.5 * 9.81 * (i ** 2))))
    h_mars.append(float(500 - (0.5 * 3.721 * (i ** 2))))
    

plt.plot(t, h, marker = 'o')
plt.plot(t, h_mars, marker = 'o')
plt.title('Falling Objects on Earth and Mars')
plt.xlabel('Time (seconds)')
plt.ylabel('Height (meters)')
plt.grid()
plt.show()
```

Script creates an array of times at a regular interval, then uses those to compute two lists of an object's height over time for differing gravitational accelerations. These are then plotted, the axes labeled, the plot given a title

##  Adding customizations to scatter plots 

#### Adding a legend 
- Create 3 different arrays:
    - an array of x values 
    - two arrays (y and z) that each depend on x
- Create a single scatter plot: 
    - y vs. x, called Data 1
    - z vs. x, called Data 2
- Make sure they are markers and different colors 
- Add a legend to the plot

```python
x = np.linspace(0, 100, 101)
y = []
z = []
for i in x:
    y.append(float((2 * i) + 14))
    z.append(math.log(i + 1)*30)
plt.scatter(x, y, 1, color = 'red')
plt.scatter(x, z, 1, color = 'blue')
plt.legend(['Data 1', 'Data 2'], loc = 'lower right')
plt.show()
```

#### Adding error bars  
 
 - For the plot below, add error bars on the y-axis of a consistent value
     - make sure you still have 'circuit.txt' in your directory


```python
x = np.linspace(0, 100, 101)
y = []
z = []
for i in x:
    y.append(float((2 * i) + 14))
    z.append(math.log(i + 1)*30)
plt.scatter(x, y, 2, color = 'red')
plt.scatter(x, z, 2, color = 'blue')
plt.errorbar(x, y, yerr = 5, elinewidth = 1, color = 'red')
plt.errorbar(x, z, yerr = 5, elinewidth = 1, color = 'blue')
plt.legend(['Data 1', 'Data 2'], loc = 'lower right')
plt.show()
```

#### Changing the marker size based on array data  
- For the plot below:
    - create an array of different values
        - make sure the array is the same length as the each data set being plotted
    - make the size of the markers equal to that array 

```python
x = np.linspace(0, 1000, 101)
t = np.linspace(0, 0, 101)
plt.scatter(x, t, s = (1 + (0.01 * (x ** 1.2))), color = 'blue')
plt.show()
```
