## Scatter Plots 


#### color-magnitude diagram 

- Download and import the "stars.txt" data 
- Make a scatter plot of magnitude (second column of data) vs. temperature (first column of data)
    - change the size and color of the points 
- label the axes 

```python
%matplotlib inline
star_data = np.loadtxt("stars.txt", float)
temp = star_data[:,0]
mag = star_data[:,1]
plt.figure(6)
plt.scatter(temp, mag, c="y", s=5)
plt.xlabel("Temperature (K)")
plt.ylabel("Magnitude")
plt.title("Big scatter plot")
plt.show()
```

##  Adding customizations to scatter plots 

#### Adding a legend 
- Create 3 different arrays:
    - an array of x values 
    - two arrays (y and z) that each depend on x
- Create a single scatter plot: 
    - y vs. x, called Data 1
    - z vs. x, called Data 2
- Make sure they are markers and different colors 
- Add a legend to the plot

```python
import numpy as np

x = np.linspace(0,20,500)
y = 5*np.sin(2*x)
z = 3*np.cos(x*3.2)
plt.figure(6)
plt.scatter(x,y, s=5, label="Data 1")
plt.scatter(x,z, s=5, label="Data 2")
plt.scatter(y,z, s=5, label="Data 3")
plt.ylim(-7,7)

plt.show()
```

#### Adding error bars  
 
 - For the plot below, add error bars on the y-axis of a consistent value
     - make sure you still have 'circuit.txt' in your directory


```python
#%matplotlib ipympl
data = np.loadtxt("circuit.txt", float)
voltage = data[:,0]
current = data[:,1]

voltage_err = np.ones(len(voltage))*15

plt.scatter(current,voltage)
plt.errorbar(current,voltage,yerr=voltage_err,fmt="None",capsize=2)
plt.show()
```
