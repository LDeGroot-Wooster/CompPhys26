#### Plotting a numpy array 
- Create a numpy array `t` of 40 points equally spaced between 0 and 20. 
- Create a second array of the equation: 
$$ N(t) = N_0 e^{-\lambda t} $$
- With $N_0 = 5.0, \lambda = 0.5$
- Make a plot of `N` vs. `t`

```python
Time_List = np.linspace(0,20,40)
N0 = 5.0
Lambda = 0.5

def N_function(T):
    return N0*m.e**(-Lambda*T)
N_List = N_function(Time_List)
plt.plot(Time_List, N_List )
```

#### Plotting the trajectory of an object in freefall 

-  Create an array of 15 different times, equally spaced between 0 and 10 s
- Using a starting height of 500 m, create an array of heights of an object above the ground as a function of your time array 
- plot height vs. time
    - change the marker/line style 
- label all axes 
- title your graph
- add gridlines 

```python
Time_Array = np.linspace(0,10,15) #seconds
Height_naught = 500 #meters
Gravity_Earth = 9.81
def Height_function(t):
    return Height_naught - Gravity_Earth*(1/2)*(t**2)
Height_Time_Array = Height_function(Time_Array)

plt.plot(Time_Array, Height_Time_Array, "--co")
plt.xlabel("Height (m)")
plt.ylabel("Time (s)")
plt.title("Ball in freeFALL!")
```
