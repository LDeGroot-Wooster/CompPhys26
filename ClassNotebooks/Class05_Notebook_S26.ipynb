#### Plotting a numpy array 
- Create a numpy array `t` of 40 points equally spaced between 0 and 20. 
- Create a second array of the equation: 
$$ N(t) = N_0 e^{-\lambda t} $$
- With $N_0 = 5.0, \lambda = 0.5$
- Make a plot of `N` vs. `t`

```python
Time_List = np.linspace(0,20,40)
N0 = 5.0
Lambda = 0.5

def N_function(T):
    return N0*m.e**(-Lambda*T)
N_List = N_function(Time_List)
plt.plot(Time_List, N_List )
```

#### Plotting the trajectory of an object in freefall 

-  Create an array of 15 different times, equally spaced between 0 and 10 s
- Using a starting height of 500 m, create an array of heights of an object above the ground as a function of your time array 
- plot height vs. time
    - change the marker/line style 
- label all axes 
- title your graph
- add gridlines 

```python
Time_Array = np.linspace(0,10,15) #seconds
Height_naught = 500 #meters
Gravity_Earth = 9.81
Mass_Ball = 1.0
def Height_function(t):
    return Height_naught - Gravity_Earth*(1/2)*(t**2)
Height_Time_Array = Height_function(Time_Array)

def Z_Kinetic_function(h):
    return Height_naught*Gravity_Earth*Mass_Ball - Mass_Ball * Gravity_Earth * h

Z_Kinetic_Array = Z_Kinetic_function(Height_Time_Array)
plt.figure(1)
plt.plot(Time_Array, Height_Time_Array, "--co")
plt.xlabel("Time (s)")
plt.ylabel("Height (m)")
plt.title("Ball in freeFALL!")

plt.figure(2)
plt.plot(Time_Array, Z_Kinetic_Array, "-ro")
plt.xlabel("Time (s)")
plt.ylabel("Energy (J)")
plt.title("Kinetic Energy!")
```

##  Adding customizations to scatter plots 

#### Adding a legend 
- Create 3 different arrays:
    - an array of x values 
    - two arrays (y and z) that each depend on x
- Create a single scatter plot: 
    - y vs. x, called Data 1
    - z vs. x, called Data 2
- Make sure they are markers and different colors 
- Add a legend to the plot

```python
X_array = np.linspace(0,50*m.pi, 1000)
Y_array = np.sin(X_array/2)
Z_array = m.pi/2 - np.arctan2(X_array,Y_array)
plt.scatter(X_array,Y_array, label="Sinusoidal Data")
plt.scatter(X_array, Z_array, label="Unknown Data")
plt.legend(fontsize=10)
plt.ylim(-1.1,1.1)
```

#### Adding error bars  
 
 - For the plot below, add error bars on the y-axis of a consistent value
     - make sure you still have 'circuit.txt' in your directory


```python


Circuitry_Array = np.loadtxt("circuit.txt", float)
Voltage_Array = Circuitry_Array[:,0]
Current_Array = Circuitry_Array[:,1]

Voltage_Unc_Array = np.ones(len(Voltage_Array))*(100 * max(Voltage_Array)/Voltage_Array)

plt.scatter(Current_Array, Voltage_Array)
plt.errorbar(Current_Array,Voltage_Array,yerr=Voltage_Unc_Array, fmt="None", capsize=2)
```
