#### Plotting a numpy array 
- Create a numpy array `t` of 40 points equally spaced between 0 and 20. 
- Create a second array of the equation: 
$$ N(t) = N_0 e^{-\lambda t} $$
- With $N_0 = 5.0, \lambda = 0.5$
- Make a plot of `N` vs. `t`

```python

t = np.linspace(0,20,40)

N0 = 5.0
lda = 0.5


Nt = N0*np.exp(-1*lda*t)

plt.plot(t,Nt)



```

#### Plotting the trajectory of an object in freefall 

-  Create an array of 15 different times, equally spaced between 0 and 10 s
- Using a starting height of 500 m, create an array of heights of an object above the ground as a function of your time array 
- plot height vs. time
    - change the marker/line style 
- label all axes 
- title your graph
- add gridlines 

```python


t = np.linspace(0,10,15)

H0 = 500 #initial height in m
a = 9.81 #m/s^2

Ht = H0 - 0.5*a*t**2
plt.figure(1)
plt.plot(t,Ht,'kH:')
plt.xlabel('Time (s)')
plt.ylabel('Height (m)')
plt.title('Height of a falling object without drag')
plt.grid()

plt.show()






#zarray


z = (H0/2)*np.cos(t)+(H0/2)



plt.figure(2)
plt.plot(t,Ht,'kH:')
plt.plot(t,z,'bh')
plt.xlabel('Time (s)')
plt.ylabel('Height (m)')
plt.title('Height of a falling object without drag')
plt.grid()

plt.show()






```

## Scatter Plots 


#### color-magnitude diagram 

- Download and import the "stars.txt" data 
- Make a scatter plot of magnitude (second column of data) vs. temperature (first column of data)
    - change the size and color of the points 
- label the axes 

```python
stars = np.loadtxt('stars.txt')
magnitude = stars[:,1]
temperature = stars[:,0]


plt.scatter(temperature,magnitude,c='m',s=0.3)
plt.xlabel('Temperature (K)')
plt.ylabel('Magnitude')
plt.title('H-R Diagram')
plt.show()
```

##  Adding customizations to scatter plots 

#### Adding a legend 
- Create 3 different arrays:
    - an array of x values 
    - two arrays (y and z) that each depend on x
- Create a single scatter plot: 
    - y vs. x, called Data 1
    - z vs. x, called Data 2
- Make sure they are markers and different colors 
- Add a legend to the plot

```python

xarr = np.linspace(0,10,500)
yarr = np.sin(xarr)
zarr = np.tan(xarr)


plt.scatter(xarr,yarr,c='r',s=1,label='Y Array')
plt.scatter(xarr,zarr,c='b',s=1,label='Z Array')
plt.ylim(-2,2)

plt.legend(fontsize=12)
plt.annotate('This graph is very cool! :)',xy=(5,0),fontsize=10)

plt.show()



```

#### Adding error bars  
 
 - For the plot below, add error bars on the y-axis of a consistent value
     - make sure you still have 'circuit.txt' in your directory


```python


data = np.loadtxt('circuit.txt')

voltage = data[:,0]
current = data[:,1]

voltage_err = np.ones(len(voltage))*15


plt.scatter(current,voltage,c='m',s=10)
plt.xlabel('Current (A)')
plt.ylabel('Voltage (V)')

plt.errorbar(current,voltage,yerr=voltage_err,fmt='None',capsize = 2,c='m')

plt.show()







```
