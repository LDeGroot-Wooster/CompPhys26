#### Plotting a numpy array 
- Create a numpy array `t` of 40 points equally spaced between 0 and 20. 
- Create a second array of the equation: 
$$ N(t) = N_0 e^{-\lambda t} $$
- With $N_0 = 5.0, \lambda = 0.5$
- Make a plot of `N` vs. `t`

```python
# create time array
t = np.linspace(0., 20., 40)
# calculate N
y = 5.0*np.exp(-0.5*t)
# plot N v. t
plt.plot(t,y)
plt.show()
```

#### Plotting the trajectory of an object in freefall 

-  Create an array of 15 different times, equally spaced between 0 and 10 s
- Using a starting height of 500 m, create an array of heights of an object above the ground as a function of your time array 
- plot height vs. time
    - change the marker/line style 
- label all axes 
- title your graph
- add gridlines 

```python
#constants
g = 9.81 #gravitational acceleration
h = 500 #starting height
#time array
t = np.linspace(0.,10., 15)

#compute current location
y = h - 0.5*g*t**2

#plot the data
plt.figure(1)
plt.plot(t,y,"bo-")
#plt.grid()
plt.xlabel("time (s)")
plt.ylabel("height (m)")
plt.title("Falling Object")

#create z array
z = t**2
plt.figure(2)
plt.plot(t, z, "k^:")
zz = t**3
plt.plot(t, zz, "rD:")

plt.show()
```

## Scatter Plots 


#### color-magnitude diagram 

- Download and import the "stars.txt" data 
- Make a scatter plot of magnitude (second column of data) vs. temperature (first column of data)
    - change the size and color of the points 
- label the axes 

```python
#read in data
star_data = np.loadtxt("stars.txt", float)
temp = star_data[:,0]
mag = star_data[:,1]

#make scatter plot
plt.scatter(temp, mag, c="r", s=1)
plt.xlabel("Temperature (K)")
plt.ylabel("Magnitude")
plt.show()
```

##  Adding customizations to scatter plots 

#### Adding a legend 
- Create 3 different arrays:
    - an array of x values 
    - two arrays (y and z) that each depend on x
- Create a single scatter plot: 
    - y vs. x, called Data 1
    - z vs. x, called Data 2
- Make sure they are markers and different colors 
- Add a legend to the plot

```python
x = np.linspace(0,20,500)
y = 5*np.sin(x*2)
z = 3*np.cos(x*3.2)

#plaot the data as a scatter plot
plt.scatter(x,y, s=10, label="Data 1")
plt.scatter(x,z, s=10, label="Data 2")
plt.legend(fontsize=12)
plt.ylim(-7,7)
plt.annotate("[as]", xy=(15,6), fontsize=12)

plt.show()
```

#### Adding error bars  
 
 - For the plot below, add error bars on the y-axis of a consistent value
     - make sure you still have 'circuit.txt' in your directory


```python
data = np.loadtxt("circuit.txt", float)
voltage = data[:,0]
current = data[:,1]

#create uncertainty array
voltage_err = np.ones(len(voltage))*15

#plot the data
plt.scatter(current,voltage)
plt.xlabel("current (A)")
plt.ylabel("Voltage (V)")

#add error bars
plt.errorbar(current, voltage, yerr=voltage_err, fmt="None", capsize=2)

plt.show()
```
