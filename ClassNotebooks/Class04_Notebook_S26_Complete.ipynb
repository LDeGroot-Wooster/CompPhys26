#### Emission lines of hydrogen (example 2.7 - page 73)

Rydberg formula:
$$ \frac{1}{\lambda} = R \left( \frac{1}{m^2} - \frac{1}{n^2} \right) $$

where $ R = 1.097 \times 10^{7}$ $\text{m}^{-1}$

Calculate the 5 longest wavelengths (in nm) for m = 1 (Lyman), m = 2 (Balmer), and m = 3 (Paschen)

```python
# define the rydberg constant 
R = 1.097e-2 # m^-1

# set up nested loop to determine 5 longest wavelengths per lower state (m)
for m in np.arange(1,4,1,float): # want m = 1, 2, 3
    for n in np.linspace(m+1, m+5,5): # define 5 n value swlarger than m 
        wavelength = 1/(R*(1/m**2 - 1/n**2))
        print("for m =", m, "and n =", n, ", wavelength =", round(wavelength,3), "nm")
```

### List comprehension

Typing out a list in brackets is not always the best or quickest way to create a list. Besides some of the methods we have talked about, `range()` for example, you can use an equation combined with a for loop and range to create a list:

`squares = [x**2 for x in range(10)]`

```python
squares = [x**2 for x in range(10)]
print(squares)

short_squares = [q for q in squares if q%2 == 0]
print(short_squares)
```

#### Defining a function

- Let's create a user-defined function that accomplishes the following:
    - the function requires two parameters: a & b
    - if a < b, the function returns the product of the two
    - if a >= b, then the function returns the results of a/b

```python
def example_function(a,b):
    if a < b:
        return a*b
    else:
        return a/b


print(example_function(3,2))

c = example_function(6,7)
print(c)
```

- Evaluate the factorial code from the book

```python
def factorial(n):
    f = 1.0
    for k in range(1,n+1):
        f *= k
    return f

factorial_val = factorial(100)
print(factorial_val)
```

- Write a function to do the following:

***Whichever you choose...***

```python
# define our function 
def net_vector2(ax, ay, az):
    mag = math.sqrt(ax**2 + ay**2 + az**2) # calculate the magnitude using the three components 
    return mag

# ask for user input of the 3 differnet vector components 
x_comp = float(input("Please give x-component of vector:"))
y_comp = float(input("Please give y-component of vector:"))
z_comp = float(input("Please give z-component of vector:"))

# call function 
net_force = net_vector2(x_comp, y_comp, z_comp)
# print out results 
print("the magnitude net force is ", net_force)
```

```python
# second option - however, missing the checks on the angle (see math.atan comment below )

import math 

# define our function 
def net_force(F1, theta1, F2, theta2):
    F1x = F1*math.cos(math.radians(theta1))
    F1y = F1*math.sin(math.radians(theta1))
    F2x = F2*math.cos(math.radians(theta2))
    F2y = F2*math.sin(math.radians(theta2))
    mag = math.sqrt((F1x+F2x)**2+(F1y+F2y)**2)
    print("Fx = ", F1x+F2x, ", Fy = ", F1y+F2y)
    direc = math.degrees(math.atan((F1y+F2y)/(F1x+F2x)))
    return(mag, direc)

Force1 = float(input("Please give the magnitude of force 1:"))
angle1 = float(input("Please give the angle relative to the +x axis of force in degrees:"))
Force2 = float(input("Please give the magnitude of force 2:"))
angle2 = float(input("Please give the angle relative to the +x axis of force in degrees:"))


net_force_val, net_force_dir = net_force(Force1, angle1, Force2, angle2)
print("the magnitude net force is ", net_force_val, "and the direction is ", net_force_dir, "degrees relative to the +x-axis")
```

## Recursion

```python
def factorial(n):
    if n==1:
        return 1
    else: 
        return n*factorial(n-1)
    

print(factorial(4))
```
