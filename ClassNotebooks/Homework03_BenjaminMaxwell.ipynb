```python
# Use this cell to import all necessary packages for the assignment 
import math
import numpy as np
import matplotlib.pyplot as plt
```

### Problem 1: Given

```python
#Defining constants
r0 = 0.450 #meters
Q = 0.250 #nC
pi = math.pi
e0 = 8.854*(10**-12) #F/m or C^2/Nm^2
# Creating an array r of values ranging from 0 to 3r with at least 100 points. 
radius = np.linspace(0, 3*r0, 500)
ErNoCon =[]
ErCon = []
# Creatig an array of electric field magnitudes for both the non-conducting and conducting sphere
#Non conducting sphere array
def electric_nonconduct(r):
    for i in range(len(radius)):
        if r[i] > r0:
            ErNoCon.append((1/(4*pi*e0))*(Q/(r[i]**2)))
        elif r[i] < r0:
            ErNoCon.append((1/(4*pi*e0))*(Q/(r0**3))*r[i])
#Conductig sphere array
def electric_conduct(r):
    for i in range(len(radius)):
        if r[i] > r0:
            ErCon.append((1/(4*pi*e0))*(Q/(r[i]**2)))
        elif r[i] < r0:
            ErCon.append(0)
#Plotting Er vs radius for nonconduct
plt.figure(1)
plt.imshow(radius, ErNoCon, origin="lower")
#Labels + Legend
plt.xlabel("Radius")
plt.ylabel("Electric Field for a nonconducting sphere")
plt.nipy_spectral()
plt.colorbar()
#Error Bars
plt.errorbar(radius, ErNoCon,yerr=np.random.rand(0.5,1), fmt="None", capsize=2)
#Plotting for conduct
plt.figure(2)
plt.imshow(radius,ErCon, origin="lower")
#Labels + Legend
plt.xlabel("Radius")
plt.ylabel("Electric Field for a conducting sphere")
plt.nipy_spectral()
plt.colorbar()
#Error Bars
plt.errorbar(radius,ErCon,yerr=np.random.rand(0.5,1), fmt="None", capsize=2)

#Show my plots
plt.show()
```
