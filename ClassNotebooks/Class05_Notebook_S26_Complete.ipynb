#### Plotting a numpy array 
- Create a numpy array `t` of 40 points equally spaced between 0 and 20. 
- Create a second array of the equation: 
$$ N(t) = N_0 e^{-\lambda t} $$
- With $N_0 = 5.0, \lambda = 0.5$
- Make a plot of `N` vs. `t`

```python
# create time array 
t = np.linspace(0., 20., 40)
# calculate N
N = 5.0*np.exp(-0.5*t)
#print(N)
plt.plot(t,N)
plt.show()
```

#### Plotting the trajectory of an object in freefall 

-  Create an array of 15 different times, equally spaced between 0 and 10 s
- Using a starting height of 500 m, create an array of heights of an object above the ground as a function of your time array 
- plot height vs. time
    - change the marker/line style 
- label all axes 
- title your graph
- add gridlines 

```python
# define an array of time for calculation
t = np.linspace(0., 10., 15)
#set value of gravitational acceleration 
g = 9.81 # m/s^2
#get initial height from user
#h = float(input("Enter initial height of ball (m):"))
h = 500

# compute location of the ball
y = h - 0.5*g*t**2

# plot data
plt.figure(1)
plt.plot(t,y,'o')
plt.xlabel('time (s)')
plt.ylabel('height (m)')
plt.title('Falling Ball')
#plt.grid()
plt.ylim(0,505)
         
plt.figure(2)
z = t*t
plt.plot(t,z,'y^:') # yellow triangle markers connected with dotted lines
zz = t*t*t
plt.plot(t,zz, 'rD')
plt.show()
        
```

##  Adding customizations to scatter plots 

#### Adding a legend 
- Create 3 different arrays:
    - an array of x values 
    - two arrays (y and z) that each depend on x
- Create a single scatter plot: 
    - y vs. x, called Data 1
    - z vs. x, called Data 2
- Make sure they are markers and different colors 
- Add a legend to the plot

```python
x = np.linspace(0,20,500)
y = 5*np.sin(x*2)
z = 3*np.cos(x*3.2)
plt.scatter(x,y, label="Data 1")
plt.scatter(x,z, label="Data 2")
plt.legend(fontsize=12)
plt.ylim(-7,7)
plt.annotate("This is a graph!", xy=(15,6), fontsize=12)

plt.show()
```

#### Adding error bars  
 
 - For the plot below, add error bars on the y-axis of a consistent value
     - make sure you still have 'circuit.txt' in your directory


```python
data = np.loadtxt("circuit.txt", float)
voltage = data[:,0]
current = data[:,1]

# create uncertainty array 
voltage_err = np.ones(len(voltage))*15

# plot the data 
plt.scatter(current, voltage)
plt.xlabel("current (A)")
plt.ylabel("voltage (V)")
# add error bars 
plt.errorbar(current, voltage, yerr=voltage_err, fmt='None', capsize=2)

plt.show()
```

#### Extra examples of interactive plotting - with a widget

```python
from ipywidgets import *
#import ipywidgets as widgets
#%matplotlib inline
%matplotlib ipympl 


x = np.linspace(0, 2 * np.pi)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
line, = ax.plot(x, np.sin(x))

def update(w = 1.0):
    line.set_ydata(np.sin(w * x))
    fig.canvas.draw()
    #fig.canvas.draw_idle()
    

interact(update);
```
