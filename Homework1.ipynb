```python
#2.6
import math
M = (1.9891*(10**30))
G = (6.6738*(10**-11))
l1 = float(input("What is the distance to the sun at perhelion? "))
v1 = float(input("What is the velocity at perhelion? "))
a = 1
b = -(2*G*M)/(v1*l1)
c = -((v1**2)-((2*G*M)/l1))
v2 = ((-b - math.sqrt(((b**2)-4*a*c)))/2*a)
l2 = (l1*v1)/v2
aa = (0.5*(l1+l2))
bb = math.sqrt(l1*l2)
t_s = (2*math.pi*aa*bb)/(l1*v1)
t_y = t_s/(365.25*24*3600)
e = (l2-l1)/(l2+l1)
print(l2)
print(v2)
print(t_y)
print(e)
```
