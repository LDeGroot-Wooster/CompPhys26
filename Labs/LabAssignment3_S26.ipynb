#### Find the sum of odds

Write a code to add all the odd numbers from 1 to $N$. 
- The function should return the sum *and* the time in milliseconds needed to perform this sum. (Google is your friend!)
- Use your function to report the sum of all the odd integers from 1 to 1,000,000 (inclusive) as well as how long it took to do this.

When your program is complete, add a markdown cell below it to explain your process and program. 

```python



def sum1N(N):
    sum = 0
    for i in range(N+1):
        if i%2 == 1:
            sum+= i
    return sum

start_time = time.time()
print(sum1N(1000000))
print('--- %s milliseconds ---' % ((time.time() - start_time)*1000))




```

To sum the values, I used a for loop with range(N+1) to count up from 1 to N. I determined if a number i was odd with i mod 2 == 1. I added the odd numbers to the sum variable, which starts at 0. Then, I returned the sum at the end of the loop.

To couint the time, I used the python time library and the time.time() command, which records the current time when it is called. Before calling the function, I stored the start time, and then I called the function by printing the sum, and then I printed the current time using a code I found by rogeriopvl on StackOverflow. The equation for standard deviation is given by: 
$$ \sigma = \left[ \frac{1}{N} \sum\limits_{i = 1}^{N} \left( a_i -\overline{a} \right)^2 \right]^{1/2} $$
where $\overline{a}$ is the average value of the array, and $a_i$ is each element of the array of length $N$. 
In the cell below, write a function to calculate the standard deviation for an input array, a:
- First compute the average of the elements in a to define $\overline{a}$ [hint: there is a function called np.mean()] **I have provided the array a for you - be sure to understand what it does**
- Next compute the sum over the squares of $a - \overline{a}$
- Then divide the sum by the number of elements in the array
- Finally take the square root (you can use np.sqrt())
Test your function on a random array, and compare to the built-in np.std() function

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
# first create input array a
a = np.random.random_sample(20)*20 #returns 20 random floats in [0.0,1.0), each multiplied by 20.
print(a)

avg = np.mean(a)
squaresdiff = (a-avg)**2
sumoversquares = np.sum(squaresdiff)
stdev = np.sqrt(sumoversquares/np.size(a))
print('Manual stdev =',stdev)

print('np.std result =',np.std(a))
```

In this program, I started by taking the mean of the array a, then I created a new array called squaresdiff that contains the squares of the differences between each element of a and the mean. Then, I summed every element in the squaresdiff array, and calculated the standard deviation by dividing the sum by the size of a.
