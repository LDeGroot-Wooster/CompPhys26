#### Find the sum of odds

Write a code to add all the odd numbers from 1 to $N$. 
- The function should return the sum *and* the time in milliseconds needed to perform this sum. (Google is your friend!)
- Use your function to report the sum of all the odd integers from 1 to 1,000,000 (inclusive) as well as how long it took to do this.

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
startTime = datetime.now()
def oddsum(N):
    sums = 0
    for i in range(N//2):
        odds = (2*i) + 1
        sums += odds
    return(sums)

print(oddsum(1000000))

print(datetime.now() - startTime)
```

To sum all the odds, I defined a function oddsum with parameter N. Then I ran a for loop over that runs N/2 times rounded to an integer after defining the odds using 2i+1. I defined sums as 0 before the loop, then added sums = sums + odds to update sums. This constantly updates until the for loop ends, summing all the odds up to the current i every time. After returning sums, I can print the result for the desired N = 1,000,000. 
To deterimine the time, I imported the datetime package and used the function datetime.now at the start and end of the program. Then, taking the difference gives the time that the program runs.
