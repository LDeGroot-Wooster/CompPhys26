#### Find the sum of odds

Write a code to add all the odd numbers from 1 to $N$. 
- The function should return the sum *and* the time in milliseconds needed to perform this sum. (Google is your friend!)
- Use your function to report the sum of all the odd integers from 1 to 1,000,000 (inclusive) as well as how long it took to do this.

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
def odd_sum(N):
    start_time = time.time() #initializing a time
    
    final_sum = 0 #initializing a variable for the numbers to be added to
    sum_array = np.linspace(1,N,N) # makes sure it's inclusive
    for i in sum_array:
        if i%2 == 1:
            final_sum += i
    
    final_time = (time.time() - start_time) * 1000 # difference in time with a conversion to milliseconds
    return int(final_sum), final_time #print stetement looks weird if the final sum is not specified to be an integer

finalSum, finalTime = odd_sum(5)
print("The program took", finalTime,"milliseconds to find the sum of odd numbers to be", finalSum)
```

I started off with figuring out how to return the sum first! I initialized a variable for the numbers to be added into and created an array for *all* the integers within the range from 1 to N. I then used a for loop to iterate through each entry in my array and checked to see of each entry was odd or not using an if statement and a modulo operator. If the entry is odd, it gets added to the variable for the whole sum.\
To keep track of how long the program took to run, I imported the time module above and took a mark of the time at the beginning of the function.
