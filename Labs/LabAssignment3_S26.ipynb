#### Find the sum of odds

Write a code to add all the odd numbers from 1 to $N$. 
- The function should return the sum *and* the time in milliseconds needed to perform this sum. (Google is your friend!)
- Use your function to report the sum of all the odd integers from 1 to 1,000,000 (inclusive) as well as how long it took to do this.

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
#Create function for calculating the sum of odd integers
def ODD(N):
    m = 0 #Create variable for the sum
    for i in range(N+1): #Go through all integers from 1 to N
        if i % 2 == 1: #Tests to see if it is odd
            m = m + i #Adds integer to total
    return m #Returns sum

#Start Timer
start_time = time.perf_counter()
start_time2 = time.time()
#Do ODD Function from 1 to 1000000
print(ODD(1000000))

#End Timer
end_time = time.perf_counter()
end_time2 = time.time()
#Calculate Elapsed Time
elapsed_time = end_time - start_time
elapsed_time2 = end_time2 - start_time2

#Changing Time into Milliseconds
elapsed_time = 1000*elapsed_time
elapsed_time2 = 1000*elapsed_time2
#Printing!
print("Execution time:", elapsed_time, "or", elapsed_time2, "milliseconds")

```

I began by creating the function requested. The function I made uses the range function to go through all values from 1 to N (It says N+1 to include N). For each value, it is divided by 2, and if it has a remainder of 1 it is odd. Then if its odd, the number is add to the sum. Then, I start both of my timers using time.time and time.perf_counter, as I wanted to test both timers. I conducted my function at 1000000, and then ended both timers. I found the elapsed times by subtracting the end and the start time, and then multiplied the elapsed times by 1000 for milliseconds. Finally, I printed everything.

### Write a script 3 (25 pts) - extra numpy practice [if time permits]

The equation for standard deviation is given by: 
$$ \sigma = \left[ \frac{1}{N} \sum\limits_{i = 1}^{N} \left( a_i -\overline{a} \right)^2 \right]^{1/2} $$
where $\overline{a}$ is the average value of the array, and $a_i$ is each element of the array of length $N$. 
In the cell below, write a function to calculate the standard deviation for an input array, a:
- First compute the average of the elements in a to define $\overline{a}$ [hint: there is a function called np.mean()] **I have provided the array a for you - be sure to understand what it does**
- Next compute the sum over the squares of $a - \overline{a}$
- Then divide the sum by the number of elements in the array
- Finally take the square root (you can use np.sqrt())
Test your function on a random array, and compare to the built-in np.std() function

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
# first create input array a
a = np.random.random_sample(20)*20
print(a)

def stand_dev(a):
    #Finding average for all elements in a
    a_bar = sum(a)/len(a)
    #Finding the sum over the squares
    aa = sum((a-a_bar)**2)
    #Dividig sum by number of elements in the array.
    a_a = aa/len(a)
    #Take Sqrt
    o = np.sqrt(a_a)
    #Return the final value
    return o
#Printing!
print(stand_dev(a))
```

The code starts with an array which is given to me. Then to start, I defined my function as stand_dev, taking the input a. I proceed to find the average by taking the sum divided by the length. Then, found the sum of the squares of the array minus the average. I divided this sum by the number of elements in the array. Finall, I take the square root of these functions and returned the value I found. Then I printed everything.
