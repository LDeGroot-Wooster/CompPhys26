#### Find the sum of odds

Write a code to add all the odd numbers from 1 to $N$. 
- The function should return the sum *and* the time in milliseconds needed to perform this sum. (Google is your friend!)
- Use your function to report the sum of all the odd integers from 1 to 1,000,000 (inclusive) as well as how long it took to do this.

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
def sum_odd(N):
    x = 0
    for i in np.linspace(1, N, N):
        if i / 2 != round(i / 2):
            x += i
    return(x)

t_i = time.time()
y = sum_odd(1000000)
t_f = time.time()
delta_t_ms = (t_f - t_i) * 1000
print('The sum of all odd numbers from 1 to 1,000,000 is:', y)
print('Calculation was performed in', delta_t_ms, 'milliseconds')
```

The defined 'sum_odd' function creates a local variable equal to zero and a 1D array of all integers from 1 to N.For each integer in that array, it checks if that number is even or odd. If the number is odd, it adds it to the local variable. Once this process has been performed for each integer from 1 to N, the function returns the value of the local variable.
The time the function takes is measured here by taking an initial and final time, calculating the difference between them, and multiplying by 1000 to convert the time (which defaults to seconds) into milliseconds. The output of the function and the time it took to execute are printed.

The equation for standard deviation is given by: 
$$ \sigma = \left[ \frac{1}{N} \sum\limits_{i = 1}^{N} \left( a_i -\overline{a} \right)^2 \right]^{1/2} $$
where $\overline{a}$ is the average value of the array, and $a_i$ is each element of the array of length $N$. 
In the cell below, write a function to calculate the standard deviation for an input array, a:
- First compute the average of the elements in a to define $\overline{a}$ [hint: there is a function called np.mean()] **I have provided the array a for you - be sure to understand what it does**
- Next compute the sum over the squares of $a - \overline{a}$
- Then divide the sum by the number of elements in the array
- Finally take the square root (you can use np.sqrt())
Test your function on a random array, and compare to the built-in np.std() function

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
# first create input array a
a = np.random.random_sample(20)*20
print(a)
```

```python
def st_dev(arr):
    mean = np.mean(arr)
    x = 0
    for i in arr:
        x += (i - mean) ** 2
    return(np.sqrt(x / len(arr)))

print('Standard Deviation (user-defined):', st_dev(a))
print('Standard Deviation (NumPy function):', np.std(a))
```

The array provided contains 20 randomly generated values between 0 and 20.
The function calculates the mean of the input array, and stores it in a local variable, as well as creating another local variable 'x' equal to 0. Then, for each element of the array, the function subtracts the mean of the array from that element, then squares it and adds it to the local variable 'x'. Lastly, the function returns the square root of the local variable 'x' divided by the length of the array. This function gives the value for standard deviation of an array as the NumPy function.
