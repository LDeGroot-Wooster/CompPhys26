```python
start = time.time()

L = 200 # number of atoms in the solid 
M = 0.0 # Medelung constant starting value 

# loop over all directions in the solid 
for i in range(-L,L+1):
    for j in range(-L,L+1):
        for k in range(-L,L+1):
            if i==0 and j==0 and k==0: #condition for if at origin 
                continue
            elif (i+j+k)%2==0: # condition for if i*j*k position is positive 
                M += 1/np.sqrt(i*i+j*j+k*k)
            else: # condition for if i*j*k position is negative 
                M -= 1/np.sqrt(i*i+j*j+k*k)

print("M =",M)
end = time.time()
print(end-start)
```

####  Calculating integrals (adapted from textbook)

Like you did with a derivative, we can also use the definition of an integral to evaluate integrals with a computer. If we want to evaluate the integral 
 

$$ I = \int_{-1}^{1} \sqrt{1-x^2} \,dx, $$


we can divide the domain of the integration in to a larger number $N$ of slices of width $h = 2/N$ each. The Riemann definition can then be applied: 
$$ I = \lim_{N \to\infty} \sum_{k=1}^{N} hy_k, $$
where 
$$ y_k = \sqrt{1-x_k^2} \hspace{0.2in} \text{and}\hspace{0.2in} x_k = -1+hk.$$


We just need to make $N$ large to get a good approximation (since we can't in practice take the limit as $N\rightarrow \infty$). The expected value of this integral is $\pi/2$. 

Write a code to do the following: 
- Define a function `f2(x)` that returns the value of $\sqrt{1-x^2}.$  [this should only take you 2 lines]
- Define a second function `f_int(N)` that takes in in an N value and uses your `f2(x)` function to evaluate the integral. Have your function return both the result of the derivative and N (in that order).
- $N = 100$ and compare with the expected result. [perhaps a percent error makes a good comparison!]
- Increase the value of $N$ to make it more accurate. If the program must run in 1 second or less, how accurate can you get? 


When your program is complete, add a markdown cell below it to explain your process and program and include any comparisions. 

```python

start = time.time()
def f2(x):
    return np.sqrt(1-x**2)
def f_int(N):
    Sum = 0
    delta = 2/N
    x_k = -1
    for k in range(N):
        x_k += delta
        y_k = f2(x_k)
        Sum += delta*y_k
    return Sum
Input_N = 1000000
print(f_int(Input_N))
end = time.time()
print(end-start, "seconds")
print(np.pi/2)
print(np.abs(f_int(Input_N)-np.pi/2))
```
