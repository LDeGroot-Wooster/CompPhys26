- Edit the code so you can measure the time it takes to do the whole calculation. Then, time how long it takes the code to run 100 atoms, 1000 atoms, and 1 million. ($L$ is the number of atoms)

- Also comment on the difference in the result as the number of terms increases. The expected value is around -1.74699. 

```python
L = 10 # number of atoms in the solid 
M = 0.0 # Medelung constant starting value 

starttime = time.time()
# loop over all directions in the solid 
for i in range(-L,L+1):
    for j in range(-L,L+1):
        for k in range(-L,L+1):
            if i==0 and j==0 and k==0: #condition for if at origin 
                continue
            elif (i+j+k)%2==0: # condition for if i*j*k position is positive 
                M += 1/np.sqrt(i*i+j*j+k*k)
            else: # condition for if i*j*k position is negative 
                M -= 1/np.sqrt(i*i+j*j+k*k)
endtime = time.time()

print("M =",M)
print('The process for L =',L,'atoms took',endtime-starttime,'seconds')
```

L = 10: 0.025 seconds

L = 100: 15.5 seconds

For every time L increases by a factor of 10, the time increases by a factor of $10^3 = 1000$, so for $L = 1000$ would take approximately 15,000 seconds or 4 hours (on my computer), 1 million would take 475,000 years, and 1 billion would take 475 quadrillion years.

```python
number = int(input('Enter the value of n.'))


def factorial_int (n):
    for i in range(1,n+1):
        n = int(n*i)
    return n


def factorial_float(n):
    for i in range(1,n+1):
        n = float(n*i)
    return n


print('the factorial of your number as an integer is', factorial_int(number))
print('the factorial of your number as a floating-point value is', factorial_float(number))
```

Using the definition of a derivative below we can use a computer take the derivative of a function $f(x)$ at a point $x$:
$$ \frac{df}{dx} = \lim_{h \to 0} \frac{f(x+h)-f(x)}{h}$$

```python

h = 1e-2
x = 1


def f1(x):
    return x*(x-1)

def f_deriv(x,h):
    dfdx = (f1(x+h)-f1(x))/h
    return dfdx,h


print(f_deriv(x,h))




```

####  Calculating integrals (adapted from textbook)

Like you did with a derivative, we can also use the definition of an integral to evaluate integrals with a computer. If we want to evaluate the integral 
 

$$ I = \int_{-1}^{1} \sqrt{1-x^2} \,dx, $$


we can divide the domain of the integration in to a larger number $N$ of slices of width $h = 2/N$ each. The Riemann definition can then be applied: 
$$ I = \lim_{N \to\infty} \sum_{k=1}^{N} hy_k, $$
where 
$$ y_k = \sqrt{1-x_k^2} \hspace{0.2in} \text{and}\hspace{0.2in} x_k = -1+hk.$$


We just need to make $N$ large to get a good approximation (since we can't in practice take the limit as $N\rightarrow \infty$). The expected value of this integral is $\pi/2$. 

Write a code to do the following: 
- Define a function `f2(x)` that returns the value of $\sqrt{1-x^2}.$  [this should only take you 2 lines]
- Define a second function `f_int(N)` that takes in in an N value and uses your `f2(x)` function to evaluate the integral. Have your function return both the result of the derivative and N (in that order).
- $N = 100$ and compare with the expected result. [perhaps a percent error makes a good comparison!]
- Increase the value of $N$ to make it more accurate. If the program must run in 1 second or less, how accurate can you get? 


When your program is complete, add a markdown cell below it to explain your process and program and include any comparisions. 

```python
#Riemann sum: divide into N slices of width h = 2/N


import math

N = 3000000
h = 2/N



def f2(x):
    return math.sqrt(1 - x**2)


def f_int(N,h):
    I = 0
    for k in range(1,N+1):
        I += h*f2(-1 + h*k) #x_k = -1+hk
    return I,N

print(f_int(N,h))






```

Since $\sqrt{1-x^2}$ is a semicircle, the result of this integral analytically is $\pi/2\approx1.57079632$.
