### Animation #1 (22 pts)
#### A Growing Cone 

- Create a cone of some initial height and radius
- Make it grow to 5 times itâ€™s size  
- Use the appropriate rate to make this occur over 10 seconds
- Make the cone a color of your choice 

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
s4 = canvas(width = 400, height = 400, background = color.black)
s4.select()
i = 1
obj_cone = cone(pos = vec(0, -2.5, 0), axis = vec(0, 1, 0), radius = 0.5, length = 1)
while i < 300:
    rate(30)
    obj_cone.radius = 0.5 + (0.5 * (1 / 75) * i)
    obj_cone.length = 1 + ((1 / 75) * i)
    i += 1
```

Script creates a scene, and a cone at (0, -2.5, 0) pointing up, with a radius of 0.5 and a length of 1. Over 10 seconds, it increases the cone's radius and length to 5 times the original value.

### Animation #2 (22 pts)
#### Animating 3D objects - a bouncing ball

- Create a green sphere starting in the upper left corner of your scene 
- Make it zig-zag it across the scene from the upper left to the lower right, overall
- Have it cross the screen at least 5 times
- Have autoscale off

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
s5 = canvas(width = 400, height = 400, background = color.black, autoscale = False)
s5.select()
i = 0
ball = sphere(pos = vec(-8,8,0), radius = 1, color = color.green)
while i < 4:
    while ball.pos.x < 8:
        rate(30)
        ball.pos = ball.pos + (vec(1, (-8 / 56), 0) * 0.1)
    if i == 3:
        break
    while ball.pos.x > -8:
        rate(30)
        ball.pos = ball.pos + (vec(-1, (-8 / 56), 0) * 0.1)
    i += 1
```

The script creates a canvas and a green sphere at (-8, 8, 0) with radius 1. The script moves the sphere along the vector <1, -8/56, 0> until it reaches x = 8, then moves it along the vector <-1, -8/56, 0> until it reaches x = -8. It then repeats the process 2 more times. On the fourth iteration, it only performs the first action in this sequence, then breaks out of the loop.
