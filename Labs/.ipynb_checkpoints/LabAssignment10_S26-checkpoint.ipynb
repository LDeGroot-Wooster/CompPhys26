## Derivatives with front, back, and central difference (25 pts)

Let's evaluate the derivative of: 

$$ f(x) = \frac{1-\cos x}{x^2} $$

I don't want to do this by hand, so let's do it computationally. 

*Note: don't forget to put brief comments in your code*

#### Step 1: Define the function `f(x)` that returns the function $f(x)$ from above. 

```python
def f(x):
    return (1-np.cos(x))/(x**2)
```

#### Step 2: Define functions that call on your function from step 1 to calculate the front, back, and central difference first derivatives of $f(x)$ at a specific location $x$. 

```python
def fwd_diff(f,x,h):   #estimates derivative using forwards difference
    return (f(x+h)-f(x))/h

def bkwd_diff(f,x,h):   #estimates derivative using backwards difference
    return (f(x)-f(x-h))/h

def cent_diff(f,x,h):   #estimates derivative using central difference
    return (f(x+(h/2))-f(x-(h/2)))/h
```

#### Step 3: Evaluate the derivative using the three different approximations using the appropriate value of $h$ for each method evaluated from $x = 0.004$ to $x = \frac{\pi}{2}$ - let's do 20 calculations. 

```python
#define appropriate h-values for each extimate
h_fwd = 1e-8
h_bkwd = 1e-8
h_cent = 1e-5

#array of 20 values between 0.004 and pi/2
x_vals = np.linspace(0.004,(math.pi)/2,20)

#calculates fwd diff results for each x-value using h_fwd
fwd_res = []
for i in range(len(x_vals)):
    fwd_res.append(fwd_diff(f,x_vals[i],h_fwd))

#calculates bkwd diff results for each x-value using h_bkwd
bkwd_res = []
for i in range(len(x_vals)):
    bkwd_res.append(bkwd_diff(f,x_vals[i],h_bkwd))

#calculates cent diff results for each x-value using h_cent
cent_res = []
for i in range(len(x_vals)):
    cent_res.append(cent_diff(f,x_vals[i],h_cent))

#use formatting to print results of above calculations
print("x                   forward diff         backward diff        central diff")
rowf = "{0:1.16f}  {1:1.16f}  {2:1.16f}  {3:1.16f}"
for i in range(len(x_vals)):
    print(rowf.format(x_vals[i],fwd_res[i],bkwd_res[i],cent_res[i]))
```

#### Step 4: Determine the equation for the derivative (by hand and then type in below this cell as markdown). Computationally calculate the true value of the derivative at the range of values from step 3 and compare to your calculated values. Explain your results in a markdown cell. 

Using quotient rule, the true value of
$$\frac{d}{dx}\left(\frac{1-\cos{x}}{x^2}\right)$$
is
$$\frac{x^2\sin{x}+2x\cos{x}-2x}{x^4}$$

```python
#function to calculate true value of the derivative
def true_val(x):
    return (((x**2)*np.sin(x))+(2*x*np.cos(x))-(2*x))/(x**4)

#fcalculates true value of the derivative for each value of x
true_vals = []
for i in range(len(x_vals)):
    true_vals.append(true_val(x_vals[i]))

#calculates error for forward difference for each value of x
fwd_err = []
for i in range(len(x_vals)):
    fwd_err.append(np.abs(fwd_res[i]-true_vals[i]))

#calculates error for backward difference for each value of x
bkwd_err = []
for i in range(len(x_vals)):
    bkwd_err.append(np.abs(bkwd_res[i]-true_vals[i]))

#calculates error for central difference for each value of x
cent_err = []
for i in range(len(x_vals)):
    cent_err.append(np.abs(cent_res[i]-true_vals[i]))

#use formatting to print results of above calculations
print("x         true value      forward diff    fwd diff err     backward diff   bkwd diff err    central diff    cent diff err")
rowf = "{0:1.6f}  {1:1.11f}  {2:1.11f}  {3:1.13f}  {4:1.11f}  {5:1.13f}  {6:1.11f}  {7:1.13f}"
for i in range(len(x_vals)):
    print(rowf.format(x_vals[i],true_vals[i],fwd_res[i],fwd_err[i],bkwd_res[i],bkwd_err[i],cent_res[i],cent_err[i]))
```

The forward and backward difference errors are usually of the order 1e-9 to 1e-7. Lower values of x produce worse estimates and higher errors on average. The central difference error is usually of the order 1e-11 to 1e-12. For every value of x tested, central difference provides a better estimate of the true value of the derivative.

#### Step 5: Evalute the derivative using the three approximations with a value of $h=10^{-6}$. Compare (computationally and in markdown) the values with each other and the true value. 

```python
new_h = 1e-6

#calculates fwd diff results for each x-value using new_h
fwd_res = []
for i in range(len(x_vals)):
    fwd_res.append(fwd_diff(f,x_vals[i],new_h))

#calculates bkwd diff results for each x-value using new_h
bkwd_res = []
for i in range(len(x_vals)):
    bkwd_res.append(bkwd_diff(f,x_vals[i],new_h))

#calculates cent diff results for each x-value using new_h
cent_res = []
for i in range(len(x_vals)):
    cent_res.append(cent_diff(f,x_vals[i],new_h))

#updates errors based on new values
fwd_err = []
for i in range(len(x_vals)):
    fwd_err.append(np.abs(fwd_res[i]-true_vals[i]))

bkwd_err = []
for i in range(len(x_vals)):
    bkwd_err.append(np.abs(bkwd_res[i]-true_vals[i]))

cent_err = []
for i in range(len(x_vals)):
    cent_err.append(np.abs(cent_res[i]-true_vals[i]))

#use formatting to print results of above calculations
print("x         true value      forward diff    fwd diff err     backward diff   bkwd diff err    central diff    cent diff err")
rowf = "{0:1.6f}  {1:1.11f}  {2:1.11f}  {3:1.13f}  {4:1.11f}  {5:1.13f}  {6:1.11f}  {7:1.13f}"
for i in range(len(x_vals)):
    print(rowf.format(x_vals[i],true_vals[i],fwd_res[i],fwd_err[i],bkwd_res[i],bkwd_err[i],cent_res[i],cent_err[i]))
```

The forward and backward difference errors are usually of the order 1e-7. Again, lower values of x produce worse estimates and higher errors on average. The central difference error is usually of the order 1e-10 to 1e-11. For every value of x tested, central difference provides a better estimate of the true value of the derivative. Using h=1e-6 slightly worsened all estimates. For backward and forward difference that h value is too large, and for central difference it is too small.

## Second derivative (22 pts)

Let's take the second derivative of the function from above. 

*Note: don't forget to put brief comments in your code*

#### Step 1: Define functions that call on your function from step 1 above to calculate the second derivative of $f(x)$ at a specific location $x$ using central difference. 

```python
#function to calculate second derivative
def sec_diff(f,x,h):
    return (f(x+h)-2*f(x)+f(x-h))/(h**2)
```

#### Step 2: Evalute the second derivative at $x = 0.004$ using a range of $h$ values from `1e-1` to `1e-12`. 

```python
x = 0.004

#make list of values for h
h_vals = []
for i in range(1,12):
    h_vals.append(10**(-i))

#calculate values of second derivative for each h
sec_res = []
for i in range(len(h_vals)):
    sec_res.append(sec_diff(f,x,h_vals[i]))

#use formatting to print results of above calculations
print("h                second derivative")
rowf = "{0:1.13f}  {1:1.16f}"
for i in range(len(h_vals)):
    print(rowf.format(h_vals[i],sec_res[i]))
```

#### Step 3: Determine the equation for the derivative (by hand/WolframAlpha and then type in below this cell as markdown). Calculate the true value of the derivative at $x = 0.004$.  

Using quotient rule, the true value of
$$\frac{d^2}{dx^2}\left(\frac{1-\cos{x}}{x^2}\right)$$
is
$$\frac{d}{dx}\left(\frac{x^2\sin{x}+2x\cos{x}-2x}{x^4}\right)$$
which in turn is, after some simplification,
$$\frac{x^2\cos{x}-4x\sin{x}-6\cos{x}+6}{x^4}$$

```python
def f_prime_prime(x):
    return ((x**2*np.cos(x))-(4*x*np.sin(x))-(6*np.cos(x))+6)/(x**4)

true_val_sec = f_prime_prime(x)
print(true_val_sec)
```

#### Step 4: Computationally compare your calculated answer to the true value for each $h$ value. Include a markdown explaining the process and result.

```python
#calculates error for second dierivative for each value of x
sec_err = []
for i in range(len(h_vals)):
    sec_err.append(np.abs(sec_res[i]-true_val_sec))

#use formatting to print results of above calculations
print("h                sec deriv true value   sec deriv estimate       sec deriv error")
rowf = "{0:1.13f}  {1:1.16f}    {2:1.16f}       {3:1.16f}"
for i in range(len(h_vals)):
    print(rowf.format(h_vals[i],true_val_sec,sec_res[i],sec_err[i]))
```

#### Explanation:

First, I wrote a function to return the estimated value of the second derivative using the formula
$$f''(x)=\frac{f(x+h)-2f(x)+f(x-h)}{h^2}$$
I created a list of h values between 1e-1 and 1e-12 and calculated the estimated value of the second derivative using the function and x=0.004 for each. Then I derived the exact value of the second derivative by hand. I calculated the errors for each value and displayed them all using table formatting.

## Automatic Differentiation 

Let's use the method of automatic differentation to determine the value for both the function 

$f(x) = x+\tan(\cos^2x+\sin^2x)$ 

and its derivative at $x = 0$ (or at $t = 0$ as described in class). Do so using the following steps. 

a) Write a Python function `f(x)` to compute the function value and derivative of a list $x$ using automatic differentiation. [Hint: you will need a few additional functions as well -> look at the class notebook example]

b) Make a plot of the function $f(x)$ vs. $x$ over the range of $-\pi$ to $\pi$ that you will be able to compare with your results in part c. 

c) Use your function determine the function and dervative values at $x = 0$ ($t=0$). Remember, that your list should be entered as `[t,1]`. How do you results compare with part b? Add a markdown cell to explain your process and compare your answer with your graph from part b. 


Hint: If you program is taking a long time to run, then you are probably doing the calculation the wrong way. It should only take a second or two to finish.

```python
# part a
#define cos, sin, tangent, power, and add for automatic differentiation
def cos(x):
    vx,dx = x
    return [math.cos(vx), -math.sin(vx)*dx]

def sin(x):
    vx,dx = x
    return [math.sin(vx),  math.cos(vx)*dx]

def tan(x):
    vx,dx = x
    return [math.tan(vx), dx/(math.cos(vx)**2)]

def power(x,c):
    vx,dx = x
    return[vx**c, dx*(c*vx**(c-1))]
    
def add(x,y):
    vx,dx = x
    vy,dy = y
    return[vx+vy, dx+dy]

#combines definitions above to make f(x)
def f(x):
    u1 = cos(x)
    u2 = power(u1, 2)
    u3 = sin(x)
    u4 = power(u3, 2)
    u5 = add(u2,u4)
    u6 = tan(u5)
    val = add(x, u6)
    return val
```

```python
# part b
#create arrays to store values of x, f(x) and f'(x)
x_vals = np.linspace(-math.pi,math.pi,100)
f_vals = np.zeros(len(x_vals))
df_vals = np.zeros(len(x_vals))

#calculate f and f'(x) for all values of x
for i in range(len(x_vals)):
    f_vals[i]= f([x_vals[i],1])[0]
    df_vals[i] = f([x_vals[i],1])[1]

#make plot of f(x) and f'(x)
plt.scatter(x_vals, f_vals, c = "green", s=8, label = "f(x)")
plt.scatter(x_vals, df_vals, c = "blue", s=8, label = "f'(x)")
plt.xlabel("x")
plt.ylabel("value")
plt.title("f(x) and f'(x) vs. x")
plt.legend(fontsize = 12)
```

```python
# part c
x = 0
f_0 = f([x,1])[0]
f_prime_0 = f([x,1])[1]

print("At x=",x,"the value f(x) is",f_0,"and the value f'(x) is",f_prime_0)
```

#### Explanation:

This was a very excessive method to do a fairly simple problem. Regardless of the value of x, $\cos^2(x)+\sin^2(x) = 1$. Then $\tan(1)\approx 1.557$. Using this fact makes the process significantly easier. Now $f(x) = x + 1.557$. The derivative at all points will be 1, as shown in parts b and c. The answer from part c makes sense given the graph generated in part b and the fact that $f(x=0) \approx 0 + 1.557$.

In part a, I defined cos, sin, tan, add, and power and combined those definitions to create a function f(x) which returns a pair of values, f and f' for a certain x. Then in part b I created an array of x values, f values, and f' values, and used my function to populate the arrays. I then plotted both f and f' vs. x onb the same plot. Part c simply prints out one of the pairs (at x=0) returned by my function f(x).
