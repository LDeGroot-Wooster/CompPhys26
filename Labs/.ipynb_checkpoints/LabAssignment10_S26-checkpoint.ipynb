## Derivatives with front, back, and central difference (25 pts)

Let's evaluate the derivative of: 

$$ f(x) = \frac{1-\cos x}{x^2} $$

I don't want to do this by hand, so let's do it computationally. 

*Note: don't forget to put brief comments in your code*

#### Step 1: Define the function `f(x)` that returns the function $f(x)$ from above. 

```python
def f(x):
    return (1-np.cos(x))/ x**2
```

#### Step 2: Define functions that call on your function from step 1 to calculate the front, back, and central difference first derivatives of $f(x)$ at a specific location $x$. 

```python
# front derivative
# define the forward derivative
def forward(f, x, h):
    return (f(x + h) - f(x)) / h

# define backward derivative
def backward(f, x, h):
    return (f(x) - f(x-h)) / h

# define central derivative
def central(f, x, h):
    return (f(x + h/2) - f(x - h/2)) / h
```

#### Step 3: Evaluate the derivative using the three different approximations using the appropriate value of $h$ for each method evaluated from $x = 0.004$ to $x = \frac{\pi}{2}$ - let's do 20 calculations. 

```python
# define an array of x's between 0.004 and pi/2
x_arr = np.linspace(0.004, np.pi/2, 20)

# create three lists of zeros for the 20 points for each method
f_arr  = np.zeros(len(x_arr))
b_arr = np.zeros(len(x_arr))
c_arr  = np.zeros(len(x_arr))

# define the h values for each method
h_forward  = 1e-8
h_backward = 1e-5
h_central  = 1e-8

# use the defined functions to calculate the values of each derivative and append to array
for i in range(len(x_arr)):
    x = x_arr[i]
    f_arr[i]  = forward(f, x, h_forward)
    b_arr[i] = backward(f, x, h_backward)
    c_arr[i]  = central(f, x, h_central)

print('x           forward           backward          central ')

for i in range(len(x_arr)):
    print(x_arr[i],"   ", f_arr[i], "   ",b_arr[i], "   ",c_arr[i])
```

#### Step 4: Determine the equation for the derivative (by hand and then type in below this cell as markdown). Computationally calculate the true value of the derivative at the range of values from step 3 and compare to your calculated values. Explain your results in a markdown cell. 

The actual value of the derivative is $$\frac{x\sin \left(x\right)-2\left(-\cos \left(x\right)+1\right)}{x^3}$$

```python
# define the derivative eq
def df(x):
    return (x*np.sin(x) - 2*(-np.cos(x) + 1)) / x**3

# create a zeros array for the derivatitive
true_val  = np.zeros(len(x_arr))

# calculate all the values for this array
for i in range(len(x_arr)):
    x = x_arr[i]
    true_val[i]  = df(x)

# create and calculate an array of error 
f_error = np.zeros(len(x_arr))
b_error = np.zeros(len(x_arr))
c_error = np.zeros(len(x_arr))

f_error = np.abs(true_val - f_arr)
b_error = np.abs(true_val - b_arr)
c_error = np.abs(true_val - c_arr)

print('actual values    forward err      backward err    central err ')

for i in range(len(x_arr)):
    print(true_val[i],"   ", f_error[i], "   ",b_error[i], "   ",c_error[i])
```

## Second derivative (22 pts)

Let's take the second derivative of the function from above. 

*Note: don't forget to put brief comments in your code*

#### Step 1: Define functions that call on your function from step 1 above to calculate the second derivative of $f(x)$ at a specific location $x$ using central difference. 

```python
def second(f, x, h):
    return (f(x+h) -2*f(x)+f(x-h))/h**2

```

#### Step 2: Evalute the second derivative at $x = 0.004$ using a range of $h$ values from `1e-1` to `1e-12`. 

```python
x = 0.004

# create array of h vals
h_second = [10**(-i) for i in range(1, 12)]
# create an array for second derivative values
second_arr = np.zeros(len(h_second))

for j in range(len(h_second)):
    second_arr[j] = second(f, x, h_second[j])

print(second_arr)
```

```python
def d2f(x):
    return (x**2*np.cos(x) - 4*x*np.sin(x) - 6*np.cos(x) + 6) / x**4

true_val_2 = d2f(0.004)

print("the true value of the second derivative at x = 0.004 is ", true_val_2)
```
