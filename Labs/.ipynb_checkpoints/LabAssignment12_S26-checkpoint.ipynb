Use the relaxation method to solve for x and y in the following nonlinear equations: 

$$ 3x - \frac{1}{y} = -1 $$
$$2x^2 - y = 0$$

##### Part 1 (double click cell to add your equation)

$x = \frac{1}{3} \left( \frac{1}{y} - 1 \right)$

$y = 2x^2$

The x equation is undefined for $y=0$, so starting value should be chosen so that condition is never computed.
The y equation is defined for all real values of x.

If starting value of y is 1, next value of x is 0, which results in the y value after that being 0, leading to an undefined value of x. Try (2, 2) as the starting point.


```python
def relax_method_vbl2(x, y, f, h, prec):
    i = 0
    while i < 4:
        x_prev, y_prev = x, y
        x, y = f(y), h(x)
        if round(x_prev, prec) == round(x, prec) and round(y_prev, prec) == round(y, prec):
            i += 1
    return x, y
```

```python
def f(y):
    return (1/3)*((1/y) - 1)
def h(x):
    return 2*x**2

x, y = relax_method_vbl2(2.0, 2.0, f, h, 6)
print(x, y) # diverges, reformat equations
```

```python
def f_alt(x):
    return 1/(3*x + 1)
def h_alt(y):
    return math.sqrt(y/2)

x, y = relax_method_vbl2(2.0, 2.0, h_alt, f_alt, 6)
print(x, y)
```

## Binary Search (25 pts)

Use the binary search method to find the roots of $f(x) = 5e^{-x} +x -5$. 

To do this: 
1) Start by defining two functions: `f(x)` which returns the given function in the problem and `binary_solve(f,x1,x2,accuracy)` which applies the binary search method for a given function $f(x)$ based on two initial values $x_1, x_2$, and continues to iterate until a given accuracy is achieved. 

2) Call the function and analyze the results. [Use a graph to estimate the interval that returns a non-zero $x$.] 

3) Plot this point as a magenta star on the graph from step 2 to see if it agrees. 


*You must clearly comment your code and include a markdown cell explaining your process*

```python
# part 1
def f3(x):
    return 5*math.exp(-x) + x -5

def binary_solve(func, x1, x2, acc):
    f1 = func(x1)
    f2 = func(x2)
    if f1*f2 < 0:
        while np.abs(x2-x1) > acc:
            x_bi = 0.5*(x2+x1)
            f_bi = func(x_bi)
            if f_bi*f1 > 0:
                x1, f1 = x_bi, f_bi
            else:
                x2, f2 = x_bi, f_bi
        result = 0.5*(x1+x2)
    else:
        print('ERROR: endpoints must have opposite signs')
        result = None
    return result
```

```python
# part 2
domain = np.linspace(-2, 8, 100)
val = []
for n in domain:
    val.append(f3(n))
zero_list = np.zeros(len(domain), float)
plt.plot(domain, val)
plt.plot(domain, zero_list, color = 'red')
plt.show()
```

We can expect the function to have two roots at approximately $x=0$ and $x=5$. We can therefore determine that setting the initial endpoints at $x=-1$ and $x=1$ for the first zero and $x=4$ and $x=6$ for the second should get a precise result.

```python
x1_0 = binary_solve(f3, -1, 1, 1e-8)
x2_0 = binary_solve(f3, 4, 6, 1e-8)
zeroes = np.array([x1_0,x2_0])
domain = np.linspace(-2, 8, 100)

val = []
for n in domain:
    val.append(f3(n))
zero_list = np.zeros(len(domain), float)
plt.plot(domain, val)
plt.plot(domain, zero_list, color = 'red')
plt.scatter(zeroes, [0, 0], color = 'magenta', marker = '*')
plt.show()
```

## Energy of an electron in a square potential well

Consider a square potential well of width $w$, with walls of height $V$: 



Using Schrödinger's equation, it can be shown that the allowed energies $E$ of a single quantum particle of mass $m$ trapped in the well are solutions of 
$$ \tan{\sqrt{w^2mE/2\hbar^2}} = \begin{cases} 
\sqrt{(V-E)/E} & \text{ for the even numbered states} \\
-\sqrt{E/(V-E)} & \text{ for the odd numbered states} 
\end{cases}
$$

where the states are labeled with integers starting from 0 and working upward, the ground state being state 0, the first excited state being state 1, and so forth. 

a) For an electron of (mass $9.1094 \times 10^{-31}$ kg) in a well with $V = 20 \text{ eV}$ and $w = 1 \text{ nm}$, write a Python program to plot the three quantities 
$$y_1 = \tan{\sqrt{w^2mE/2\hbar^2}} \hspace{0.2in} y_2 = \sqrt{\frac{V-E}{E}} \hspace{0.2in} y_2 = -\sqrt{\frac{E}{V-E}}, $$

on the same graph, as a function of $E$ from $E = 0$ to $E = 20 \text{ eV}$. **Hint** You may find it useful to define functions to calculate $y_1$, $y_2$, and $y_3$. 

b) From your plot make approximate estimates of the energies of the first six energy levels of the particle. 

c) Write a program to calculate the values of the first six energy levels in electron volts to an accuracy of $0.001 \text{ eV}$ using binary search. 

*hint, remember that energies need to be converted to joules in the $y_1$ equation above*

*You must clearly comment your code and include a markdown cell explaining your process*

```python
#a)
def y1(E):
    return math.tan(math.sqrt(((1e-9)**2 * (9.1094e-31) * (1.60218e-19 * E)) / (2 * (1.0546e-34)**2)))
def y2_even(E):
    return math.sqrt((20*1.60218e-19 - (1.60218e-19 * E))/(1.60218e-19 * E))
    # Undefined when E = 0.
def y2_odd(E):
    return -1 * math.sqrt((1.60218e-19 * E)/((1.60218e-19 * 20) - (1.60218e-19 * E)))
    # Undefined when E = V. Probably need to exclude the endpoints of the (0, 20) domain.

domain = np.linspace(0, 20, 202, endpoint = False)
domain = domain[1:len(domain)-1]
y1_list, y2e_list, y2o_list = [], [], []
for n in domain:
    if abs(y1(n)) <= 50:
        y1_list.append(y1(n))
    elif y1(n) > 0:
        y1_list.append(50)
    elif y1(n) < 0:
        y1_list.append(-50)
    y2e_list.append(y2_even(n))
    y2o_list.append(y2_odd(n))
plt.plot(domain, y1_list, color = 'red', label = "y1")
plt.plot(domain, y2e_list, color = 'green', label = "y2 - even")
plt.plot(domain, y2o_list, color = 'blue', label = "y2 - odd")
plt.legend()
plt.show()
```

The jumps in the graph are due to the frequent undefined values caused by the tangent function, so I will not be counting the points where $y_1$ jumps across the two $y_2$ curves.

```python
#c)
def y_diff_even(E):
    return y1(E) - y2_even(E)
def y_diff_odd(E):
    return y1(E) - y2_odd(E)

elevel = []
elevel.append(binary_solve(y_diff_odd, 0.5, 2.5, 0.001))
elevel.append(binary_solve(y_diff_even, 2.5, 3.0, 0.001))
elevel.append(binary_solve(y_diff_odd, 4.0, 7.5, 0.001))
elevel.append(binary_solve(y_diff_even, 7.5, 9.0, 0.001))
elevel.append(binary_solve(y_diff_odd, 10.0, 12.5, 0.001))
elevel.append(binary_solve(y_diff_even, 13.5, 17.5, 0.001))

for n in elevel:
    print('Energy Level', elevel.index(n)+1, ':', n, 'eV')
```
