### Applying Trapezoidal Rule (22 pts)


Write a code to do the following: 
- Create a user-defined function `trapezoidal_int(a,b,N,f)` to determine the integral using the trapezoidal rule using your function defined above 

    - here: a = the lower limit of the integrand, b = upper limit, N = number of steps, and f = the function from above 
- Do so using $N = 100$ steps
- Do so again by doubling the number of steps and seeing how much the result accuracy changes by. 
- Estimate the error on the result using the equation discussed in class. 

When your program is complete, add a markdown cell below it to explain your process and program. Don't forget to comment in your code

```python
# Function to integrate using Trapezoidal Rule
def trapezoidal_int(a,b,N,f):
    h = (b-a)/N #number of steps
    k_sum = 0
    #for loop to determine sum within definition
    for k in range(1,N):
        val = f(a+k*h)
        k_sum += val
    I = h*(0.5*f(a)+0.5*f(b)+k_sum) #definition of integral
    return I
```

```python
# N = 100 steps 
N = 100
a = 1.1
b = 10.1
f = f
I1_trap = trapezoidal_int(a,b,N,f)
print(I1_trap)
```

```python
# double steps 
N = 200
a = 1.1
b = 10.1
f = f
I2_trap = trapezoidal_int(a,b,N,f)
print(I2_trap)
```

```python
# estimate error 
delta = np.abs((1/3)*(I2_trap-I1_trap)) #definition of error for trapezoidal
print(delta)
```

First I defined a function that takes four inputs to define the trapezoidal estimate of the area under the curve. The function first calculates h, then calculates the sum of f(a+k*h) for k being each step from N=1 to N=the maximum number of steps. f(x) is defined above as sin(x)/ln(x). It then adds that sum together with f(a)/2 and f(b)/2, and multiplies the whole thing by h. I ran this function twice, both times with a = 1.1, b=10.1, f=f=(sin(x)/ln(x)). The first time N=100 and the second time N=200. Finally I calculated the error between the two calculations with different N's using the formula delta = (I2-I1)/3.

### Applying Simpson's Rule (22 pts)

Write a code to do the following: 
- Create a user-defined function `simpsons_int(a,b,N,f)` to determine the integral using Simpson's rule using your function defined above 

    - here: a = the lower limit of the integrand, b = upper limit, N = number of steps, and f = the function from above 
- Do so using $N = 50$ steps [note this is half of the initial steps used in trapezoidal]
- Do so again by doubling the number of steps and seeing how much the result accuracy changes by. 
- Estimate the error on the result using the equation discussed in class. 

When your program is complete, add a markdown cell below it to explain your process and program. Don't forget to comment in your code. 

```python
# Function to integrate using Simpson's Rule
# Function to integrate using Trapezoidal Rule
def simpsons_int(a,b,N,f):
    h = (b-a)/N #number of steps
    k_sum = 0
    #for loop to determine sum within definition
    for k in range(1,N):
        if k%2 == 0:
            val = f(a+k*h)
            k_sum += 2*val #for even k's add 2*f(a+kh)
        elif k%2 == 1:
            val = f(a+k*h)
            k_sum += 4*val #for odd k's add 4*f(a+kh)
    I = (1/3)*h*(f(a)+f(b)+k_sum) #definition of integral
    return I
```

```python
#N = 50 steps
N = 50
a = 1.1
b = 10.1
f = f
I1_simp = simpsons_int(a,b,N,f)
print(I1_simp)
```

```python
# double steps
N = 100
a = 1.1
b = 10.1
f = f
I2_simp = simpsons_int(a,b,N,f)
print(I2_simp)
```

```python
# error
delta = np.abs((1/15)*(I2_simp-I1_simp)) #definition of error for simpson's
print(delta)
```

First I defined a function that takes four inputs to define the Simpson's estimate of the area under the curve. The function first calculates h, then calculates the sum of f(a+k*h) for k being each step from N=1 to N=the maximum number of steps. If k is even, this value is multiplied by 2 and added to the sum. If k is odd, it is instead multiplied by 4. f(x) is defined above as sin(x)/ln(x). It then adds that sum together with f(a) and f(b), and multiplies the whole thing by h/3. I ran this function twice, both times with a = 1.1, b=10.1, f=f=(sin(x)/ln(x)). The first time N=50 and the second time N=100. Finally I calculated the error between the two calculations with different N's using the formula delta = (I2-I1)/15.

    $$ I_i = \frac{1}{2}I_{i-1} + h_i\sum_{k=1 (k \text{ odd})}^{N_i-1} f(a+kh_i)$$


Trapezoidal had an error estimate of about 0.0131, which is about 4 times larger than the error estimate for Simpson's of 0.00296. Simpson's only does 100 steps to acheive this accuracy, while trapezoidal needs 200. However, if the target accuracy is 0.003, we need even more steps for trapezoidal top be accurate enough. The cell below calculates the number of steps needed to acheive the desired error. I used a while loop to check if the error was small enough. If not, I calculated I1, doubled N, and used the equation above to calculate N2. This required using a for loop and an if statement to check if each k was odd. Then I recalculated delta. This was repeated the necessary number of times until the error was small enough. I ended up doubling N three times, until it reached 800 steps, to acheive a small enough error.

```python
# Value of I calculated with 200 steps above
N = 100
a = 1.1
b = 10.1
f = f

delta = 1 #starting value of delta to be replaced when while loop below is run

#using a while loop to find when delta becomes low enough, starting at N=100
double_count = 0
while delta > 0.003:
    I1_trap = trapezoidal_int(a,b,N,f)
    N = N * 2
    h = (b-a)/N
    double_count += 1
    k_sum = 0
    for k in range(1,N):
        if k%2 == 1:
            val = f(a+k*h)
            k_sum += val #for odd k's add f(a+kh)
    I2_trap = 0.5*I1_trap+h*k_sum #equation above to save computational time. I2=0.5I1+h*k_sum
    delta = np.abs((1/3)*(I2_trap-I1_trap))
print("Starting at N = 100, I doubled the number of steps", double_count, "times to yield an error of about", round(delta,5), "at N =",\
     N, "steps.")
```

### Integrating Data (22 pts)

In GitHub, you will find a file called `forces.txt`, which contins two columns of numbers. The first column reprsents the distance $x$ in meters and the second is the force $F$ along the x-direction in Newtons on the particle. We want to determine the work done on the particle, by doing the following: 

- read in the data
- use the trapezoidal rule to calculate the approximate work done on the particle in the $x$-direction. 
- plot the original force vs. distance data 
- plot the total work as a function of distance on the same graph - what is the work done? 
    - does this look familiar? 

Remember that: 

$$ W = \int_a^b F(x) \,d x $$

When your program is complete, add a markdown cell below it to explain your process and program. Don't forget to comment in your code. 


```python
data = np.loadtxt("forces.txt", float)
dist = data[:,0]
force = data[:,1]

total_work = [0]
work = 0
for i in range(1, len(dist)):
    b=i
    a=i-1
    N=1
    h=(dist[b]-dist[a])/N
    f=force
    I = h*0.5*(f[b]+f[a]) #definition of integral
    work += I
    total_work.append(work)


plt.scatter(dist, force, c = "green", s=20, label = "Force")
plt.scatter(dist, total_work, c = "blue", s=20, label = "Total Work")
plt.legend(fontsize = 12)
plt.xlabel("Distance (meters)")
plt.ylabel("Force (N) and Work (N*m)")
```

I had to go back to the basic definition of the area of a trapezoid here. For each index of the data list (starting at 1) I calculated h (the distance at the index minus the distance at the previous index) and multiplied that by 0.5 times the sum of the force at the first index and the force at the second index. I appended this all to a list containing a zero (the total work done at the 0th point) and plotted the result.
