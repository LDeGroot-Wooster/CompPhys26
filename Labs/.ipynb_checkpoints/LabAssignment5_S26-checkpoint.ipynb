### Write a script 1 
#### Making a density plot of a function

Write a code to do the following: 
- Create a 2D array that shows how the values change linearly from the center outwards based on the equation: 
$$ B = 7.5r - 2.8 $$
    - the overall grid should be 200 x 200 units 
    - you should use a 1000x1000 array
- Plot the array as a density plot, change the color scale to one of your choice, and include a color bar

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
arr_size = 1000
grid_size = 200
x = np.linspace(-grid_size/2,grid_size/2,arr_size) #gives 1000 values of x between -100 and +100
y = np.linspace(-grid_size/2,grid_size/2,arr_size) #same for y
B = np.zeros([arr_size,arr_size],float) #zero array to be filled with B values
for i in range(arr_size):
    for j in range(arr_size):
        r = np.sqrt(x[i]**2+y[j]**2) #calculates r using r^2 = x^2+y^2
        B[i,j] = 7.5*r-2.8
plt.imshow(B, origin= "lower",extent=[-100,100,-100,100]) #plots B on 200by200 grid
plt.colorbar()
```

I assumed r was the distance from the center (0,0) like it usually is. First I defined the array size and the grid size. To get an array of 1000 values on a 200 by 200 grid, I calculated 1000 values between -100 and +100 for x and y. I created an array of zeroes to store B. I then used a nested for loop to calculate all the values of r and B and put the B values in the array. I plotted the correct region of the data and added a colorbar.
