### Write a script 2 (25 pts)
#### Find the sum of odds

Write a code to add all the odd numbers from 1 to $N$. 
- The function should return the sum *and* the time in milliseconds needed to perform this sum. (Google is your friend!)
- Use your function to report the sum of all the odd integers from 1 to 1,000,000 (inclusive) as well as how long it took to do this.

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
def timely_add(N):
    sum = 0
    for i in range(N+1): #includes all numbers between 1 and N including N
        if i%2 != 0: #selects odd numbers
            sum += i
    return sum

start_time = time.perf_counter() #starts stopwatch
result = timely_add(1000000) #does calculation
end_time = time.perf_counter() #stops stopwatch

elapsed_time = end_time - start_time #calculates elapsed time
time_ms = round(elapsed_time*1000,1) #calculates milliseconds

print("Adding all odd numbers from 1 to 1000000 to get", result, "took", time_ms, "milliseconds.")
```

I first defined a function that sums every odd number (found using modulus) between 1 and 1 above the given number N (in order to include the given number). It returns the sum. I use built in function time.perf_counter() to time how long getting the result took. The result is in seconds so I converted it to milliseconds and printed the result.

### Write a script 3 (25 pts) - extra numpy practice [if time permits]

The equation for standard deviation is given by: 
$$ \sigma = \left[ \frac{1}{N} \sum\limits_{i = 1}^{N} \left( a_i -\overline{a} \right)^2 \right]^{1/2} $$
where $\overline{a}$ is the average value of the array, and $a_i$ is each element of the array of length $N$. 
In the cell below, write a function to calculate the standard deviation for an input array, a:
- First compute the average of the elements in a to define $\overline{a}$ [hint: there is a function called np.mean()] **I have provided the array a for you - be sure to understand what it does**
- Next compute the sum over the squares of $a - \overline{a}$
- Then divide the sum by the number of elements in the array
- Finally take the square root (you can use np.sqrt())
Test your function on a random array, and compare to the built-in np.std() function

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
# first create input array a
a = np.random.random_sample(20)*20 #prints 20 random numbers from 0 to 20
print(a)
def stand_dev(a):
    avg_a = np.mean(a) #finds average of array
    sum_diff_sq = 0
    for i in a:     #for loop to sum all squared differences
        diff_sq = (i-avg_a)**2     #finds the squared difference between each element and the average
        sum_diff_sq += diff_sq     
    stdev = np.sqrt(sum_diff_sq/len(a))    #divides sum of squared differences by number of numbers in array and sqrts
    return stdev
print(stand_dev(a))
print(np.std(a))
```

To create my own standard deviation function, I first computed the average of the random array using np.mean(). Then I used a for loop that calculated each squared difference (a-a_avg) and added it together. I then divided that by the length of the array (20) and square rooted it to get the standard deviation.
