## Gaussian Quadrature 1 (22 pts)

Consider the integral from last lab $$\int_{1.1}^{10.1} \frac{\sin{x}}{\ln{x}} \,dx $$ a) Re-evaulate the integral using Gaussian quadrature with 𝑁=10 points 

b) Repeat the evaluation with 2𝑁 points and evaluate the error.  

c) Compare these results with your results for both the trapezoidal and Simpson’s rules


```python
def gauss_quad(a, b, N, f):
    x, w = gaussxw(N, a, b)
    s = 0.0
    for k in range(N):
        s += w[k] * f(x[k])
    return s
# takes the integration bounds, number of points, and the function to be integrated, calculates the points and weights for Gaussian quadrature,
# sums up products of the weights and the values of the function at the defined points and returns the sum.
def f(x):
    return math.sin(x) / math.log(x)

print(gauss_quad(1.1, 10.1, 10, f))
```

```python
int_10 = gauss_quad(1.1, 10.1, 10, f)
int_20 = gauss_quad(1.1, 10.1, 20, f)
print('N = 10:', int_10)
print('N = 20:', int_20)
gauss_err = int_20 - int_10
print('Error:', abs(gauss_err))
# uses the previously defined function to estimate the error using the formula discussed in class
```

```python
def trapezoidal_int(a, b, N, f):
    h = (b-a) / N
    s = 0.5*f(a)+0.5*f(b)
    for k in range(1, N):
        s += f(a+k*h)
    return(h*s)
def simpsons_int(a, b, N, f):
    h = (b-a) / N
    s = f(a) + f(b)
    for k in range(1, N):
        if k%2 == 1:
            s += 4*f(a+k*h)
        else:
            s += 2*f(a+k*h)
    return (1/3)*h*s
# these functions were copied from previous lab, but I decided it wasn't worth bringing over the error calculation functions, so the error is just
# calculated in the print command instead

print('Trapezoidal Rule:')
print('N = 50:', trapezoidal_int(1.1, 10.1, 50, f))
print('N = 100:', trapezoidal_int(1.1, 10.1, 100, f))
print('Error:', abs((1/3) * (trapezoidal_int(1.1, 10.1, 100, f) - trapezoidal_int(1.1, 10.1, 50, f))))

print("Simpson's Rule:")
print('N = 50:', simpsons_int(1.1, 10.1, 50, f))
print('N = 100:', simpsons_int(1.1, 10.1, 100, f))
print('Error:', abs((1/15) * (trapezoidal_int(1.1, 10.1, 100, f) - trapezoidal_int(1.1, 10.1, 50, f))))
```

## Gaussian Quadrature 2 (22 pts)

One of the caveats of Gaussian Quadrature is that the function being integrated must be relatively smooth. With this in mind, let’s integrate the function $$ f(x) = e^{-x^2/2}$$ from 𝑥=0 to 𝑥=1.5 using 2 sample points. 


a) First, make a plot of $f(x)$ vs. $x$ for 50 points within the limits of integration. Is this function smooth? Do we expect the method of GQ to return an accurate result? 

b) Determine and print out the values of the sample points and weights for this problem. 
Evaluate the integral. 

c) Double the number of sample points and apply parts (b) and (c) until your error is less than our desired accuracy of $10^{−5}$


```python
def g(x):
    return math.e**(-(x**2) / 2)
x = np.linspace(0, 1.5, 50)
gx = []
for n in x:
    gx.append(g(n))
plt.plot(x, gx)
plt.show()
# defines the function g(x), and calculates it across a linspace array so it can be plotted.
```

Function is smooth, Gaussian quadrature will return an accurate result.

```python
x, w = gaussxw(2, 0, 1.5)
print(x, w)
print(gauss_quad(0.0, 1.5, 2, g))
```

```python
N = 2
gauss_val = [gauss_quad(0.0, 1.5, N, g)]
print('N = ', N, ':', gauss_val[-1])
gauss_err = 100000 # this starting error isn't significant, it just needs to be high enough for the program to not skip the loop
while gauss_err > 1e-5:
    N = 2*N
    gauss_val.append(gauss_quad(0.0, 1.5, N, g))
    gauss_err = abs(gauss_val[-1] - gauss_val[-2])
    print('N = ', N, ':', gauss_val[-1])
    print('Error:', gauss_err)
#increments N by two, performs Gaussian quadrature for that many points, calculates the error, prints both, and repeats unless desired precision reached
```

## Infinite integrals (22 pts)

Let's evaluate the integral: 

$$ \int_0^\infty \frac{x^2}{\sqrt{x}}e^{-x}dx $$

I don't want to do this by hand, so let's do it computationally. 

*Note: don't forget to put brief comments in your code*

#### Step 1: Define the variable substitution you will make for to solve this problem. This should be done in markdown using LaTex formatting. 

$z=\frac{x}{x+1}$, $\mathrm{d}x = \frac{\mathrm{d}z}{(1-z)^2}$

#### Step 2: Define the function to use in the next step for integration

```python
def sub(z):
    return -z / (z-1)
def j(z):
    return (sub(z)**2 * (math.e ** (-1*sub(z)))) / (math.sqrt(sub(z)) * (1-z)**2)
# just defines the function with the appropriate substitution made, but I used a second function for the substitution to make it neater
```

#### Step 3: Evaluate the integral using Gaussian Quadrature with 20 sample points

```python
print('Gaussian Quadrature:', gauss_quad(0, 1, 20, j))
print('Known Value:', (3*math.sqrt(math.pi))/4)
# I put the known value here for reference, so I could see it in decimal form
```

#### Step 4: Include a markdown explaining the process and result. Include how your answer compares to the answer of $\frac{3\sqrt{\pi}}{4}$

The substitution for x $z=\frac{x}{x+1}$ is equal to 1 when $x$ is infinite, so making this substitution turns this indefinite integral into a definite one. After rearranging the substitution to isolate $x$ and taking the derivative to find the substitution for $\mathrm{d}x$, the function to be integrated, $j(z)$, can be defined in the code. The program then uses the previously defined function for performing Gaussian quadrature to evaluate the integral of $j(z)$. The value this evaluation of the integral gives is $8.027*10^{-4} \%$ greater than the known value of $\frac{3\sqrt{\pi}}{4}$.

## Multiple integrals (22 pts)

Let's find the mass of a metal plate that is 3 meters wide (in the $x$ direction) and 2 meters tall ($y$ direction), with a nonuniform density of (in kg/$\text{m}^2$)

$$ \sigma(x,y) = (\sin(\pi x)+1)y $$

Hint 1: the total mass is given by: $$M = \iint_R \sigma(x,y) \hspace{1mm} dA$$

*Note: don't forget to put brief comments in your code*

#### Step 1: Write the full equation for the double integral in markdown. (Hint: you'll need to swithch the cell to markdown and then you can use some of my latex coding from the cell above this)

$$M = \int_0^2 \int_0^3 y(\sin(\pi x)+1) \hspace{1mm} dx dy$$

#### Step 2: Define the function to use in the next step for integration

```python
def sigma(x, y):
    return y*(math.sin(math.pi*x)+1)
```

#### Step 3: Evaluate the double integral using double Gaussian quadrature for 100 sample points along each axis. 

```python
N = 100
ax, bx, ay, by = (0, 3, 0, 2)
xx, wx = gaussxw(N, ax, bx)
yy, wy = gaussxw(N, ay, by)
# bounds of integration defined, and weights and points for Gaussian quadrature are computed. Same as normal, just another dimension to account for.
s = 0.0
for i in range(N):
    for j in range(N):
        s += wx[i]*wy[j]*sigma(xx[i], yy[j])
# I can't use the quadrature function I defined earlier, since that's only meant for single integrals, so I had to write this out in full;
# it's the exact same thing, just iterating over two different variables, and more parameters for the equation
print(s)
```

#### Step 4: Include a markdown explaining the process and result. Include how your answer compares to the answer of $\frac{4}{\pi}+6 \simeq 7.2732395$

As per Fubini's Theorem, the double integral over a rectangular domain $R$ is given by $$M = \int_{a_y}^{b_y} \int_{a_x}^{b_x} f(x, y) \hspace{1mm} dx dy,$$ where the bounds of integration are the bounds of the domain in each respective variable. The program defines these integration bounds and the number of points used, then uses them to calculate the points and weights required for Gaussian quadrature. It then iterates across each variable and sums up the product of the weights and the value of the function $\sigma(x, y)$ at each point. This evaluation of the integral has a percent error of $4.885*10^{-16} \%.$
