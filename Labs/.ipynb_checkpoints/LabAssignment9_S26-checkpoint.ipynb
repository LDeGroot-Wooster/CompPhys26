## Gaussian Quadrature 1 (22 pts)

Consider the integral from last lab $$\int_{1.1}^{10.1} \frac{\sin{x}}{\ln{x}} \,dx $$ a) Re-evaulate the integral using Gaussian quadrature with 𝑁=10 points 

b) Repeat the evaluation with 2𝑁 points and evaluate the error.  

c) Compare these results with your results for both the trapezoidal and Simpson’s rules


```python
# part a
def f(x):
    return math.sin(x)/math.log(x)

N = 10
a = 1.1
b = 10.1
x,w = gaussxw(N,a,b) #calculate sample points and weights

s = 0.0
#perform sum over all sample points
for k in range(N):
    s += w[k]*f(x[k])
print(s)
```

```python
# part b
x2,w2 = gaussxw(2*N,a,b) #calculate sample points and weights

s2 = 0.0
#perform sum over all sample points
for k in range(2*N):
    s2 += w2[k]*f(x2[k])
print(s2)

delta_gauss = np.abs(s2 - s) #error = I(2N)-I(N)
print("The error is", delta_gauss)
```

```python
# part c
#function to calculate integral using trapezoidal method
def trapezoidal_int(a,b,N,f):
    h = (b-a)/N #number of steps
    k_sum = 0
    #for loop to determine sum within definition
    for k in range(1,N):
        val = f(a+k*h)
        k_sum += val
    I = h*(0.5*f(a)+0.5*f(b)+k_sum) #definition of integral
    return I

def trap_error(I1, I2):
    delta_trap = np.abs((1/3)*(I2-I1)) #definition of error for trapezoidal
    return delta_trap

# Function to integrate using Simpson's Rule
def simpsons_int(a,b,N,f):
    h = (b-a)/N #number of steps
    k_sum = 0
    #for loop to determine sum within definition
    for k in range(1,N):
        if k%2 == 0:
            val = f(a+k*h)
            k_sum += 2*val #for even k's add 2*f(a+kh)
        elif k%2 == 1:
            val = f(a+k*h)
            k_sum += 4*val #for odd k's add 4*f(a+kh)
    I = (1/3)*h*(f(a)+f(b)+k_sum) #definition of integral
    return I

def simp_error(I1, I2):
    delta_simp = np.abs((1/15)*(I2-I1)) #definition of error for simpson's
    return delta_simp

#calculate results and errors for trapezoidal and simpsons
trap_res = trapezoidal_int(a,b,N,f)
simp_res = simpsons_int(a,b,N,f)
trap_err = trap_error(trapezoidal_int(a,b,N,f), trapezoidal_int(a,b,2*N,f))
simp_err = simp_error(simpsons_int(a,b,N,f), simpsons_int(a,b,2*N,f))

print("Performing 20 steps of Gaussian Quadrature yields a result of", s2, "with an error of", delta_gauss, ". \
Performing 20 steps of Trapezoidal Rule yields a result of", trap_res, "with an error of", trap_err,". \
Performing 20 steps of Simpson's Rule yields a result of", simp_res, "with an error of", simp_err, ".")
```

First, I calculated the value of the integral using gaussian quadrature for 10 and 20 steps. I used this by using gaussxw to calculate the correct sample points for each number of steps and their weights. I calculated the error in gaussian quadrature by simply subtracting these values. I then added some of my code from last lab to calculate the same integrals using trapezoidal and simpsons method and their errors. Gaussian Quadrature was significantly more accurate than either method with a much smaller error.

## Gaussian Quadrature 2 (22 pts)

One of the caveats of Gaussian Quadrature is that the function being integrated must be relatively smooth. With this in mind, let’s integrate the function $$ f(x) = e^{-x^2/2}$$ from 𝑥=0 to 𝑥=1.5 using 2 sample points. 


a) First, make a plot of $f(x)$ vs. $x$ for 50 points within the limits of integration. Is this function smooth? Do we expect the method of GQ to return an accurate result? 

b) Determine and print out the values of the sample points and weights for this problem. 
Evaluate the integral. 

c) Double the number of sample points and apply parts (b) and (c) until your error is less than our desired accuracy of $10^{−5}$


```python
#part a

#define function to plot
def f(x):
    return math.exp(-x**2/2)

a = 0
b = 1.5
N = 2

all_xs = np.linspace(a,b,50) #50 points between a and b
all_pts = np.zeros([50,2], float) #empty array to store values of x and f(x)

#populate data array
for x in range(50):
    all_pts[x,0] = all_xs[x]
    all_pts[x,1] = f(all_xs[x])

#plot function
plt.scatter(all_pts[:,0], all_pts[:,1], c = "green", s=20, label = "Force")
plt.xlabel("x")
plt.ylabel("f(x)")
```

This looks fairly smooth to me. It also looks like if I do like 5 dots and connect them, it will give a pretty good idea of the shape of the plot. I expect GQ to work fairly well. However, with a value as low as two, the flat part at the beginning of the function will be kind of ignored I think. 2 points might be less accurate than usually expected from GQ.

```python
x,w = gaussxw(N,a,b) #calculate sample points and weights
print("The sample points are", x, "and their weights are", w)

s = 0.0
#perform sum over all sample points
for k in range(N):
    s += w[k]*f(x[k])
print("The result with", N, "sample points is", s)
```

```python
x2,w2 = gaussxw(2*N,a,b) #calculate sample points and weights

s2 = 0.0
#perform sum over all sample points
for k in range(2*N):
    s2 += w2[k]*f(x2[k])
print(s2)

delta_gauss = np.abs(s2 - s) #error = I(2N)-I(N)
print("The result with", N*2, "sample points is", s2, "and the error is", delta_gauss, ".")

#now a while loop is used to calculate the number of steps needed to get an error less than the desired error
desired_err = 1e-5
num_doubles = 1  #stores how many times the while loop has run
current_N = N*2  #stores the number of steps used in the most recent while loop
current_s = 0 #stores the result of the integral from the most recent while loop
current_delta = delta_gauss #stores the error from the most recent while loop
old_s = s2  #stores the result of the integral from the previous while loop (for calculating error)

while current_delta >= desired_err: #checks whether error is small enough
    current_N *= 2 #doubles number of steps
    x, w = gaussxw(current_N,a,b)
    current_s = 0.0
    for k in range(current_N):
        current_s += w[k]*f(x[k])
    current_delta = np.abs(current_s-old_s) #calculates new error from old value of integral and value just calculated
    old_s = current_s #replaces old value of integral with new one
    num_doubles += 1

print("After", num_doubles, "doublings, the result with", current_N, "sample points is", new_s, \
      "and the error is", current_delta, ".")
```

It looks like I was correct, in this case 2 sample points was not enough to be very accurate. In this problem, I first defined a, b, f, and N. I then calculated 50 values of x between a and b and plotted the respective values of f(x) for each x. Then I performed a GQ estimation with N = 2. I repeated the calculation with N = 4. This single doubling of N was not enough to acheive the desired error, so I wrote a while loop to recalculate the value of the integral and delta until a small enough error was reached. It turns out that N = 8 was enough points. To show the versatility of the while loop, I also set desired_err = 1e-8 and got

"After 3 doublings, the result with 16 sample points is 1.085853317666042 and the error is 2.55351295663786e-14 ."

The error here is reaching the limits of computer rounding accuracy, so N=16 is extremely accurate.

## Infinite integrals (22 pts)

Let's evaluate the integral: 

$$ \int_0^\infty \frac{x^2}{\sqrt{x}}e^{-x}dx $$

I don't want to do this by hand, so let's do it computationally. 

*Note: don't forget to put brief comments in your code*

#### Step 1: Define the variable substitution you will make for to solve this problem. This should be done in markdown using LaTex formatting. 

I will be using the substitutions
$$x=\tan{z}$$
and
$$dx = \frac{dz}{(\cos{z})^2}$$

This means that $a=0$ and $b=\pi/2$.

#### Step 2: Define the function to use in the next step for integration

```python
def f(z):
    return (math.tan(z))**2/(np.sqrt(math.tan(z)))*math.exp(-math.tan(z))/(math.cos(z)**2)
```

#### Step 3: Evaluate the integral using Gaussian Quadrature with 20 sample points

```python
#define boundaries of integral
N = 20
a = 0.0
b = math.pi/2
x,w = gaussxw(N,a,b) #calculate sample points and their weights

#sum each sample point multiplied by its weight
s = 0.0
for k in range(N):
    s += w[k]*f(x[k])

actual = 3*np.sqrt(math.pi)/4

print(s)
print(actual)

err = np.abs((s-actual)/actual) #calculate error in result

print(err)
```

#### Step 4: Include a markdown explaining the process and result. Include how your answer compares to the answer of $\frac{3\sqrt{\pi}}{4}$

First I defined f(z) by replacing every x in f(x) with $x=\tan{z}$. Then I computed a GQ estimate with N=20. A and b also had to be changed based on the x-z conversion. I got a result of 1.3293800316113007. The actual result is $3\sqrt{\pi}/4$ which is roughly equal to 1.329340388179137. This means that my calculated result is only off by 2.9821881977100983e-05.

## Multiple integrals (22 pts)

Let's find the mass of a metal plate that is 3 meters wide (in the $x$ direction) and 2 meters tall ($y$ direction), with a nonuniform density of (in kg/$\text{m}^2$)

$$ \sigma(x,y) = (\sin(\pi x)+1)y $$

Hint 1: the total mass is given by: $$M = \iint_R \sigma(x,y) \hspace{1mm} dA$$

*Note: don't forget to put brief comments in your code*

#### Step 1: Write the full equation for the double integral in markdown. (Hint: you'll need to swithch the cell to markdown and then you can use some of my latex coding from the cell above this)

Substituting the above equation 1 into equation 2 yields
$$M = \int_A (\sin(\pi x)+1)y  \hspace{1mm} dA$$
To integrate over a rectangular area we simply have to integrate over both the x and y directions. This means that the above equation is equivalent to
$$M = \int_x \int_y (\sin(\pi x)+1)y \hspace{1mm} dy  \hspace{1mm} dx$$

#### Step 2: Define the function to use in the next step for integration

```python
def f(x,y):
    return y*(math.sin(math.pi*x)+1)
```

#### Step 3: Evaluate the double integral using double Gaussian quadrature for 100 sample points along each axis. 

```python
#I am defining the origin (the bottom left corner of the plate) as (0,0) for convenience
N = 100
x_a = 0
x_b = 3
y_a = 0
y_b = 2

#calculates sample points and weights on both axes
x,w_x=gaussxw(N,x_a,x_b)
y,w_y =gaussxw(N,y_a,y_b)

#iterates through x and y sample points and calculates f(x,y) at each value to calculate sum
s = 0.0
for i in range(N):
    for j in range(N):
        s += w_x[i]*w_y[j]*f(x[i],y[j])

actual = (4/math.pi)+6

print(s)
print(actual)

err = np.abs((s-actual)/actual) #calculate error in result

print(err)
```

#### Step 4: Include a markdown explaining the process and result. Include how your answer compares to the answer of $\frac{4}{\pi}+6 \simeq 7.2732395$

First I define the function. N = 100. I set x_a and y_a equal to 0 and x_b and y_b equal to the wifth and height of the metal plate, respectively. Then I calculated the 100 sample points and their weights on each axis. For each pair of sample points, I calculated the product of their weights times the value of f(x,y) for those points. I summed all of these values together in a for loop. The result I got has an error of only 4.884637247197754e-16 from the correct answer. This error is approximately the rounding error of the computer, so N=100 gives as accurate of a result as possible.
