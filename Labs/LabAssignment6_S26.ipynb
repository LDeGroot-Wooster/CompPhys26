### Animation #1 (22 pts)
#### A Growing Cone 

- Create a cone of some initial height and radius
- Make it grow to 5 times itâ€™s size  
- Use the appropriate rate to make this occur over 10 seconds
- Make the cone a color of your choice 

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
# define canvas and initial cone
scene = canvas(height=500, width=700, autoscale=False)
my_cone = cone(pos = vector(0,0,0), axis=vector(0,1,0), radius=2.0)

# define step and rate things
scale_factor = 5.0
time = 10
fps = 100
nsteps = time * fps


axis_change = (my_cone.axis * (scale_factor )) / nsteps
radius_change = (my_cone.radius * (scale_factor )) / nsteps

for i in range(nsteps):
    rate(fps)
    my_cone.axis += axis_change
    my_cone.radius += radius_change

# my_cone.axis = vector(0,5,0)
# my_cone.radius = 10.0
```

I defined my cone and my scale factor and everything that I need to adjust the cone. Then, in a for loop, I used the .axis and .radius to change the axis and radius of the cone, while the rate() command animates the process.

### Animation #2 (22 pts)
#### Animating 3D objects - a bouncing ball

- Create a green sphere starting in the upper left corner of your scene 
- Make it zig-zag it across the scene from the upper left to the lower right, overall
- Have it cross the screen at least 5 times
- Have autoscale off

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
scene3 = canvas(height=500, width=500, autoscale=False)

# Scene boundaries
left = -10
right = 10
top = 10
bottom = -10

# create sphere at initial pos
ball = sphere(pos=vector(left, top, 0), radius=0.8, color=color.green)

# define animation things
crossings = 5         
time = 10
fps = 100
nsteps = time * fps


# downward drift per step
down_change = -(20 / nsteps)

# horizontal change per step
#side_change = 20 / (steps / crossings)
side_change = 20 / (nsteps / crossings)

# define a variable for the direction of the ball
direction = 1

for i in range(nsteps):
    rate(fps)

    # down motion
    ball.pos.y += down_change

    # side to side
    ball.pos.x += side_change * direction

    # get the ball to change direction at a boundary
    if ball.pos.x >= right:
        ball.pos.x = right
        direction = -1
    elif ball.pos.x <= left:
        ball.pos.x = left
        direction = 1
```

I looked at the two different motions that were taking place here, downward and side to side motion. THe downward motion was easy after I had the ball and scene limits all defined, it was just a constant decrease in position. For the side to side I used a similar process but was multiplying the horizontal change by a variable that changed from 1 to -1 at the right boundary, and -1 to 1 at the left boundary. This controlled the horizontal motoin direction, and allowed for the bouncy behavior.
