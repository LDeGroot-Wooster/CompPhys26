- Make it grow to 5 times itâ€™s size  
- Use the appropriate rate to make this occur over 10 seconds
- Make the cone a color of your choice

```python
import time

scenea1 = canvas(width=400,length=400,autoscale=False)

carrot = cone(axis=vector(0,-1,0),radius=0.25,color=color.orange)

timestart = time.time()
scale = 1
while scale<5:
    rate(4)
    scale += 0.1
    carrot.axis = vector(0,-1*scale,0)
    carrot.radius = 0.25*scale
    #count += 1
timefinish = time.time()
elapsed = timefinish-timestart
elapsed / 1000 #ms to s
print('The process took',elapsed,'seconds.')


```

To make sure the cone grew to 5x its size, I used a while loop looking for the scale factor to be less than or equal to 5. I increased the scale by 0.1 each time, so the scale goes from 1 to 5 over 40 iterations. For 40 iterations to take 10 seconds is a rate of 4. I used the time package to time the process.

#### Animating 3D objects - a bouncing ball

- Create a green sphere starting in the upper left corner of your scene 
- Make it zig-zag it across the scene from the upper left to the lower right, overall
- Have it cross the screen at least 5 times
- Have autoscale off

When your program is complete, add a markdown cell below it to explain your process and program. 

```python

scenea2 = canvas(width=400,length=400,range=10,autoscale=False)

#starting coordinates
x = -10
y = 10

ball = sphere(pos=vector(x,y,0),radius=1,color=color.green)


bounces = 0
while True:
    while x < 10:
        rate(10)
        x += 1
        y -= 0.2
        ball.pos=vector(x,y,0)
    bounces += 1
    if bounces == 5:
        break
    while x > -10:
        rate(10)
        x -= 1
        y -= 0.2
        ball.pos=vector(x,y,0)
    bounces += 1
    if bounces == 5:
        break

```

I used a while loop that always runs, and to end it I used a break command that executes when the ball has crossed the screen 5 times. The reason I did this is because I made the ball zigzag with more while loops where the y position steadily decreases and the x position alternates between bounces increasing or decreasing.
