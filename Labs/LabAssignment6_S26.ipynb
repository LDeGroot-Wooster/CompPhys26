```python
scene4 = canvas(width=200, height=200, background=color.white, autoscale=False)
scalar = 1
c = cone(pos=vector(0,-1,0), axis=vector(0,0.2*scalar,0), radius =0.1*scalar, color=color.cyan)

for i in range(1,11,1):
    rate(1)
    scalar = scalar+0.5*i
    c.axis = vector(0,0.2*scalar,0)
    c.radius = 0.1*scalar
```
