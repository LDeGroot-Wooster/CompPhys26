### Animation #1 (22 pts)
#### A Growing Cone 

- Create a cone of some initial height and radius
- Make it grow to 5 times itâ€™s size  
- Use the appropriate rate to make this occur over 10 seconds
- Make the cone a color of your choice 

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
scene3 = canvas(width = 400, height = 400, background = color.white, range=10,autoscale=False)
hgt = 2
rad = 1
c = cone(pos=vector(0,0,0),axis=vector(hgt,0,0),radius = rad, color=color.blue)

#determin rate for radius
total_grow_rad = 4*rad
step_size_rad = 0.1*rad
total_time = 10 #seconds
nsteps_rad = int(total_grow_rad/step_size_rad)
time_between_rad = int(nsteps_rad/total_time)

#determin rate for height
total_grow_hgt = 4*hgt
step_size_hgt = 0.1*hgt
total_time = 10 #seconds
nsteps_hgt = int(total_grow_hgt/step_size_hgt)
time_between_hgt = int(nsteps_hgt/total_time)

#grow cone over 10 seconds
for i in range(nsteps_rad):
    rate(time_between_rad)
    rad+=step_size_rad
    c.radius = rad
    hgt+=step_size_hgt
    c.axis = vector(hgt,0,0)
    
```

I created a scene with a cone with a height and radius set at 1. I made sure autoscale was off so we can see the cone growth. I calculated the correct rate. I want an increase of 4 in radius and height (x+4x=5x), over 10 seconds. I chose to change the size by increments of 0.1 times the value, so height changes at twice the rate of radius, since height is two times greater than radius.

I made a for loop to increase the radius and height at the same time by the correct values at the calculated rate.

### Animation #2 (22 pts)
#### Animating 3D objects - a bouncing ball

- Create a green sphere starting in the upper left corner of your scene 
- Make it zig-zag it across the scene from the upper left to the lower right, overall
- Have it cross the screen at least 5 times
- Have autoscale off

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
#create background
r = 10
rad = 1
scene4 = canvas(width = 400, height = 400, background = color.white, range=r+rad,autoscale=False)

#sphere in upper left
rad = 1
xpos = -r
ypos = r
s=sphere(pos=vector(xpos,ypos,0),radius=rad,color=color.green)

#calculate rate
total_dist = 2*r
step_size_x = 0.1
total_time = 10 #seconds
nsteps = int(total_dist/step_size_x)
time_between = int(nsteps/total_time)

step_size_y = 5*step_size_x #different than x so ball moves faster in y and bounces

#moves ball
for i in np.linspace(-r,r,nsteps):
    rate(time_between)
    xpos += step_size_x
    ypos -= step_size_y
    s.pos=vector(xpos,ypos,0)
    if (ypos <= -r or ypos >= r): #turns ball around if it hits the top or bottom
        step_size_y = -step_size_y
```

First I created a background with a sphere in the upper left by setting the position of the sphere to (-r,r) where r+radius of sphere is the range of the canvas. I then calculated the rate for the sphere to move. I wanted it to cover the entire canvas (a distance of 2r) in the x direction in 10 seconds. In order for it to bounce 5 times, I made the step size in y 5 times greater than the step size in x so it would reach the bottom of the screen after traversing 1/5 of the screen in the x direction.

I used a for loop to move the ball in x and y. Finally, I had to turn the ball around when it hit either the top or bottom of the canvas, so I used an if statement to change the sign of the step size when that happened.

### Animation #3 (22 pts)
#### Animating 3D objects - a spinning sign 

- Create a new scene of your choice 
- Create a rectangular sign (large, flat box)

- Have it continue to rotate CCW once every second for 5 seconds 

Note: The axis attribute may be helpful here! 


```python
scene5 = canvas(width = 400, height = 400, background = color.white, range=8,autoscale=False)

l = 6
h = 4
sign = box(pos=vector(0,0,0),length=l,width=0.1, height=h, color=color.red,axis=vector(l,h,0))

num_steps=400
for theta in np.linspace(0, 10*np.pi, num_steps): #rotate 5 times
    rate(num_steps/5) #correct rate to rotate once per second
    x_axis = l*np.cos(theta) #rotate counter clock wise
    y_axis = l*np.sin(theta)
    sign.axis= vector(x_axis,y_axis,0)
    
    
```

I created a scene with a red sign in the middle. In order to easily rotate the sign, I set the vector of the axis to (length, height, 0). Then I used a for loop going from 0 to 10pi over 400 steps (smoothly rotating 5 times). The x axis and y axis were set as the length times cos(theta) and the length times sin(theta) respectively for counter clockwise motion.
