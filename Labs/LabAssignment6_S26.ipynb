### Animation #1 (22 pts)
#### A Growing Cone 

- Create a cone of some initial height and radius
- Make it grow to 5 times itâ€™s size  
- Use the appropriate rate to make this occur over 10 seconds
- Make the cone a color of your choice 

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
scene3 = canvas(width=600, height=600, autoscale=False, range=12)

# make a cone
rad = 1
height = 2
max_height = height*5
c = cone(pos=vec(0, -1, 0), axis=vec(0, height, 0), radius=rad, color=color.cyan)

# calculate the rate
total_time = 10 #seconds
growth_rate = 0.01 

nsteps = int((max_height - height) / growth_rate)

fps = int(nsteps/total_time)

# make the cone grow
for i in range(nsteps):
    rate(fps)
    rad += growth_rate
    height += growth_rate
    c.axis = vector(0, height, 0)
    c.radius = rad
```

I started off with the one easy part by making a cone and a canvas. I decided to leave finding the rate for later and set some basic constants to move forward with. I was able to make a pretty simple for loop which increased the height and radius of the cone.\
Then I tried the rate again. I eventually found the number of steps taken for the cone to grow (which was inputted as the range for the for loop) by dividing the change in height by the growth rate. I then calulated the frames per second (fps) to put into the rate function by dividing the number of steps by the total time.
