### Animation #3 (22 pts)
#### Animating 3D objects - a spinning sign 

- Create a new scene of your choice 
- Create a rectangular sign (large, flat box)
- Make it rotate counter-clockwise smoothly 
- Have it continue to rotate CCW once every second for 5 seconds 

Note: The axis attribute may be helpful here! 


```python
scene5 = canvas(width = 500, height = 500, background = color.white, range=10)
scene5.autoscale = False
#Determining Rate needed
total_time=5 #sec
nsteps = ((10*np.pi)/0.1)
time_between = int(nsteps/total_time)
#Define variables to rotate the clock
x_ax = 0
y_ax = 0
#Define the sign
sign = box(pos=vector(0, 0, 0), axis=vector(x_ax,y_ax,0), length=12, width=.1, height=6, color=color.blue)
#Rotates it counterclockwise
for theta in np.arange(0, 10*np.pi, 0.1):
    rate(time_between)
    x_ax = 12*np.cos(-theta)
    y_ax = 12*np.sin(theta)
    sign.axis = vector(x_ax, y_ax, 0)
    
```

I started out my code by making a wonderful canvas. I also turned off autoscale just in case I needed to. Then I determined the total time for rate(), using the number of steps (5 rotations, so 10pi divided by 0.1, which is used in my for loop) divided by total time (5s). The, I defined a y_ax and x_ax variable to rotate my sign. Then, I defined my sign, making it blue, and look sign-like using the box function. Then, I went through all values of theta from 0 to 10pi, going up by 0.1 per step. Every step, I updated the axis using rcos(theta) and rsin(theta). To make it go counterclockwise, I made the theta in cos negative. Then, I updated the axes of the box.
