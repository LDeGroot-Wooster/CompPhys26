#### Calculate kinetic and potential energy without air resistance 

Use numpy arrays to complete the following:
- Create a 1D array of heights ranging from 0 to 50 m counting by 0.5 m
- Define the mass of the object 
- If the object is at rest at 50 m, calculate two arrays of the kinetic and potential energy of the object at each height. Ignore air resistance. 
- print the height array, kinetic energy array, and potential energy array 
    - do the arrays look like you would expect?

```python
import numpy as np
y_arr = np.arange(0, 50.5, 0.5)
u_arr = []
k_arr = []
m = 1
i = 0
e_tot = m * 9.81 * 50 

while i < 101:
    u_arr.append(float(round(m * 9.81 * y_arr[i], 3)))
    k_arr.append(float(round(e_tot - u_arr[i], 3)))
    i += 1

print('Height:', y_arr)
print('Kinetic Energy:', k_arr)
print('Potential Energy:', u_arr)
```

Arrays look as expected, with kinetic energy being greatest at 0 m and potential energy being greatest at 50 m. Rounded the calculated values to three decimal places to make printing them out more compact.

Script sets up initial array and variables required, along with an index variable and two empty arrays to fill later. Mass is assumed to be 1 kilogram here for simplicity. The potential and kinetic energy is calculated for each height and appended to the corresponding empty array defined earlier. Calculated values of kinetic and potential energy are rounded to three decimal places, since otherwise those arrays would take up much more space. The values appended to those arrays must be converted into floats, since otherwise they will default to np.float64() values, which makes the arrays more difficult to read when printed.
