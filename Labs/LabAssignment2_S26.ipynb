```python
meas_list = []
for i in range(10):
    val = float(input(f"Enter measurement {i+1} (between 1-10): "))
    meas_list.append(val)
print(len(meas_list))
print(sum(meas_list) / len(meas_list))
print(max(meas_list))
print(min(meas_list))
velocity = 55
distances = [t * velocity for t in meas_list]
print(distances)
```
