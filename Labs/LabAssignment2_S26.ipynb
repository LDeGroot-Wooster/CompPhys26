Use numpy arrays to complete the following:

- If the object is at rest at 50 m, calculate two arrays of the kinetic and potential energy of the object at each height. Ignore air resistance. 
- print the height array, kinetic energy array, and potential energy array 
    - do the arrays look like you would expect?

```python
height_array = np.arange(0,50.5,0.5)
print(height_array)
mass = float(input("Mass of the object: "))
gravity = float(9.81)
Energy_max = mass*gravity*height_array[len(height_array)-1]
print(Energy_max)

#Defining Potential Energy Array using U = m*g*h
potential_array = mass*gravity*height_array
print(potential_array)
#Defining Kinetic Energy Array using T = E_max - U
kinetic_array = Energy_max - potential_array
print(kinetic_array)
```

Firstly I define the height array using np.arrange as we have a bounded array with set intervals, before printing the height array. I then ask the user for an input mass, before defining gravity. I then define a maximum energy given by m.g.h_max which I then print. I then define potential and kinetic energy array before printing them.
