```python
#Defining a list
meas_list = []
#Asking the user for a value 10 times and adding them to the list
n=0
while n<10:
    meas_list.append(int(input("Give me a number between 1 and 10")))
    n=n+1
#Pringint Results
print(meas_list)
#Calculating Avg, Min, and Max
meas_list_avg = sum(meas_list)/len(meas_list)
meas_list_max = min(meas_list)
meas_list_min = max(meas_list)
#More Printing!
print(meas_list_avg)
print(meas_list_max)
print(meas_list_min)
```

I began by defining an empty list named meas_list. I the made use of a while loop to ask the user for a number between 1-10, ten times. I made sure this value was an integer, and appended these values to the list as they were added. Then, I printed the list to check my results. Once completed, I used the sum and len commands to calculate the avg, (which is found using avg=sum/len), and then found the min and max using the min and max commands. I then printed my results.

```python
#Array of Heights from 0 to 50m counting by 0.5m
heights = np.arange(0,50.1,0.5,float)
#Asking The user for a mass
mass = float(input("What is the mass of your object in kg"))
#Calculating array of potential at each height
potential = heights*10.*mass
#Calculating array of kinetic at each height
kinetic = 0.5*mass*(2.*10*(50-heights))
#Printing!
print(heights)
print(potential)
print(kinetic)
```

I began by defining an array of heights from 0m to 50m, going up by 0.5m at every value (being to have the second value be 50.1 as to include 50, and also to define it as a float). Then, I used an input to ask the user for a mass in kg. Next, I calculated the potential energy using the formula $p_e=mgh$ where $p_e$ is potential energy, $m$ is the mass (given by user) $g$ is gravity ($10 m/s^2$) and $h$ is the height of the object at a specific height (given by our array of heights). By using this equation, it creates an array of all of the potential energies of an object at the given heights wihtin the array. I then calculated the kinetic energy using the equation $k_e = \frac{1}{2}mv^2$ where $k_e$ is the kinetic energy, $m$ is the mass (given by the user), and $v$ is velocity at a specifc point. The velocity at a certain point can be found usig the equation $v^2=V_0^2+2ax$. $V_0$ is initial velocity, which is 0 $m/s^2$, a is acceleration from gravity ($10 m/s^2$), and x is the distance the object has traveled (which in this case is 50m minus the array of heights.)
