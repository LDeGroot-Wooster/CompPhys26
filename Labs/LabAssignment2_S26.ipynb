#### Calculate kinetic and potential energy without air resistance 

Use numpy arrays to complete the following:
- Create a 1D array of heights ranging from 0 to 50 m counting by 0.5 m
- Define the mass of the object 
- If the object is at rest at 50 m, calculate two arrays of the kinetic and potential energy of the object at each height. Ignore air resistance. 
- print the height array, kinetic energy array, and potential energy array 
    - do the arrays look like you would expect?

```python
#Define constants
m = 10.0 # mass of object in kg
g = 9.81 #accel due to gravity on Earth (m/s^2)

#Define height array
heights = np.arange(0, 50.5, 0.5) # heights in m
print(heights)


#Start with potential --> mgh
pot_energy = m*g*heights #definition of gravitational potential energy
print(pot_energy)

#Now, do kinetic --> 1/2*m*v^2
#So, we need to get v for every height --> use v^2 = v0^2 + 2*a*height

vel = -np.sqrt(2*g*(50-heights)) # these are velocities at every height - include negative for direction
#print(vel) 

kin_energy = 0.5*m*vel**2
print(kin_energy)
```

The arrays look how I'd expect. As the heights increase, the potential energies increase linearly. Conversely, as the heights and PE increase, the kinetic energy decreases. Also, the values make sense, as the KE at height = 50 m is the same as the PE at height = 0 m. Energy is being conserved as expected.
