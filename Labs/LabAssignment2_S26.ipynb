```python
meas_list = [] #assuming time in hours

for _ in range(10):
    meas_list.append(float(input('Enter a number between 1 and 10.'))) #needs to be a float or an int because the default of input() yields a str

print('The length of the list is',len(meas_list))

avg = sum(meas_list)/len(meas_list)
maximum = max(meas_list)
minimum = min(meas_list)

print('The average time is',avg,'hours.')
print('The maximum time is',maximum,'hours.')
print('The minimum time is',minimum,'hours.')

v_obj = 55 #miles per hour
v_obj_m = v_obj*1610 #m per hour (approximately)


avg *= v_obj_m
maximum *= v_obj_m
minimum *= v_obj_m


print('The average distance is',avg,'meters.')
print('The maximum distance is',maximum,'meters.')
print('The minimum distance is',minimum,'meters.')


```

To ask the user for the times, I used a for loop that runs 10 times and the append() command to add to the empty list.

To calculate the average, I used the sum of the list divided by the length of the list, and I used the built-in max and min commands to produce the maximum and minimum values from the list.

Then, I used the approximation that 1 mi is 1.61 km to convert the speed in miles per hour to meters per hour. 
Finally, I simply multiplied the average, max, and min values by the speed to yield the distance travelled in those times.

#### Calculate kinetic and potential energy without air resistance 

Use numpy arrays to complete the following:
- Create a 1D array of heights ranging from 0 to 50 m counting by 0.5 m
- Define the mass of the object 
- If the object is at rest at 50 m, calculate two arrays of the kinetic and potential energy of the object at each height. Ignore air resistance. 
- print the height array, kinetic energy array, and potential energy array 
    - do the arrays look like you would expect?

```python

heights = np.arange(0,50.1,0.5,float) #stops at 50.1 to include 50
print(heights) #in meters
print()

m = 5 #object mass in kg


pe = np.copy(heights) #potential energy in J
pe = pe*m*9.81 #multiplies each entry by the mass and 9.81 m/s2 via the formula U = mgh
print(pe)
print()

Etot = max(pe) #maximum potential energy is the same as mechanical energy, and will be the maximum kinetic energy as well


ke = (pe*-1)+Etot #kinetic energy in J


print(ke)


```

First, I used the np.arange() command to produce increasing heights from 0 to 50 m. I then defined an arbitrary object mass of 5 kg.

Then, I used the formula of U = mgh to produce an array of potential energies. As expected, the potential energy at [0] is zero, and the potential energy at the maximum height is the maximum PE. 

Since the object is at rest at the maximum height, the maximum PE is the total mechanical energy E of the object. So, I know that max(pe) is also max(ke), where the maximum kinetic energy is at height 0. So to translate from pe to ke, I used the formula of E - U = K. As expected, the kinetic energy at the maximum height is 0, and the kinetic energy at [0] is equal to the maximum potential energy.
