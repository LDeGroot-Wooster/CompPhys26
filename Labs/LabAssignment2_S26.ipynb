### Write a script 1 (25 pts) - Lists 

Let's build and analyze a list of measurements:
- Create an empty list called "meas_list"
- Ask the user 10 times for a number between 1 and 10 & append to the list 
    - Print the length of the list to make sure it is correct! 
- Calculate the average, maximum, and minimum of the list & print these values
- Assuming all of these are values of time, convert the values into distance (in meters) by multiplying by the constant velocity the object is traveling (55 mph) & print your results

When your program is complete, add a markdown cell below it to explain your process and program. 

```python
meas_list = []
counter = 0
new_num = 0
#Asks for 10 inputs
while counter < 10:
    new_num = int(input("Input a number between 1 and 10"))
    meas_list.append(new_num)
    counter += 1
#Prints list statistics
print(len(meas_list))
avg_meas = sum(meas_list)/len(meas_list)
max_meas = max(meas_list)
min_meas = min(meas_list)
print(avg_meas)
print(max_meas)
print(min_meas)
#Convert mph to m/s
meters_per_mile = 1609.34
seconds_per_hour = 3600
speed_ms = 55 * meters_per_mile / seconds_per_hour
#Calculate and print distances
avg_dist = avg_meas * speed_ms
max_dist = max_meas * speed_ms
min_dist = min_meas *speed_ms
print("Average distance is", round(avg_dist,2), "m")
print("Maximum distance is", round(max_dist,2), "m")
print("Minimum distance is", round(min_dist,2), "m")
```

I used a while list with an increasing counter to ask for 10 values between 1 and 10. I then print the length of the list and it's average, maximum, and minimum using built in functions. I then convert 55 mph to m/s and calculate and print the corresponding distances. When printing the distances I rounded to two decimal places.

#### Calculate kinetic and potential energy without air resistance 

Use numpy arrays to complete the following:
- Create a 1D array of heights ranging from 0 to 50 m counting by 0.5 m
- Define the mass of the object 
- If the object is at rest at 50 m, calculate two arrays of the kinetic and potential energy of the object at each height. Ignore air resistance. 
- print the height array, kinetic energy array, and potential energy array 
    - do the arrays look like you would expect?

```python
heights = np.arange(0,50.5,0.5,float) #0 to 50 counting by 0.5 inclusive
mass = 5
g = 9.8
v=np.sqrt(2*g*(50-heights)) #calculates velocity using v^2 = v0^2 + 2ax
T = (1/2)*mass*(v**2) #calculates KE using T = (1/2)mv^2
U = mass*g*heights #calculates PE using U = mgh
print(heights)
print(U)
print(T)
```

These arrays look like I would expect. Potential energy starts at 0 and increases, kinetic energy starts high and decreases to 0. At all points T+U = constant.
