```python
meas_list=[]
counter = 0
while counter < 10:
    temp = [int(input("Please input a time from 1-10 hours."))]
    meas_list=meas_list + temp
    counter = counter + 1
print(meas_list)
print("The length is", len(meas_list))
total = sum(meas_list)
average=total/len(meas_list)
print("The average is",average)
print("The highest value is", max(meas_list))
print("The lowest value is", min(meas_list))
distance = (np.array(meas_list))*55*1609.34
print("The distances traveled for each of these times of an object moving at 55 mph are:", distance)
```

We start with an empty list to append values to. We also introduce a counter to memory so we can track how many things are within the list. The list is then appended to as many times as the counter incraments (in this case it goes to 10). The average is taken with the sum of the list divided by its length, and the min and max are calculated using other functions native to python. The distance is calculated by converting the list to an array for ease of calculation and scaling the whole list by 55x1609.34.
